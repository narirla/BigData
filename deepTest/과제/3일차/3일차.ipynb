```python
import tensorflow as tf
from keras.optimizers import Adam
from keras import Sequential, Model
from keras.layers import Dense, Input
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sb
from scikeras.wrappers import KerasClassifier, KerasRegressor # л¶„лҘҳ KerasClassifier , нҡҢк·Җ(м—°мҶҚк°’) KerasRegressor
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score

import matplotlib
matplotlib.rcParams['font.family']='Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_csv('../../data/pima-indians-diabetes.data.csv')
df.head(3)
```

```python
x_data = df.iloc[:,:-1]
y_data = df.iloc[:, -1]
```

```python
x_data.shape
```

```python
y_data
```

```python
y_data.shape
```

---

train, test лӮҳлҲ„кё°

```python
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.2, random_state=42, stratify=y_data)
```

---

м •к·ңнҷ”

```python
scaler = StandardScaler()
x_train_s = scaler.fit_transform(x_train)
x_test_s = scaler.transform(x_test)
```

```python
x_train_s.shape
```

```python
y_train.shape
```

---

л”Ҙлҹ¬лӢқ л°Ҹ нҢҢмқҙн”„лқјмқё

```python
# dense л‘җлІҲмңјлЎң мӨ„мқҙкі  4X1 , л°ҳліө 30нҡҢлЎң мӨ„мқҙкё°
def get_model():
    myin = Input(shape=(8,))
    dense1 = Dense(units=32, activation='relu')
    dense2 = Dense(units=16, activation='relu')
    dense3 = Dense(units=1, activation='sigmoid')
    model = Sequential([myin, dense1, dense2, dense3])
    model.compile(loss='binary_crossentropy', metrics=['acc'])
    return model
```

```python
mode_sci = KerasClassifier(model=get_model, epochs=500) # н•ҷмҠө мӢң мЈјлҠ” мқёмһҗлҠ” KerasClassifier() м•Ҳм—җ л„Јм–ҙ мЈјкё°
pipe = Pipeline([('scaler', StandardScaler()),
                ('keras', mode_sci)])
pipe.fit(x_train_s, y_train)
```

```python
pred_tr = (pipe.predict(x_train_s) > 0.5) + 0
pred_tr
```

```python
pred_ts = (pipe.predict(x_test_s) > 0.5) + 0
pred_ts
```

---

accuracy_score, f1_score

```python
print('accuracy_score(Train)',accuracy_score(y_train, pred_tr))
print('accuracy_score(Test)',accuracy_score(y_test, pred_ts))
```

```python
print('f1_score(Train)',f1_score(y_train, pred_tr))
print('f1_score(Test)',f1_score(y_test, pred_ts))
```

---

confusion_matrix

```python
c_tr = confusion_matrix(y_train, pred_tr)
c_tr
```

```python
c_ts = confusion_matrix(y_test, pred_ts)
c_ts
```
