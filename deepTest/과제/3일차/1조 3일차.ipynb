```python
import numpy as np
import pandas as pd
from keras.optimizers import Adam
from keras import Sequential
from keras.layers import Dense, Input
from scikeras.wrappers import KerasClassifier
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/pima-indians-diabetes.data.csv')
df.head(3)
```

```python
x = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

---
лҚ°мқҙн„°л¶„н• 

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,train_size=0.7, stratify=y)
```

---
м •к·ңнҷ” л°Ҹ л”Ҙлҹ¬лӢқ нҢҢмқҙн”„лқјмқё

```python
def get_model():
    myin = Input(shape=(8,))
    dense1 = Dense( units=100, activation='relu')
    dense2 = Dense( units=50,  activation='relu')
    dense3 = Dense(units=1, activation='sigmoid')
    model = Sequential([myin, dense1,dense2,dense3])
    model.compile(loss='binary_crossentropy',optimizer=Adam(0.01),metrics=['acc'] )
    return model
```

```python
model_sci = KerasClassifier( model=get_model, epochs=300)
pipe = Pipeline( [('scaler', StandardScaler() ), ('keras', model_sci) ] )
pipe.fit( x_train,y_train)
```

---
train data


```python
pred = pipe.predict(x_train)
accuracy_score(y_train,pred)
```

```python
cm = confusion_matrix(y_train, pred)
print(cm)
f1 = f1_score(y_train,pred)
print(f1)
```

---
test data

```python
pred2 = pipe.predict(x_test)
accuracy_score(y_test,pred2)
```

```python
cm2 = confusion_matrix(y_test, pred2)
print(cm2)
f2 = f1_score(y_test, pred2)
print(f2)
```
