```python
import tensorflow as tf
from keras.optimizers import Adam
from keras import Sequential
from keras.layers import Dense
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
x1 = tf.constant([1,0,3,4,5], tf.float32) # 공부한 시간
x2 = tf.constant([0,1,2,3,4], tf.float32) # 출석일수
y = tf.constant([50,60,70,80,90], tf.float32) # 점수
```

```python
w1 = tf.Variable( tf.random.uniform([1]) )
w2 = tf.Variable( tf.random.uniform([1]) )
b = tf.Variable( tf.random.uniform([1]) )
```

```python
lr = 0.01
for i in range(2000):
    with tf.GradientTape() as tape: # 미분 대상 함수 with로 지정
        hx = w1*x1 + w2*x2 + b
        cost = tf.reduce_mean( (hx-y)**2 )
    
    w1_grad, w2_grad, b_grad = tape.gradient( cost, [w1, w2, b] )
    w1.assign_sub(lr*w1_grad)
    w2.assign_sub(lr*w2_grad)
    b.assign_sub(lr*b_grad)
    print( 'cost', cost.numpy()) #y축

print('최종w1', w1.numpy()) # 최종w1 [8.71829e-05]
print('최종w2', w2.numpy()) # 최종w2 [9.999976]
print('최종b', b.numpy()) # 최종b [49.99973]
```

### 행렬곱

```python
# x1 = tf.constant([1,0,3,4,5], tf.float32) # 공부한 시간
# x2 = tf.constant([0,1,2,3,4], tf.float32) # 출석일수
x = tf.constant( [ [1,0],[0,1],[3,2],[4,3],[5,4] ] , tf.float32)
y = tf.constant([50,60,70,80,90], tf.float32) # 점수
```

```python
# w1 = tf.Variable( tf.random.uniform([1]) )
# w2 = tf.Variable( tf.random.uniform([1]) )
w = tf.Variable(tf.random.uniform([2,1])) # [특정 데이터 갯수, 라벨의 갯수]
b = tf.Variable( tf.random.uniform([1]) ) # [라벨의 갯수]
```

```python
lr = 0.01
for i in range(2000):
    with tf.GradientTape() as tape: # 미분 대상 함수 with로 지정
        # hx = w1*x1 + w2*x2 + b
        hx = tf.matmul(x,w) + b
        cost = tf.reduce_mean( (hx-y)**2 )
    
    # w1_grad, w2_grad, b_grad = tape.gradient( cost, [w1, w2, b] )
    w_grad, b_grad = tape.gradient( cost, [w, b] )
    w.assign_sub(lr*w_grad)
    b.assign_sub(lr*b_grad)
    print( 'cost', cost.numpy()) #y축 cost 200.0

print('최종w', w.numpy()) #최종w [[-0.00113996]
                        #  [ 0.00173019]]
print('최종b', b.numpy()) # 최종b [69.99931]
```
