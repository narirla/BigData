```python
import tensorflow as tf
import numpy as np
```

```python
x = tf.Variable( 4.0 )
with tf.GradientTape() as tape: # with: 기록 구간 지정
    y = x**2 

dy_dx = tape.gradient( y, x) # y를 x에 대해서 미분
dy_dx
```

```python
x = tf.constant( [1,2,3,4,5], tf.float32)
y = tf.constant( [3,5,7,9,11], tf.float32)
```

```python
w = tf.Variable( 10.0 , dtype=tf.float32)
b = tf.Variable( 10.0 , dtype=tf.float32)
```

```python
lr = 0.01
for i in range(2000):
    with tf.GradientTape() as tape: # 미분 대상 함수 with로 지정
        hx = w*x + b
        cost = tf.reduce_mean( (hx-y)**2 )
    
    w_grad, b_grad = tape.gradient( cost, [w,b] )
    w.assign_sub(lr*w_grad)
    b.assign_sub(lr*b_grad)
    print( 'cost', cost.numpy()) #y축

print('최종w', w.numpy()) # 최종w 1.999994
print('최종b', b.numpy()) # 최종b 1.0000207
```
