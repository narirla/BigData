```python
x_trainR = x_train.reshape( 60000, 28*28)
x_trainR = x_trainR/255
```

```python
y_trainOne = to_categorical( y_train)
y_trainOne
```

```python
y_trainOne.shape
```

```python
#dense1 = Dense( units=10, input_dim=784 ) 
model = Sequential()
model.add( Dense(units=64, input_dim=784, activation='relu') ) #w:784x64, b=64
model.add( Dense(units=32, input_dim=64, activation='relu') ) #w:64x32 ,b =32
model.add( Dense(units=10, input_dim=32, activation='softmax') ) #w: 32x10, b10
model.compile( loss='categorical_crossentropy', optimizer='adam',metrics=['acc'])
model.fit(x_trainR, y_trainOne, epochs=10)
```

```python
model = Sequential()
model.add( Flatten( input_shape=(28,28) ) ) # x = x.reshape(n,28*28)
model.add( Dense( units=64, activation='relu')) #w:784x64  z = matmul(x,w)+b hx=relu(z)
model.add( Dense( units=32, activation='relu')) #w: 64x32   z = matmul(hx, w) +b hx=relu(z)
model.add( Dense( units=10, activation='softmax')) #w: 32x10  z= matmul(hx,w)+b hx=softmax(z)
model.compile( loss='sparse_categorical_crossentropy', optimizer='adam',metrics=['acc'])
model.fit(x_train/255, y_train, epochs=10)
```
