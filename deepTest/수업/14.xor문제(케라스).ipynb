```python
import tensorflow as tf
from keras.optimizers import Adam
from keras import Sequential
from keras.layers import Dense, Input
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
x = np.array([[0,0], [0,1], [1,0], [1,1]])
y = np.array([[0],[1],[1],[0]])
```

```python
# deepн•ҳкІҢ мһЎм•„м„ң acc к°’ 1.00мңјлЎң м¶ңл ҘлҗЁ
dense1 = Dense(units=100, input_dim=2, activation='sigmoid') # 2X100 b:100
dense2 = Dense(units=1, input_dim=100, activation='sigmoid') # 100X1 b:1
model = Sequential( [dense1, dense2] )
model.compile(loss='binary_crossentropy', optimizer=Adam(0.1), metrics=['acc']) # acc : м •нҷ•лҸ„лҘј м¶ңл Ҙн•ҙмӨҢ
h = model.fit(x, y, epochs=500)
```

```python
# forward м—°мӮ°
# 4X2 2X100 = 4X100
hx1 = dense1(x) # z1 = matmul(x, w1) + b1  hx1 = sigmoid(z1)
# 4X100 100X1 = 4X1
hx = dense2(hx1) # z = matmul(hx1, w2) + b2  hx = sigmoid(z)
hx
```

```python
model.predict(x)
```

```python
(model.predict(x) > 0.5) + 0 
```

```python
# deepн•ҳкІҢ мһЎм•„м„ң acc к°’ 1.00мңјлЎң м¶ңл ҘлҗЁ
dense1 = Dense(units=32, input_dim=2, activation='sigmoid')  # 2X32 b:32
dense2 = Dense(units=16, input_dim=32, activation='sigmoid') # 32X16 b:16
dense3 = Dense(units=1, input_dim=16, activation='sigmoid')  # 16X1 b:1
model = Sequential( [dense1, dense2, dense3] )
model.compile(loss='binary_crossentropy', optimizer=Adam(0.1), metrics=['acc']) # acc : м •нҷ•лҸ„лҘј м¶ңл Ҙн•ҙмӨҢ
h = model.fit(x, y, epochs=500)
```

```python
# Input лӮҙмһҘн•ЁмҲҳлҘј нҶөн•ҙ input_dim мғқлһө к°ҖлҠҘ (к¶ҢмһҘмӮ¬н•ӯ)
myin = Input(shape=(2,)) # мһ…л Ҙмёө: Input layer
dense1 = Dense(units=32, activation='sigmoid') # 2X32  b:32  # hidden layer
dense2 = Dense(units=16, activation='sigmoid') # 32X16 b:16  # hidden layer
dense3 = Dense(units=1,  activation='sigmoid') # 16X1  b:1   # output layer
model = Sequential( [myin, dense1, dense2, dense3] )
model.compile(loss='binary_crossentropy', optimizer=Adam(0.1), metrics=['acc']) # acc : м •нҷ•лҸ„лҘј м¶ңл Ҙн•ҙмӨҢ
h = model.fit(x, y, epochs=500)
```

```python
# 4X2 2X32 = 4X32
hx1 = dense1(x)   # z1 = matmul(x, w1) + b1    hx1 = sigmoid(z1)
# 4X32 32X16 = 4X16
hx2 = dense2(hx1) # z2 = matmul(hx1, w2) + b2  hx2 = sigmoid(z2)
# 4X16 16X1 = 4X1
hx = dense3(hx2)  # z = matmul(hx2, w3) + b3   hx = sigmoid(z)
hx
```

```python
model.predict(x)
```

```python
(model.predict(x) > 0.5) + 0 
```

### hidden layer : ReRU

```python
myin = Input(shape=(2,)) # мһ…л Ҙмёө: Input layer
dense1 = Dense(units=32, activation='relu')    # 2X32  b:32  # hidden layer
dense2 = Dense(units=16, activation='relu')    # 32X16 b:16  # hidden layer
dense3 = Dense(units=1,  activation='sigmoid') # 16X1  b:1   # output layer
model = Sequential( [myin, dense1, dense2, dense3] )
model.compile(loss='binary_crossentropy', optimizer=Adam(0.1), metrics=['acc']) # acc : м •нҷ•лҸ„лҘј м¶ңл Ҙн•ҙмӨҢ
h = model.fit(x, y, epochs=100)
```

```python
### нҖҙмҰҲ: iris dataset л”Ҙлҹ¬лӢқмңјлЎң кө¬м„ұн•ҙм„ң train, test мқҳ м •нҷ•лҸ„ нҷ•мқё (м •к·ңнҷ”X)
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from keras.utils import to_categorical
from sklearn.metrics import accuracy_score
```

```python
iris = load_iris(as_frame=True)
iris['frame']
```

```python
x = iris['data'].values
y = to_categorical(iris['target'])
```

```python
x.shape
```

```python
y.shape
```

```python
myin = Input(shape=(4,)) # мһ…л Ҙмёө: Input layer
dense1 = Dense(units=32, activation='relu')    # hidden layer
dense2 = Dense(units=16, activation='relu')    # hidden layer
dense3 = Dense(units=3,  activation='softmax') # output layer
model = Sequential( [myin, dense1, dense2, dense3] )
model.compile(loss='categorical_crossentropy', optimizer=Adam(0.01), metrics=['acc']) # acc : м •нҷ•лҸ„лҘј м¶ңл Ҙн•ҙмӨҢ
h = model.fit(x, y, epochs=100)
```

```python
y_pred = model.predict(x).argmax(axis=1)
y_pred
```

```python
accuracy_score(y.argmax(axis=1), y_pred)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42, stratify=y)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
myin = Input(shape=(4,)) # мһ…л Ҙмёө: Input layer
dense1 = Dense(units=32, activation='relu')    # hidden layer
dense2 = Dense(units=16, activation='relu')    # hidden layer
dense3 = Dense(units=3,  activation='softmax') # output layer
model = Sequential( [myin, dense1, dense2, dense3] )
model.compile(loss='categorical_crossentropy', optimizer=Adam(0.01), metrics=['acc']) # acc : м •нҷ•лҸ„лҘј м¶ңл Ҙн•ҙмӨҢ
h = model.fit(x_train, y_train, epochs=500)
```

```python
pred = model.predict(x_train).argmax(axis=1)
pred
```

```python
accuracy_score(y_train.argmax(axis=1), pred)
```

```python
pred_t = model.predict(x_test).argmax(axis=1)
pred_t
```

```python
accuracy_score(y_test.argmax(axis=1), pred_t)
```
