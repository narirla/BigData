```python
import tensorflow as tf
from keras.optimizers import Adam
from keras import Sequential
from keras.layers import Dense
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
x = tf.constant([[1,3],[2,2],[3,1],[4,6],[5,5],[6,4]], tf.float32) # [공부한 시간, 출석일수]
y = tf.constant([[0],[0],[0],[1],[1],[1]], tf.float32)
```

```python
initX = tf.initializers.GlorotUniform() # 초기 값을 랜덤하게 잘 부여하도록하는 알고리즘
w = tf.Variable( initX( shape=[2,1] ) ) # 랜덤하게 2행1열, [특정데이터 갯수, 라벨의 갯수]
b = tf.Variable( initX( shape=[1] ) ) # [라벨의 갯수]
```

```python
lr = 0.01
optimizer = tf.keras.optimizers.Adam(learning_rate=lr)
for i in range(2000):
    with tf.GradientTape() as tape: # 미분 대상 함수 with로 지정
        z = tf.matmul(x,w) + b
        cost_i = tf.nn.sigmoid_cross_entropy_with_logits(labels=y, logits= z )
        cost = tf.reduce_mean( cost_i)
    
    gradients = tape.gradient( cost, [w,b] )
    optimizer.apply_gradients(zip(gradients, [w,b]))
    # w.assign_sub(lr*w_grad)
    # b.assign_sub(lr*b_grad)
    print( 'cost', cost.numpy()) #y축

print('최종w', w.numpy()) #최종w [[1.0458604]
                        #  [1.0469794]]
print('최종b', b.numpy()) # 최종b [-6.872084]
```
