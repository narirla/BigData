```python
import tensorflow as tf
from keras.optimizers import Adam
from keras import Sequential, Model
from keras.layers import Dense, Input
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
from scikeras.wrappers import KerasClassifier, KerasRegressor # 분류 KerasClassifier , 회귀 KerasRegressor
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
```

```python
iris = load_iris()
x,y = iris.data, iris.target
```

```python
def get_model():
    myin = Input(shape=(4,))
    dense1 = Dense(units=100, activation='relu')
    dense2 = Dense(units=3, activation='softmax')
    model = Sequential([myin, dense1,dense2])
    model.compile(loss='sparse_categorical_crossentropy', metrics=['acc'])
    return model
```

```python
mode_sci = KerasClassifier(model=get_model, epochs=100) # 학습 시 주는 인자는 KerasClassifier() 안에 넣어 주기
pipe = Pipeline([('scaler', StandardScaler()),
                ('keras', mode_sci)])
pipe.fit(x,y)
```

```python
pipe.predict(x)
```

```python
pipe.predict_proba(x) # predict_proba : 확률값으로 보여줌
```
