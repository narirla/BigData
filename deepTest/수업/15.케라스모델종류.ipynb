```python
import tensorflow as tf
from keras.optimizers import Adam
from keras import Sequential, Model
from keras.layers import Dense, Input
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
x = np.array([[0,0], [0,1], [1,0], [1,1]], np.float32)
y = np.array([[0],[1],[1],[0]], np.float32)
```

### model 1

```python
# deepн•ҳкІҢ мһЎм•„м„ң acc к°’ 1.00мңјлЎң м¶ңл ҘлҗЁ
myin = Input(shape=(2,))
dense1 = Dense(units=100, activation='relu')   # 2X100 b:100
dense2 = Dense(units=1,  activation='sigmoid') # 100X1 b:1
model = Sequential( [dense1, dense2] )
model.compile(loss='binary_crossentropy', optimizer=Adam(0.1), metrics=['acc']) # acc : м •нҷ•лҸ„лҘј м¶ңл Ҙн•ҙмӨҢ
h = model.fit(x, y, epochs=100)
```

### model 2 (мӮ¬мҡ© еӨҡ)

```python
model = Sequential()
model.add(Input(shape=(2,)))
model.add( Dense(units=100, activation='relu'))
model.add( Dense(units=1,   activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer=Adam(0.1), metrics=['acc'])
h = model.fit(x, y, epochs=100)
```

### model 3

```python
x_train = np.array([[0,0], [0,1], [1,0], [1,1]], np.float32)
y_train = np.array([[0],[1],[1],[0]], np.float32)

myin = Input(shape=(2,))
z = Dense(units=100, activation='relu')(myin) # (myin) -> __call__ лҸҷмқј
hx = Dense(units=1,  activation='sigmoid')(z)

model = Model(inputs=myin, outputs=hx)
model.compile(loss='binary_crossentropy', optimizer=Adam(0.1), metrics=['acc'])
h = model.fit(x_train, y_train, epochs=100)
```

### model 4 (pytorch мӮ¬мҡ© еӨҡ)

```python
# Modelмқ„ мғҒмҶҚ л°ӣлҠ” нҒҙлһҳмҠӨ мғқм„ұ
class MyXOR(Model):
    def __init__(self):
        super(MyXOR, self).__init__() # Model мғқм„ұмһҗ нҳём¶ң

        self.dense1 = Dense(units=100, activation='relu')
        self.dense2 = Dense(units=1, activation='sigmoid')

    def call(self, inputs):
        hx1 = self.dense1(inputs)
        return self.dense2(hx1)
    
model = MyXOR()
model.compile(loss='binary_crossentropy', optimizer=Adam(0.1), metrics=['acc'])
h = model.fit(x_train, y_train, epochs=100)
```

```python
model.predict(x_train) # model.call(x_train)
```

```python
from sklearn.datasets import load_iris
from keras.utils import to_categorical
from sklearn.metrics import accuracy_score

iris = load_iris()
x = iris['data']
y = to_categorical(iris['target'])
```

```python
# Modelмқ„ мғҒмҶҚ л°ӣлҠ” нҒҙлһҳмҠӨ мғқм„ұ
class MyIris(Model):
    def __init__(self):
        super(MyIris, self).__init__() # Model мғқм„ұмһҗ нҳём¶ң

        self.dense1 = Dense(units=32, activation='relu')
        self.dense2 = Dense(units=16, activation='relu')
        self.dense3 = Dense(units=3, activation='softmax')

    def call(self, inputs):
        hx1 = self.dense1(inputs)
        hx2 = self.dense2(hx1)
        return self.dense3(hx2)
    
model = MyIris()
model.compile(loss='categorical_crossentropy', optimizer=Adam(0.01), metrics=['acc'])
h = model.fit(x, y, epochs=500)
```

```python
pred = model.predict(x).argmax(axis=1)
# accuracy_score(y.argmax(axis=1), pred)
accuracy_score(iris['target'], pred)
```
