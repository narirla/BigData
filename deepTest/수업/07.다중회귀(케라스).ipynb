```python
import tensorflow as tf
from keras.optimizers import Adam
from keras import Sequential
from keras.layers import Dense
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('data/test.csv', header=None)
df.columns = ['q1', 'q2', 'mid', 'final']
df
```

```python
x = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
x.shape
```

```python
# 1. hx = matmul(x,w) + b
# 2. w[3,1], b[1]
dense = Dense(units=1, input_dim=3)
print(dense(x)) #  hx = matmul(x,w) + b
w, b = dense.get_weights() # н•ҷмҠөлҗҳм§Җ м•ҠмқҖ мғҒнғңм—җм„ң мҳҲмёЎ к°’ м¶”м¶ңн•ҙ лҙ„
print(w)
print(b)
```

```python
dense = Dense(units=1, input_dim=3)
model = Sequential( [dense] ) # denseмқҳ wмҷҖ bк°Җ н•ҷмҠө лҢҖмғҒ
model.compile( loss='mse', optimizer=Adam(0.1))
model.fit(x, y, epochs=500)
```

```python
w, b = dense.get_weights() # н•ҷмҠөлҗң мғҒнғң
print(w)
print(b)
```

```python
dense(x) # tensor к°қмІҙлқј np.arrayлЎң мЈјм–ҙм•јн•Ё
```

```python
model.predict(x) # 1. кё°ліё м „мІҳлҰ¬ 2. dense(x) 3. кІ°кіјлҘј numpyлЎң м¶ңл Ҙ
```
