```python
import tensorflow as tf
from keras.optimizers import Adam
from keras import Sequential
from keras.layers import Dense
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
x_data = np.array([[0,0], [0,1], [1,0], [1,1]])
y_data = np.array([[0],[1],[1],[0]])
```

```python
x = tf.constant(x_data, tf.float32)
y = tf.constant(y_data, tf.float32)
```

```python
# 더 deep하게 늘림
initX = tf.initializers.GlorotUniform()
w1 = tf.Variable( initX(shape=[2,32]) ) #[2, ] 고정 2X1
b1 = tf.Variable( initX(shape=[32]) )
w2 = tf.Variable( initX(shape=[32,16]) )
b2 = tf.Variable( initX(shape=[16]) )
w3 = tf.Variable( initX(shape=[16,1]) ) #[ ,1] 고정 특성 데이터 2X1 , 정해진 건 없지만 보통 16배수 多 줌
b3 = tf.Variable( initX(shape=[1]) )
```

```python
lr = 0.01
optimizer = tf.keras.optimizers.Adam(learning_rate=lr)
for i in range(500):
    with tf.GradientTape() as tape: # 미분 대상 함수 with로 지정
        z1 = tf.matmul(x, w1) + b1
        hx1 = tf.nn.sigmoid(z1)
        z2 = tf.matmul(hx1, w2) + b2
        hx2 = tf.nn.sigmoid(z2)
        z = tf.matmul(hx2, w3) + b3
        
        cost_i = tf.nn.sigmoid_cross_entropy_with_logits(labels=y, logits= z )
        cost = tf.reduce_mean( cost_i)
    
    gradients = tape.gradient( cost, [w1, w2, w3, b1, b2, b3] )
    optimizer.apply_gradients(zip(gradients, [w1, w2, w3, b1, b2, b3]))
    print( 'cost', cost.numpy()) #y축

# print('최종w1', w1.numpy()) 
# print('최종w2', w2.numpy()) 
# print('최종w3', w2.numpy()) 
# print('최종b1', b1.numpy()) 
# print('최종b2', b2.numpy()) 
# print('최종b3', b3.numpy())  
```

```python
def hxFn(xd):
    xd = tf.constant(xd, tf.float32)
    z1 = tf.matmul(xd, w1) + b1
    hx1 = tf.sigmoid(z1)
    z2 = tf.matmul(hx1, w2) + b2
    hx2 = tf.sigmoid(z2)
    z = tf.matmul(hx2, w3) + b3
    hx = tf.sigmoid(z)
    return (hx.numpy() > 0.5) + 0
```
