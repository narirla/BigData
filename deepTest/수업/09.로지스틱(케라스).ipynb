```python
import tensorflow as tf
from keras.optimizers import Adam
from keras import Sequential
from keras.layers import Dense
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
x = np.array([[1,3],[2,2],[3,1],[4,6],[5,5],[6,4]], np.float32) # [кіөл¶Җн•ң мӢңк°„, м¶ңм„қмқјмҲҳ]
y = np.array([[0],[0],[0],[1],[1],[1]], np.float32)
```

## 1.forward
z = tf.matmul( x, w) + b

hx

## 2. w: 2X1 , b:1 

```python
dense = Dense( units=1, input_dim=2, activation='sigmoid')
dense( x )
```

```python
w , b = dense.get_weights() # н•ҷмҠө лҗҳм§Җ м•ҠмқҖ мғҒнғң
print(w)
print(b)
```

```python
z = tf.matmul( x, w )
hx = tf.nn.sigmoid(z)
hx # dense( x ) к°’кіј лҸҷмқј
```

```python
dense = Dense(units=1, input_dim=2, activation='sigmoid')
model = Sequential( [dense] )
model.compile(loss='binary_crossentropy', optimizer=Adam(0.1), metrics=['acc']) # acc : м •нҷ•лҸ„лҘј м¶ңл Ҙн•ҙмӨҢ
h = model.fit(x, y, epochs=500)
```

```python
model.predict(x)
```

```python
w , b = dense.get_weights() # н•ҷмҠөлҗң мғҒнғң
print(w)
print(b)
```

### нҖҙмҰҲ : н”јл§Ҳмқёл””м–ё лҚ°мқҙн„°лҘј мјҖлқјмҠӨлЎң н•ҷмҠө нӣ„ м •нҷ•лҸ„, loss м°ЁнҠёлҘј к·ёлҰ¬мӢңмҳӨ
### нҠ№м„ұ лҚ°мқҙн„° 1лІҲм§ё н–үмқҳ мҳҲмёЎк°’

```python
df = pd.read_csv('data/pima-indians-diabetes.data.csv')
df.head()
```

```python
x = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
x.shape
```

```python
dense = Dense(units=1, input_dim=8, activation='sigmoid')
model = Sequential( [dense] ) # Sequential = мһ…л Ҙ вҶ’ мқҖлӢүмёө вҶ’ м¶ңл Ҙмёөмқ„ мқјл ¬(м§Ғм„ )лЎң м—°кІ°н•ң лӘЁлҚё
model.compile(loss='binary_crossentropy', optimizer=Adam(0.01), metrics=['acc']) # acc : м •нҷ•лҸ„лҘј м¶ңл Ҙн•ҙмӨҢ
h = model.fit(x, y, epochs=500)
```

```python
h.history.keys()
```

```python
plt.plot(h.history['acc']) # м •к·ңнҷ” н•„мҡ”
plt.show()
```

```python
plt.plot(h.history['loss'])
plt.show()
```
