```python
import tensorflow as tf
import numpy as np
```

```python
x = tf.constant( [1,2,3,4,5], tf.float32)
y = tf.constant( [3,5,7,9,11], tf.float32)
```

```python
w = tf.Variable( 10.0 , dtype=tf.float32)
b = tf.Variable( 10.0 , dtype=tf.float32)
```

### low level api

```python
lr = 0.01
optimizer = tf.keras.optimizers.SGD(learning_rate=lr)
for i in range(2000):
    with tf.GradientTape() as tape: # 미분 대상 함수 with로 지정
        hx = w*x + b
        cost = tf.reduce_mean( (hx-y)**2 )
    
    gradients = tape.gradient( cost, [w,b] )
    ## 계산된 변수 SGD를 적용해서 업데이트 해라
    optimizer.apply_gradients(zip(gradients, [w,b])) # 기본 미분에 옵티마이저 적용
    # w.assign_sub(lr*w_grad)
    # b.assign_sub(lr*b_grad)
    print( 'cost', cost.numpy()) #y축

print('최종w', w.numpy()) # 최종w 1.999994
print('최종b', b.numpy()) # 최종b 1.0000207
```

```python
w = tf.Variable( 10.0 , dtype=tf.float32)
b = tf.Variable( 10.0 , dtype=tf.float32)

lr = 0.1
optimizer = tf.keras.optimizers.Adam(learning_rate=lr)
for i in range(2000):
    with tf.GradientTape() as tape: # 미분 대상 함수 with로 지정
        hx = w*x + b
        cost = tf.reduce_mean( (hx-y)**2 )
    
    gradients = tape.gradient( cost, [w,b] )
    ## 계산된 변수 SGD를 적용해서 업데이트 해라
    optimizer.apply_gradients(zip(gradients, [w,b])) # 기본 미분에 옵티마이저 적용
    # w.assign_sub(lr*w_grad)
    # b.assign_sub(lr*b_grad)
    print( 'cost', cost.numpy()) #y축

print('최종w', w.numpy()) # 최종w 1.999994
print('최종b', b.numpy()) # 최종b 1.0000207
```
