### 사칙연산 : add(+), sub(-), multiply(*), div(/)
- broad casting
- element wise

```python
a = tf.constant(3)
b = tf.constant( [1,2,3] )
# c = tf.add(a,b)
# c = a+b
# c = tf.multiply(a,b)
c = a*b
c
```

```python
c.numpy()
```

```python
d = tf.add(a,b)
d
```

```python
e = c+d
e
```

```python
f = tf.constant([ 1, 2, 3] )
g = tf.constant([ [1, 2, 3], [4, 5, 6], [7, 8, 9],] )
h = f+g # tf.add(f,g)
h
```

### 행렬곱

```python
o = tf.constant([[1,2],[3,4]])
p = tf.constant([[5,6],[7,8]])
q = tf.multiply(o,p)
r = tf.matmul(o,p)
```

```python
q.numpy()
```

```python
r.numpy()
```

### 많이 사용하는 연산

```python
s = tf.constant([1,2,3])
# t = tf.square(s)
t = s**2
t
```

```python
u = tf.reduce_mean(s)
u
```

```python
v = tf.reduce_sum(s)
v
```

```python
v = tf.Variable(10) # Variable의 경우 값 변경 가능
v
```

```python
v.assign(20) # 10 -> 20 변경
v
```

```python
v.assign_add(2) # 20 + 2 
v
```

```python
v.assign_sub(3) # 22 - 3 , 가장 사용 多 
v
```
