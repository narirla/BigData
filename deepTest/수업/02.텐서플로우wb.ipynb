```python
import tensorflow as tf
import numpy as np
```

```python
x = tf.constant( [1,2,3,4,5], tf.float32)
y = tf.constant( [3,5,7,9,11], tf.float32)
```

```python
w = tf.Variable( 10.0 , dtype=tf.float32)
b = tf.Variable( 10.0 , dtype=tf.float32)
epochs = 3000 # 반복 횟수
lr = 0.01
n = len(x)
```

```python
for i in range(epochs):
    hx = w*x + b # tf.add( tf.multiply(w,x), b)
    c = tf.reduce_sum( (hx-y)**2)/n
    gradientW = tf.reduce_sum((w*x + b-y)*2*x)/n
    gradientB = tf.reduce_sum((w*x + b-y)*2)/n
    print('w=', w.numpy(), 'b=', b.numpy(),
            'w미분', gradientW.numpy(), 'b미분', gradientB.numpy())
    print('cost', c.numpy())
    w.assign_sub(lr*gradientW) # w = w - lr*미분
    b.assign_sub(lr*gradientB) # b = b - lr*미분
print('최종w', w.numpy(), '최종b', b.numpy())
```
