```python
import matplotlib.pyplot as plt
import numpy as np
```

### 다중

```python
x1 = np.array( [1,2,3,4,5] )#공부한시간
x2 = np.array( [0,1,2,3,4] )#출석일수
y = np.array( [60,70,80,90,100] )#점수
```

```python
w1 = 10
w2 = 10
b = 10
epochs=2000
learning_rate = 0.01
n = len( x1)
```

w1*x1 + w2*x2 + w3*x3 + ... wn*xn +b

```python
for i in range( epochs):
    hx = w1*x1+ w2*x2  + b
    cost = np.sum( (hx-y)**2)/n
    gradienW1 = np.sum((w1*x1+w2*x2+b-y )*2*x1 )/n
    gradienW2 = np.sum((w1*x1+w2*x2+b-y )*2*x2 )/n
    gradienB = np.sum((w1*x1+w2*x2+b-y )*2 )/n
    w1 = w1-learning_rate*gradienW1
    w2 = w2-learning_rate*gradienW2
    b = b-learning_rate*gradienB
    print('cost:',cost,'w1:',w1,'w2:',w2,'b:',b)

print('최종w1', w1,'최종w2', w2, '최종b', b)
```

### 단일선형 회귀

```python
x = np.array( [1,2,3,4,5] )
y = np.array( [3,5,7,9,11] )
```

```python
plt.ylim( (0,15) )
plt.xlim( (0,10) )
plt.scatter( x, y)
plt.show()
```

```python
w = 10
b = 10
epochs=2000
learning_rate = 0.01
n = len( x)
```

```python
for i in range( epochs):
    hx = w*x + b
    cost = np.sum( (hx-y)**2)/n
    gradienW = np.sum((w*x+b-y )*2*x )/n
    gradienB = np.sum((w*x+b-y )*2 )/n
    w = w-learning_rate*gradienW
    b = b-learning_rate*gradienB
    print('cost:',cost,'w:',w,'b:',b)

print('최종w', w, '최종b', b)
```
