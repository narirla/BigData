##  단순 선형 회귀 공식

기울기($w$)와 절편($b$)을 사용하여 예측값 $\hat{Y}$를 나타내는 공식:

$$
\hat{Y}_i = w X_i + b
$$

---

### 🔍 공식 요소 설명

| 기호 | 의미 | 설명 |
| :--- | :--- | :--- |
| $\hat{Y}_i$ | $i$번째 데이터의 **예측값** (Predicted Value) | 모델이 입력 $X_i$에 대해 예측한 종속 변수의 값. |
| $X_i$ | $i$번째 데이터의 **독립 변수** (Independent Variable) | 예측을 수행하는 데 사용되는 입력 데이터(특징)의 값입. |
| $w$ | **가중치/기울기** (Weight or Slope) | $X$가 한 단위 변할 때 $\hat{Y}$가 얼마나 변하는지를 나타냅니다. 모델 학습을 통해 결정되는 핵심 계수. |
| $b$ | **절편** (Bias or Intercept) | $X$가 0일 때 $\hat{Y}$의 값. 모델 학습을 통해 결정되는 값. |

---

**평균 제곱 오차(Mean Squared Error, MSE)** 공식
---

##  MSE(평균 제곱 오차) 공식

MSE는 회귀 분석에서 모델의 예측값과 실제값 사이의 **오차의 제곱에 대한 평균**을 나타내는 지표입니다.

$$
\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (Y_i - \hat{Y}_i)^2
$$

---

### 공식 요소 설명

| 기호 | 의미 | 설명 |
| :--- | :--- | :--- |
| $n$ | 데이터 포인트의 총 개수 | 전체 데이터 쌍의 수입니다. |
| $i$ | 개별 데이터 포인트의 인덱스 | 1부터 $n$까지의 순서를 나타냅니다. |
| $Y_i$ | $i$번째 데이터의 **실제값** (참값) | 모델이 예측하려고 시도하는 실제 관측값입니다. |
| $\hat{Y}_i$ | $i$번째 데이터의 **예측값** | 모델이 계산한 값입니다. |
| $(Y_i - \hat{Y}_i)^2$ | $i$번째 오차의 **제곱** | 실제값과 예측값의 차이(오차)를 제곱한 값입니다. 제곱을 통해 양수/음수 오차를 동일하게 처리하고 큰 오차에 더 큰 페널티를 부여합니다. |
| $\sum_{i=1}^{n}$ | 총합 (시그마) | $i=1$부터 $n$까지 모든 오차 제곱을 더합니다. |
| $\frac{1}{n}$ | 평균 | 총합을 데이터 개수 $n$으로 나누어 평균을 구합니다. |

---



```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_data = [1,2,3]
y_data = [1,2,3]
```

```python
for i in range(3): #[0,1,2]
    print(i)
```

```python
# x = [1,2,3]
# y = [1,2,3]
# w = 1
def mse( x, y, w ):
    c = 0
    for i in range( len(x)):
        hx = w*x[i] # hx = 2x3
        c = c+ (hx-y[i])**2 # 1+4+9
    return c/len(x)  # 1+4+9 /3
```

```python
print( mse( x_data, y_data, -1))
print( mse( x_data, y_data, 0))
print( mse( x_data, y_data, 1))
print( mse( x_data, y_data, 2))
print( mse( x_data, y_data, 3))
```
