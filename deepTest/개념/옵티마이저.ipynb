## 옵티마이저

<img src='https://velog.velcdn.com/images/ksolar03/post/07572402-4c7a-44e2-b16c-6ac54e5082b6/image.png'>

딥러닝에서 옵티마이저는 크게 **'방향'** 을 개선하는 계보(Momentum계)와 **'학습률(보폭)'** 을 개선하는 계보(Adagrad계)로 나뉘며, 이 둘을 합친 것이 현재 가장 대중적인 **Adam**

###  케라스 주요 옵티마이저 요약 표

| 구분 | 옵티마이저 | 핵심 아이디어 | 장점 (특징) | 단점 (한계) |
| --- | --- | --- | --- | --- |
| **기본** | **GD** (Batch) | 전체 데이터 미분 | 수렴이 매우 안정적임 | 데이터가 크면 계산이 너무 느림 |
|  | **SGD** | 무작위 샘플 미분 | 속도가 빠르고 메모리 절약 | 경로가 지그재그로 불안정함 |
| **방향 개선** | **Momentum** | 가속도(관성) 추가 | 로컬 미니마(웅덩이) 탈출 용이 | 관성 때문에 최적점을 지나칠 수 있음 |
|  | **NAG** | 앞을 미리보고 가속 | Momentum보다 안정적인 멈춤 | 계산이 다소 복잡함 |
| **보폭 개선** | **Adagrad** | 자주 변하는 축은 작게 | 피처별 맞춤형 학습 가능 | 학습이 길어지면 보폭이 0이 됨 |
|  | **RMSprop** | 지수 이동 평균 사용 | Adagrad의 조기 중단 문제 해결 | 하이퍼파라미터() 설정 필요 |
|  | **Adadelta** | Adagrad 보완 | 학습률 설정이 아예 필요 없음 | 가끔 학습이 너무 느려짐 |
| **통합** | **Adam** | **Momentum + RMSprop** | **가장 무난하고 성능이 좋음** | 특정 데이터에서 일반화 성능 저하 |
|  | **Nadam** | **NAG + RMSprop** | Adam보다 빠른 수렴 속도 | 하이퍼파라미터에 민감할 수 있음 |

---

###  주요 옵티마이저 상세 분석

#### 1. SGD (Stochastic Gradient Descent)

* **특징:** 매 스텝 전체 데이터가 아닌 1개(또는 배치 단위)의 데이터만 사용하여 가중치를 업데이트.
* **용도:** 데이터셋이 너무 커서 메모리에 다 올릴 수 없을 때 필수.

#### 2. Momentum (관성)

* **특징:** 경사 방향으로 힘을 받아 '가속도'가 붙는 물리적 원리를 도입.
* **효과:** 진행하던 방향으로 계속 가려는 성질 덕분에 '지그재그' 현상이 줄어들고 속도가 빠르다.

#### 3. RMSprop

* **특징:** 제프리 힌튼 교수가 제안한 방식으로, 최근의 기울기 위주로 학습률을 조정.
* **효과:** Adagrad가 학습을 너무 일찍 멈추는 단점을 해결하여, 순환 신경망(RNN) 등에서 특히 효과적.

#### 4. Adam (Adaptive Moment Estimation)

* **특징:** 현재 딥러닝에서 **가장 많이 쓰이는 "국룰" 옵티마이저**.
* **효과:** 방향(Momentum)과 보폭(RMSprop)을 모두 자동으로 조절해주기 때문에, 처음 모델을 만들 때 가장 먼저 시도하기 좋다.

###  어떤 것을 써야 할까

1. **처음 시작할 때:** 무조건 **`'adam'`** 을 추천. 대부분의 문제에서 준수한 성능을 보인다.
2. **이미지 처리(CNN) 등 정교한 튜닝이 필요할 때:** **`SGD(momentum=0.9)`** 를 사용하여 세밀하게 학습률을 깎아나가는 방식이 최종 정확도가 더 높게 나오기도 한다.
3. **자연어 처리나 RNN 계열:** **`RMSprop`** 이나 **`Adam`** 이 유리.
