```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
# x_data = np.array( [1,2,3,4,5,6], np.float32)
# y_data = np.array( [0,0,0,1,1,1], np.float32)
x_data = tf.constant( [1,2,3,4,5,6], tf.float32)
y_data = tf.constant( [0,0,0,1,1,1], tf.float32)
# y_data = np.array( [1,2,3,4,5,6])
```

```python
import math
def sigmoid(z):
    return 1/(1+math.e**-z)
```

```python
def cost( x, y, w):
    # z = w*x
    # hx = sigmoid(z)
    # c = np.mean((hx-y)**2)
    # c = tf.reduce_mean((hx-y)**2)
    z = w*x
    # hx = tf.nn.sigmoid( z ) # 0~1
    # cost_i = -( y*tf.math.log(hx) + (1-y)*(tf.math.log(1-hx)) ) # 시그모이드를 통해 그래프 펴주는 작업
    cost_i = tf.nn.sigmoid_cross_entropy_with_logits(labels=y, logits= z ) # 위 두가지 코드가 한번에 들어가 있음
    c = tf.reduce_mean( cost_i)
    return c
```

```python
for w in np.linspace( -3, 5, 50):
    c = cost( x_data,y_data, w )
    plt.plot( w, c, 'ro')
plt.show()
```
