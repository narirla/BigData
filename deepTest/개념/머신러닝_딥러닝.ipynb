## 딥러닝(Deep Learning)

 인간의 뇌 구조를 모방한 **인공신경망(Artificial Neural Networks)**을 통해 컴퓨터가 스스로 데이터를 학습하는 머신러닝의 한 분야입니다.

<img src='image/deep.jpg'>

---

## 1. 딥러닝의 핵심 구조: 인공신경망

딥러닝의 'Deep'은 신경망의 층(Layer)이 깊게 쌓였다는 뜻입니다. 기본적으로 세 가지 층으로 구성됩니다.

* **입력층 (Input Layer):** 데이터를 받아들이는 층.
* **은닉층 (Hidden Layer):** 데이터의 특징을 추출하고 계산이 일어나는 복잡한 층 (여러 층일수록 '딥'해집니다).
* **출력층 (Output Layer):** 최종 결과를 내보내는 층.

---

## 2. 어떻게 학습하나요? (핵심 원리)

딥러닝 모델이 정답을 맞춰가는 과정은 크게 두 단계로 나뉩니다.

1. **순전파 (Forward Propagation):** 데이터를 입력받아 가중치()와 편향()을 곱하며 결과값을 예측합니다.
2. **손실 함수 (Loss Function):** 예측값과 실제 정답 사이의 오차를 계산합니다.
3. **역전파 (Backpropagation):** 오차를 줄이기 위해 뒤에서부터 앞으로 가중치를 수정해 나갑니다. 이 과정에서 **경사하강법(Gradient Descent)**이라는 최적화 알고리즘이 사용됩니다.

---

## 3. 주요 알고리즘 유형

데이터의 특성에 따라 사용하는 신경망의 종류가 다릅니다.

| 종류 | 풀네임 | 주요 용도 |
| --- | --- | --- |
| **ANN** | 인공신경망 | 가장 기본적인 형태의 신경망 |
| **CNN** | 합성곱 신경망 | **이미지 인식**, 사물 검출 (시각 정보 처리) |
| **RNN** | 순환 신경망 | **텍스트, 음성**, 시계열 데이터 (순서가 중요한 데이터) |
| **Transformer** | 트랜스포머 | ChatGPT 같은 **생성형 AI**의 핵심 기술 |

---

## 4. 딥러닝이 강력한 이유

* **특징 자동 추출:** 기존 머신러닝은 사람이 직접 '이건 귀 모양이야'라고 특징을 알려줘야 했지만, 딥러닝은 데이터만 주면 스스로 중요한 특징을 찾아냅니다.
* **비정형 데이터 처리:** 이미지, 영상, 텍스트, 음성 등 복잡한 데이터를 처리하는 데 압도적인 성능을 보입니다.

---

## 5. 한계점

* **데이터 의존성:** 학습을 위해 엄청난 양의 데이터가 필요합니다.
* **연산 비용:** 고성능 GPU와 같은 하드웨어 자원이 많이 소모됩니다.
* **블랙박스 문제:** 결과가 왜 이렇게 나왔는지 내부 과정을 완벽히 설명하기 어렵습니다.

#  머신러닝 vs 딥러닝 기술 종류 비교표

| 구분         | 주요 기술/알고리즘                     | 설명 |
|--------------|----------------------------------------|------|
| 머신러닝 (ML) | **회귀 (Regression)**                 | 연속적인 수치를 예측하는 모델 (ex. 집값 예측) |
|              | **의사결정나무 (Decision Tree)**       | 분할을 통해 의사결정 규칙을 학습 |
|              | **랜덤 포레스트 (Random Forest)**      | 여러 개의 결정트리를 결합한 앙상블 모델 |
|              | **서포트 벡터 머신 (SVM)**             | 초평면을 통해 데이터를 분류 |
|              | **K-최근접 이웃 (K-NN)**               | 주변 데이터와의 거리로 분류/회귀 수행 |
|              | **나이브 베이즈 (Naive Bayes)**        | 조건부 확률 기반 분류 모델 |
|              | **K-평균 클러스터링 (K-Means)**        | 비지도 학습 기반 데이터 클러스터링 |
|              | **PCA (주성분 분석)**                  | 차원 축소 기법 |
|              | **Gradient Boosting (XGBoost 등)**     | 성능 좋은 앙상블 기법 중 하나 |




| 구분         | 주요 기술/알고리즘                     | 설명 |
|--------------|----------------------------------------|------|
| 딥러닝 (DL)   | **인공신경망 (ANN)**                   | 가장 기본적인 신경망 구조 |
|              | **합성곱 신경망 (CNN)**                | 이미지/영상 처리에 강력한 구조 |
|              | **순환 신경망 (RNN)**                  | 순차적 데이터 (시계열, 텍스트 등)에 적합 |
|              | **장단기 메모리 (LSTM)**              | RNN의 장기 의존성 문제를 해결 |
|              | **GRU (Gated Recurrent Unit)**        | LSTM보다 간단한 구조의 순환 신경망 |
|              | **변환기 (Transformer)**              | 병렬 처리에 강력한 자연어 처리 모델 구조 |
|              | **BERT, GPT, T5 등**                  | Transformer 기반 사전학습 언어모델 |
|              | **GAN (생성적 적대 신경망)**           | 데이터를 생성하는 딥러닝 모델 |
|              | **AutoEncoder**                       | 입력을 압축/복원하는 비지도 학습 구조 |
|              | **Deep Reinforcement Learning**       | 강화학습과 딥러닝의 결합 |


## 딥러닝 주요 모델 개념 비교

| 모델 | 풀네임 | 주요 아이디어 | 주로 쓰이는 분야 | 장점 | 단점 |
|------|--------|--------------|------------------|------|------|
| **CNN** | Convolutional Neural Network | 이미지나 2D 데이터에서 **공간적 패턴**을 합성곱(Convolution)으로 추출 | 이미지 분류, 객체 인식, 영상 분석 | 이미지 특징 추출에 강함, 파라미터 수 적음 | 순차 데이터 처리엔 부적합 |
| **RNN** | Recurrent Neural Network | **순차 데이터**를 처리하기 위해 이전 단계의 출력을 다음 단계로 전달 | 시계열 예측, 자연어 처리 | 과거 정보 반영 가능 | 긴 시퀀스에서 **기울기 소실/폭발** 문제 발생 |
| **LSTM** | Long Short-Term Memory | RNN의 한계(장기 의존성 문제)를 해결한 구조, **셀 상태**와 게이트로 정보 보존/제거 | 기계 번역, 음성 인식, 주가 예측 | 긴 문맥 유지 가능, 장기 의존성 처리 | 구조가 복잡하고 계산량 많음 |
| **GRU** | Gated Recurrent Unit | LSTM을 단순화한 RNN 변형, **업데이트/리셋 게이트** 사용 | 채팅봇, 시계열 예측, NLP | 학습 속도 빠르고 성능 유사 | 장기 의존성 처리 능력은 LSTM보다 살짝 약함 |
| **GAN** | Generative Adversarial Network | **생성자(Generator)**와 **판별자(Discriminator)**가 경쟁하며 데이터 생성 | 이미지 생성, 스타일 변환, 데이터 증강 | 매우 사실적인 데이터 생성 가능 | 학습 불안정, 모드 붕괴 문제 |
| **LLM** | Large Language Model | 대규모 텍스트 데이터로 학습한 **거대 신경망 언어 모델** | 챗봇, 요약, 번역, 코드 생성 | 다양한 언어 작업 가능, 범용성 높음 | 대규모 연산 자원 필요, 환각(hallucination) 가능성 |

 **요약**  
- **CNN** → 이미지 전문가   
- **RNN/LSTM/GRU** → 순차 데이터 전문가   
- **GAN** → 창작형 AI   
- **LLM** → 언어 전천후
