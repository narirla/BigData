# 딥러닝 라이브러리

딥러닝 라이브러리는 모델을 처음부터 직접 프로그래밍할 필요 없이, 복잡한 수학 연산과 신경망 구조를 쉽게 구현할 수 있도록 도와주는 도구 모음입니다.

2026년 현재 가장 널리 쓰이는 주요 라이브러리와 프레임워크
---

## 1. 3대 핵심 딥러닝 프레임워크

가장 대중적이고 생태계가 큰 세 가지입니다.

| 라이브러리 | 개발사 | 주요 특징 | 추천 대상 |
| --- | --- | --- | --- |
| **PyTorch** | Meta | 직관적이고 Python다운 코드(Pythonic). 동적 계산 그래프 지원으로 디버깅이 쉬움. | **연구자**, 대학생, 최신 논문 구현용 |
| **TensorFlow** | Google | 강력한 배포(Deployment) 기능과 확장성. 대규모 상용 서비스 운영에 최적화. | **기업 실무**, 대규모 인프라 운영 |
| **JAX** | Google | 고성능 수치 계산(TPU 최적화) 및 자동 미분에 특화. 속도가 매우 빠름. | 초고성능 모델 연구, 대규모 LLM 학습 |

---

## 2. 생산성을 높여주는 고수준 라이브러리

복잡한 코드를 줄여주고 핵심 로직에만 집중하게 해주는 도구들입니다.

* **Keras (케라스):** 텐서플로 위에서 동작하는 API로, "딥러닝의 레고"라고 불릴 만큼 쉽습니다. 초보자가 모델을 빠르게 프로토타이핑할 때 필수입니다.
* **PyTorch Lightning:** 파이토치의 복잡한 반복 코드(학습 루프 등)를 표준화하여 깔끔하게 정리해 줍니다. 연구 생산성을 극대화합니다.
* **fastai:** 파이토치 기반의 라이브러리로, 최소한의 코드로 최첨단(SOTA) 모델 성능을 낼 수 있도록 설계되었습니다.

---

## 3. 특정 분야 특화 라이브러리

특정 데이터를 다룰 때 유용한 도구들입니다.

* **Hugging Face (Transformers):** **자연어 처리(NLP)**의 표준입니다. ChatGPT 같은 거대언어모델(LLM)을 활용하거나 미세 조정(Fine-tuning)할 때 가장 많이 쓰입니다.
* **OpenCV / torchvision:** **컴퓨터 비전(이미지/영상)** 처리용 라이브러리입니다. 이미지 변환, 증강, 데이터 로딩에 필수적입니다.
* **scikit-learn:** 딥러닝 전단계인 **전통적 머신러닝**과 데이터 전처리에 가장 널리 쓰이는 기본 라이브러리입니다.
