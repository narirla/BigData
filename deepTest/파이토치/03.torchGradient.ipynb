```python
import torch
from torch.optim import Adam
import matplotlib.pyplot as plt
```

```python
x = torch.tensor( 2.0, requires_grad=True )
y = 2*x**2
y.backward() # 미분
print( x.grad)

y = 2*x**2
y.backward() # 미분 누적
print( x.grad)

y = 2*x**2
y.backward() # 미분 누적
print( x.grad)
```

```python
w = torch.tensor( 2.0, requires_grad=True) # 미분을 통해 수정되는 텐서객체
for epoch in range( 10):
    y = 2*w
    y.backward() #미분
    print( w.grad )
```

```python
x = torch.FloatTensor( [1,2,3,4,5] )
y = torch.FloatTensor( [2,4,6,8,10] )
```

```python
w = torch.tensor( 10.0, requires_grad=True )
b = torch.tensor( 10.0, requires_grad=True )
```

```python
optimizer = Adam( [w,b], lr=0.1 )
for i in range( 2000 ):
    hx = w*x + b
    cost = torch.mean( (hx-y)**2)
    optimizer.zero_grad() # 초기화
    cost.backward() # 미분 w,b
    optimizer.step() # w= w-0.1*w미분값, b = b-0.1xb미분값
    print(i, 'cost', cost.item() )
```
