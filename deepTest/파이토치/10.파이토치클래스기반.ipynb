```python
import torch
from torch.optim import Adam
from torch.nn import Linear, MSELoss, Sequential, BCELoss, Sigmoid, ReLU, Module
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x_data = np.array( [[0,0],[0,1],[1,0],[1,1]])
y_data = np.array( [[0],[1],[1],[0]])
```

```python
x = torch.FloatTensor(x_data)  # 공부한 시작,출석일수
y = torch.FloatTensor(y_data ) 
```

```python
class XORModel( Module ):
    def __init__(self):
        super().__init__()
        self.nn1 = Linear( 2,32)
        self.relu1 = ReLU()
        self.nn2 = Linear( 32,1 )
        self.sig1 = Sigmoid()

    # def __call__(self,x):
    #     return forward(x)
    def forward( self, x):
        print('forward call')
        x = self.nn1(x)
        x = self.relu1(x)
        x = self.nn2(x)
        x = self.sig1(x)
        return x
```

```python
model = XORModel()
# model(x) #model.__call__(x) 
# model.forward(x)
print( model.parameters() )
loss_fn = BCELoss()
optimizer = Adam( model.parameters(), lr=0.01)
for epoch in range(1000):
    optimizer.zero_grad()
    hx = model(x) #model.forward(x)
    cost = loss_fn( hx, y)
    cost.backward()
    optimizer.step()
    print( 'epoch', epoch, 'cost:', cost.item())
```
