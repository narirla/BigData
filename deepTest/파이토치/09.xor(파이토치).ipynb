```python
import torch
from torch.optim import Adam
from torch.nn import Linear, MSELoss, Sequential, BCELoss, Sigmoid, ReLU
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x_data = np.array( [[0,0],[0,1],[1,0],[1,1]])
y_data = np.array( [[0],[1],[1],[0]])
```

```python
x = torch.FloatTensor(x_data)  # 공부한 시작,출석일수
y = torch.FloatTensor(y_data ) #합격여부
```

```python
model = Sequential()
model.add_module( 'nn1', Linear( 2,32 )) 
model.add_module( 'relu1', ReLU())
model.add_module( 'nn2', Linear( 32,1 )) 
model.add_module( 'sig1', Sigmoid())

loss_fn = BCELoss()
optimizer = Adam( model.parameters(), lr=0.1)

for epoch in range( 3000 ):
    optimizer.zero_grad()
    # z1 = matmul( x, w1) + b1   hx1 = ReLU(z1)   z= matmul( hx1, w2)+b2  hx= Sigmoid(z)
    hx = model( x )  # 4x2 2x32 4x32 32x1 4x1
    pred = (hx.detach().numpy()>0.5) +0
    print( '정확도', (y_data==pred).mean() )
    cost = loss_fn( hx, y )
    cost.backward()
    optimizer.step()
    print( 'cost:', cost.item())
    print('==========')
```
