```python
import torch
from torch.optim import Adam
from torch.nn import Linear, Sequential, CrossEntropyLoss, ReLU, Softmax
from sklearn.datasets import load_iris
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris = load_iris()
X = MinMaxScaler().fit_transform(iris['data'])
y = OneHotEncoder(sparse_output=False).fit_transform(iris['target'].reshape(-1,1))
X,y = torch.FloatTensor(X), torch.FloatTensor(y)
```

```python
model = Sequential()
model.add_module( 'nn1', Linear( 4,32 )) 
model.add_module( 'relu1', ReLU())
model.add_module( 'nn2', Linear( 32,3 )) 
model.add_module( 'sig1', Softmax(dim=1))

loss_fn = CrossEntropyLoss()
optimizer = Adam( model.parameters(), lr=1e-1)

for epoch in range( 1, 3001 ):
    optimizer.zero_grad()
    # z1 = matmul( x, w1) + b1   hx1 = ReLU(z1)   z= matmul( hx1, w2)+b2  hx= Sigmoid(z)
    hx = model( X )  # 4x2 2x32 4x32 32x1 4x1
    pred = (hx.detach().numpy()>0.5) +0
    cost = loss_fn( hx, y )
    cost.backward()
    optimizer.step()
    print(f'\repoch: {epoch}, acc: {(y.numpy()==pred).mean():.4f}, cost: {cost.item():.4f}',end='')
```
