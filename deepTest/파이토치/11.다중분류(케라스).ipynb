```python
import tensorflow as tf
from keras.optimizers import Adam
from keras import Sequential
from keras.layers import Dense
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('../data/softmax.txt', sep=' ', header=None)
df
```

```python
x_data = df.iloc[:, :2].values.astype(np.float32) # кіөл¶Җн•ң мӢңк°„, м¶ңм„қ мқјмҲҳ
y_data = df.iloc[:, 2:].values # н•ҷм җ
```

```python
y_data
```

```python
y_data.shape
```

```python
dense = Dense(units=3, input_dim=2, activation='softmax') # w: 2X3 (нҠ№м„ұлҚ°мқҙн„° к°ҜмҲҳ, лқјлІЁмқҳ к°ҜмҲҳ), b:[лқјлІЁмқҳ к°ҜмҲҳ]
dense(x_data)
```

```python
w, b = dense.get_weights() # н•ҷмҠөлҗҳм§Җ м•ҠмқҢ
print(w)
print('-'*40)
print(b)
```

```python
z = tf.matmul(x_data, w) + b
hx = tf.nn.softmax(z)
hx.numpy().argmax(axis=1)
```

```python
w.shape
```

```python
dense = Dense(units=3, input_dim=2, activation='softmax')
model = Sequential( [dense] )
model.compile(loss='categorical_crossentropy', optimizer=Adam(0.1), metrics=['acc']) # acc : м •нҷ•лҸ„лҘј м¶ңл Ҙн•ҙмӨҢ
h = model.fit(x_data, y_data, epochs=500)
```

```python
w, b = dense.get_weights() # н•ҷмҠөлҗң лҚ°мқҙн„°
print(w)
print('-'*40)
print(b)
```

```python
dense(x_data).numpy().argmax(axis=1)
```

```python
y_data.argmax(axis=1)
```

```python
model.predict(x_data).argmax(axis=1)
```

## iris лҚ°мқҙн„° м…Ӣ мјҖлқјмҠӨлЎң н•ҷмҠөн•ҳкі 
## 1,2 лІҲм§ё н–үмқҳ мҳҲмёЎк°’мқ„ кө¬н•ҳмӢңмҳӨ.

```python
from keras.utils import to_categorical
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris(as_frame=True)
iris['frame']
```

```python
xdata = iris['data'].values
ydata = to_categorical(iris['target'])
```

```python
xdata.shape
```

```python
ydata
```

```python
ydata.shape
```

```python
dense = Dense(units=3, input_dim=4, activation='softmax') # w: 2X3 (нҠ№м„ұлҚ°мқҙн„° к°ҜмҲҳ, лқјлІЁмқҳ к°ҜмҲҳ), b:[лқјлІЁмқҳ к°ҜмҲҳ]
dense(xdata)
```

```python
w, b = dense.get_weights() # н•ҷмҠөлҗҳм§Җ м•ҠмқҢ
print(w)
print('-'*40)
print(b)
```

```python
z = tf.matmul(xdata, w) + b
hx = tf.nn.softmax(z)
hx.numpy().argmax(axis=1)
```

```python
dense = Dense(units=3, input_dim=4, activation='softmax') # units= y, input_dim= x
model = Sequential( [dense] )
model.compile(loss='categorical_crossentropy', optimizer=Adam(0.1), metrics=['acc']) # acc : м •нҷ•лҸ„лҘј м¶ңл Ҙн•ҙмӨҢ
h = model.fit(xdata, ydata, epochs=500)
```

```python
model.predict(xdata[[0,1]]).argmax(axis=1) # 2X4 4X3
```
