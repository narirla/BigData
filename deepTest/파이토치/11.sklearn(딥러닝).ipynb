```python
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
import numpy as np
```

```python
x_data = np.array( [[0,0],[0,1],[1,0],[1,1]] )
y_data = np.array( [[0],[1],[1],[0] ] )
```

```python
model_logi = LogisticRegression( max_iter=1000, verbose=1)
model_logi.fit( x_data, y_data )
```

```python
model_logi.predict( x_data )
```

```python
model_logi.predict_proba( x_data )
```

```python
model_logi.coef_
```

```python
model_logi.intercept_
```

```python
model_mlp = MLPClassifier(max_iter=1000 ,hidden_layer_sizes=(100,) ,verbose=True )
model_mlp.fit( x_data, y_data )
```

```python
model_mlp.predict( x_data )
```

```python
model_mlp.predict_proba( x_data )
```

```python
model_mlp.coefs_
```

```python
model_mlp.coefs_[0].shape
```

```python
model_mlp.coefs_[1].shape
```

```python
model_mlp.intercepts_[0].shape
```

```python
model_mlp.intercepts_[1].shape
```

```python
model_mlp = MLPClassifier(max_iter=1000 ,hidden_layer_sizes=(64,32) ,verbose=True )
model_mlp.fit( x_data, y_data )
```

```python
model_mlp.predict( x_data )
```
