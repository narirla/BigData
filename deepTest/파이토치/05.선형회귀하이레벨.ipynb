```python
import torch
from torch.optim import Adam
from torch.nn import Linear, MSELoss, Sequential
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/cars.csv', index_col='Unnamed: 0')
df
```

```python
x = torch.FloatTensor( df[['speed']].values )
y = torch.FloatTensor( df[['dist']].values )
```

```python
linear = Linear( 1, 1 ) # w:1x1, b:1, hx = w*x + b
print( list( linear.parameters() )  )
```

```python
linear( torch.FloatTensor( [[15],[20]] ) )
```

```python
linear.forward( torch.FloatTensor( [[15],[20]] ) )
```

```python
linear = Linear( 1,1)
model = Sequential()
model.add_module( 'nn1', linear )
```

```python
list(model.parameters())
```

```python
loss_fn = MSELoss()
optimizer = Adam( model.parameters(),  lr=0.1)
for step in range(1000):
    optimizer.zero_grad() # 미분이 계속 누적되기 때문에 초기화 과정을 거침
    hx = model(x) # linear(x) --> w*x + b
    cost = loss_fn( hx, y ) 
    cost.backward() # dmse/dw, dmse/db
    optimizer.step() # w = w-0.1*dmse/dw , b = b-0.1*dmse/db
    print( step, cost.item() )
```

```python
model( torch.FloatTensor( [[15],[20]] ) )
```
