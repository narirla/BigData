```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic' 
```

```python
# 국어, 영어, 수학 성적 데이터 (5명의 학생)
arr = np.array([
    [85, 92, 78],
    [90, 88, 95],
    [75, 80, 82],
    [95, 90, 88],
    [80, 75, 90]
])
```

```python
xname = ['홍길동','이순신','임꺽정','김철수','이황']
```

```python
kor = arr[:,0]
eng = arr[:,1]
math = arr[:,2]
```

```python
kor
```

```python
plt.plot( kor )
```

```python
plt.plot(xname, kor)
plt.show()
```

```python
plt.plot(xname, kor)
plt.plot(xname, eng)
plt.legend( ['국어','영어'] )
plt.show()
```

```python
plt.xlabel('국어점수')
plt.ylabel('영어점수')
plt.scatter(kor , eng, s=100)
plt.show()
```

```python
plt.xlabel('국어점수')
plt.ylabel('국어점수')
plt.scatter(kor , kor, s=100)
plt.show()
```
