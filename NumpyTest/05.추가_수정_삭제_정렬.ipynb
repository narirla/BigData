```python
import numpy as np
```

```python
arr = np.array([11,22,33,44,55])
arr
```

### CRUD

### 추가

```python
arr
```

```python
np.append(arr, [100,200]) #추가할 값을 , 뒤로 배치
```

```python
np.insert(arr, 1,[1,2]) #(대상 array, 해당 인덱스 , 추가할 값)
```

### 수정

공간 = 값

```python
arr[1] = 200 # 해당 공간에 값을 주게 됨
```

```python
arr
```

```python
arr[1:3] = [1,2]
```

```python
arr[1:3] = 10
```

```python
arr[3:5] = arr[3:5]+2
```

```python
arr
```

```python
### Q) 30이상인 데이터를 기존 값에다 2를 더한 것으로 수정하시오.
### 결과; arr([11,22,35,46,57])
arr[arr>=30] = arr[arr>=30]+2
arr
```

### 삭제

```python
# np.delete(대상 배열,삭제할 인덱스)
# arr = np.delete(arr,[0,2]) ## arr= 을 추가해줘야 arr에 삭제가 적용
np.delete(arr,[0,2])
```

```python
np.where(arr<30) #해당 조건의 인덱스를 반환됨
```

```python
np.where(arr>=30, 'A','B') #True = 'A', False = 'B'
```

```python
np.select([arr>=40,arr>=30,arr>=0],['A','B','C'],'F')
```

```python
arr
```

### 정렬

```python
arr[1] = 100
```

```python
np.sort(arr)
```

```python
np.sort(arr)[::-1] # 내림차순은 따로 없음으로 역슬라이싱으로 해야함
```

```python
arr[np.argsort(arr)] # arr[[0, 2, 3, 4, 1]] 복수개 인덱싱
```
