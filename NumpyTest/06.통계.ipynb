```python
a = np.array([10,20,20,20,30,30,40])
values, counts = np.unique(a, return_counts=True)
print(values)
print(counts)
```

```python
a = np.array([10,20,20,20,30,30,40])
np.unique(a)
```

```python
a = np.array([10,20,20,20,30,30,40,55,60,75,80,95])
bins = [0, 50, 70, 100]
freq, edges = np.histogram(a, bins=bins)
print( freq) 
print( edges )
```

| 구간 | 의미 |
|------|------|
| 0 ~ 50 | **0 ≤ d < 50** |
| 50 ~ 70 | **50 ≤ d < 70** |
| 70 ~ 100 | **70 ≤ d < 100** |
