## reshape 함수 (사용 빈도 多)

```python
arr = np.array([1,2,3,4,5,6])
arr
```

```python
arr.reshape(2,3) #동일한 데이터를 유지한 채, 새로운 형태(모양)의 배열을 반환
```

```python
arr.reshape(2,-1) #(2행, 알아서)
```

```python
arr.reshape(-1,2) #(알아서, 2열)
```

```python
arr1 = np.array( [[1,2,3],[4,5,6]] )
arr1.reshape(-1) 
# arr1.reshape(6) 
```
