## 넘파이 연산(산관논 연산자 사용 가능)
- element wise : 요소별 연산(동일한 인덱스 끼리 연산)
- broad casting : 차원이 높은 쪽으로 자동 확장 

```python
import numpy as np
arr = np.array([11,22,33,44,55])
arr
```

```python
arr+2 #[11,22,33,44,55] + [2,2,2,2,2] = [13, 24, 35, 46, 57] 벡터로 확장되어 동일한 인덱스끼리 연산됨
```

```python
arr*2  #[11,22,33,44,55] * [2,2,2,2,2] = [ 22,  44,  66,  88, 110] 벡터로 확장되어 동일한 인덱스끼리 연산됨
```

```python
arr+[1,2,3,4,5] # [11,22,33,44,55] + [1,2,3,4,5]
```

```python
### 퀴즈: 세금 3.3%를 제외한 실수령액을 구하시오.
salary = np.array([1000,2000,3000,4000,5000])
salary
```

```python
salary*(1-0.033)
```

### 관계 연산자

```python
arr>30 # [11,22,33,44,55] > [30,30,30,30,30]
```

```python
mylist = [11,22,33,44,55]
[n for n in mylist if n>30]
```

```python
arr[arr>30] ##arr[[False, False,  True,  True,  True]] -> boolean indexing이 적용되어 True 값인 [33, 44, 55]이 출력
```

```python
arr<=20 #[11,22,33,44,55] <= [20,20,20,20,20]
```

```python
arr[arr<=20] #arr[[ True, False, False, False, False]] -> True 값만 출력
```

### 논리 연산자
- and: &  -> 둘다 True인 경우만 True
- or: | -> 둘중 하나만 True인 경우도 True
- not: ~

(arr==22) | (arr==44)
1. step: borad casting
([11,22,33,44,55] == [22,22,22,22,22]) |
([11,22,33,44,55] == [44,44,44,44,44])
2. element wise
([F,T,F,F,F]) | ([F,F,F,T,F])
= [F,T,F,T,F]

```python
(arr==22) | (arr==44) 
```

```python
arr[(arr==22) | (arr==44)]
```

```python
(arr>=20) & (arr<=40)
```

```python
arr[(arr>=20) & (arr<=40)]
```

```python
~((arr==22) | (arr==44))
```

```python
arr[~((arr==22) | (arr==44))]
```
