### 6.  random  <- 딥러닝, 머신러닝 파트에서 주로 사용하게 됨

```python
np.random.randint(1,10,5) # 1<=r<10 1~9까지의 숫자를 랜덤하게 array를 만들어줌
```

```python
np.random.randint(1,10,(2,3)) #2행3열짜리로 랜덤하게 만들어줌
```

```python
np.random.randn(5) #표준 정규분포(평균0 , 편차1) 
```

```python
arr = np.arange(1,11)
np.random.shuffle(arr) # 중복 제거 시 suffle 사용
arr[:5]
```

```python
np.random.seed(10) #seed: 랜덤하게 생성되는 값이 고정되어 동일하게 출력됨
np.random.randint(1,10,5)
```
