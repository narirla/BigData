### 관계 연산자

````python
arr = np.array([
    [85, 92, 78],
    [90, 88, 95],
    [75, 80, 82],
    [95, 90, 88],
    [80, 75, 90]
])
````

````python
arr
````

````python
arr[:,0] >= 80
````

````python
arr[arr[:,0] >= 80] #국어 점수가 80점 이상인 행을 선택
````

````python
# 영어 점수가 80점 이상 이거나 90점 이상인 행을 선택
arr[(arr[:,1] == 80) | (arr[:,1]==90)] # |(or)는 둘 중 하나라도 True면 True
````

````python
# 수학 점수가 85점이상 95점 이하인 행을 선택
arr[(arr[:,2] >= 85) & (arr[:,2]<=95)]
````

### 데이터 읽기

````python
# birthArr = np.loadtxt('data/births.csv',delimiter=',',dtype=np.int64, skiprows=1)
birthArr = np.loadtxt('data/births.csv',delimiter=',',dtype=np.int64)
birthArr
````

1. 행의 갯수를 구하시오.
2. 남아, 여아 출생 평균을 구하시오.
3. 남아, 여아 표준편차를 구하시오.
4. 여아 사분위수를 구하시오.
5. 남아 여아 상관관계를 구하시오.
6. 2000년대 데이터를 구하시오
7. 남아 수가 150000명 이상인 연도만 추출하시오
8. 연도별 총 출생아 수(남아+여아)를 새로운 열로 추가하시오.

````python
#1. 행의 갯수를 구하시오.
print(birthArr.shape[0])
print(len(birthArr))
````

````python
#2. 남아, 여아 출생 평균을 구하시오.
birthArr[:,1:].mean(axis=0).round(2)
````

````python
# 3. 남아, 여아 표준편차를 구하시오.
birthArr[:,1:].std(axis=0).round(2)
````

````python
# 4. 여아 사분위수를 구하시오.
np.quantile(birthArr[:,2],[0.25,0.5,0.75])
````

````python
# 5. 남아 여아 상관관계를 구하시오.
np.corrcoef(birthArr[:,1:], rowvar=False)
````

````python
# 6. 2000년대 데이터를 구하시오
birthArr[birthArr[:,0]>=2000]
````

````python
# 7. 남아 수가 150000명 이상인 연도만 추출하시오
birthArr[birthArr[:,1]>=150000][:,0]
````

````python
# 8. 연도별 총 출생아 수(남아+여아)를 새로운 열로 추가하시오.
total = birthArr[:,1] + birthArr[:,2]
# total.reshape(-1,1)
# np.hstack((birthArr,total.reshape(-1,1)))
np.c_[birthArr, total] # 속성, 벡터로 가능
````

````python
print('년도','남아','여아','총 출생아 수',sep="\t")
print("="*40)
for y,b,g,t in np.hstack((birthArr,total.reshape(-1,1))):
    print(y,b,g,t, sep="\t")
````
