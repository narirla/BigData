##  NumPy 정의
NumPy(Numerical Python)는  
고성능 **수치 계산과 배열(행렬) 연산을 위한 파이썬 라이브러리**이다.  
과학계산, 데이터 분석, 머신러닝, 이미지 처리 등에서 필수적으로 사용된다.

---

##  NumPy의 핵심 특징

### ✅ 1. 빠르고 효율적인 다차원 배열 객체(ndarray)
- 파이썬 리스트보다 **10~100배 빠름**
- 대규모 데이터 연산에 최적화됨
- 내부적으로 C로 구현되어 성능이 뛰어남

### ✅ 2. 벡터화(Vectorization) 연산 지원
- 반복문 없이 **배열 전체에 일괄 연산 가능**
- 예: `arr * 2`, `arr + 5`
- 코드가 짧고 성능이 뛰어남

### ✅ 3. 브로드캐스팅(Broadcasting)
- 서로 다른 크기의 배열 간 연산을 자동 확장하여 수행  
  → 반복문 없이 바로 계산 가능

### ✅ 4. 다양한 수학/과학 연산 함수 제공
- 선형대수, 통계, 확률, FFT, 삼각함수 등 내장
- 복잡한 수학 계산을 쉽게 수행 가능

### ✅ 5. 타 라이브러리의 기반
- Pandas, Matplotlib, Scikit-learn, TensorFlow 등  
  대부분의 과학·AI 라이브러리가 **NumPy 배열 기반으로 동작**

---

## 🔹 NumPy가 필요한 이유

| 이유 | 설명 |
|------|------|
| **속도** | 순수 파이썬보다 훨씬 빠른 연산 |
| **편의성** | 반복문 없이 배열 전체 계산 |
| **일관성** | 데이터 과학 라이브러리들과 통일된 배열 구조 사용 |
| **확장성** | 고성능 계산에 최적화 |

---

## 🔹 한 줄 요약
> **NumPy는 파이썬에서 빠르고 효율적인 수치 계산을 가능하게 하는 핵심 라이브러리로, 배열(행렬) 연산을 기반으로 데이터 사이언스·머신러닝 분야의 사실상 표준이다.**



# 📌 NumPy 핵심 문법 정리표

## 1. 배열 생성 (Array Creation)

| 기능 | 코드 예시 | 설명 |
|------|-----------|------|
| 리스트로 배열 생성 | `np.array([1,2,3])` | 기본 배열 생성 |
| 0으로 채운 배열 | `np.zeros((2,3))` | 2×3 배열 생성 |
| 1로 채운 배열 | `np.ones((3,3))` | 3×3 배열 생성 |
| 특정 값으로 채우기 | `np.full((2,2), 7)` | 7로 채운 2×2 |
| 범위 배열 | `np.arange(0, 10, 2)` | 0~10, step 2 |
| 균일 분포 배열 | `np.linspace(0, 1, 5)` | 0~1 사이 5개 점 |
| 단위 행렬 | `np.eye(4)` | 4×4 Identity Matrix |
| 난수 배열 | `np.random.rand(2,3)` | 0~1 랜덤값 |

| 속성 | 예시 | 설명 |
|------|------|------|
