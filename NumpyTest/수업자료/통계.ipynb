##  NumPy 통계 함수 표 (1차원 기준)


```python

import numpy as np
a = np.array([1,2,3])
```

| 함수 | 설명 | 예시 코드 | 결과 예시 |
|------|------|-----------|-----------|
| `np.sum()` | 모든 요소 합계 | `np.sum(a)` | 1+2+3 → 6 |
| `np.mean()` | 평균 | `np.mean(a)` | (1+2+3)/3 |
| `np.median()` | 중앙값 | `np.median(a)` | 2 |
| `np.std()` | 표준편차 | `np.std(a)` | 0.816... |
| `np.var()` | 분산 | `np.var(a)` | 0.666... |
| `np.min()` | 최솟값 | `np.min(a)` | 1 |
| `np.max()` | 최댓값 | `np.max(a)` | 3 |
| `np.argmin()` | 최솟값의 인덱스 | `np.argmin(a)` | 0 |
| `np.argmax()` | 최댓값의 인덱스 | `np.argmax(a)` | 2 |
| `np.ptp()` | 최대–최소 범위 | `np.ptp(a)` | 3 - 1 = 2 |
| `np.percentile()` | 백분위수 | `np.percentile(a, 50)` | 2 |
| `np.quantile()` | 분위수 | `np.quantile(a, 0.25)` | 1.5 |
| `np.cumsum()` | 누적합 | `np.cumsum(a)` | [1,3,6] |
| `np.diff()` | 1차 차분 | `np.diff(a)` | [1,1] |
| `np.sort()` | 정렬 | `np.sort(a)` | [1,2,3] |

---



`np.diff(x)`는 배열 `x`의 **인접한 원소들 사이의 차이(1차 차분)** 를 계산.  
수식으로 쓰면 

$
\text{diff}[i] = x[i+1] - x[i]
$  
즉, 오른쪽 원소에서 왼쪽 원소를 뺀다

---

```python

diff 함수 옵션
n : 차분의 차수 (기본 1). n만큼 반복 차분을 계산함.
axis : 다차원 배열에서 차분할 축(기본 마지막 축).
prepend, append : 차분 전에/후에 값을 붙여 결과 길이를 조절할 수 있음(경계 처리).

import numpy as np
a = np.array([1, 2, 3])
print(np.diff(a))        # [1 1]
print(np.diff(a, n=1))   # 동일, 1차 차분
print(np.diff(a, n=2))   # 2차 차분 -> [0]

np.diff(a, prepend=0)   # [1-0, 2-1, 3-2] -> [1,1,1]




#  NumPy 2차원 ndarray 통계 함수   

```python

import numpy as np

b = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

| axis     | 의미              | 예시          |
| -------- | --------------- | ----------- |
| `axis=0` | **세로 방향(열 기준)** | 각 열별 통계     |
| `axis=1` | **가로 방향(행 기준)** | 각 행별 통계     |
| 없음       | 전체              | 전체 원소 기준 통계 |


| 함수                | 설명     | 기본 사용                  | 축(axis) 사용                     | 결과 예시                         |
| ----------------- | ------ | ---------------------- | ------------------------------ | ----------------------------- |
| `np.sum()`        | 전체 합   | `np.sum(b)`            | `np.sum(b, axis=0)`            | 전체=21, 열합=[5,7,9]             |
| `np.mean()`       | 평균     | `np.mean(b)`           | `np.mean(b, axis=1)`           | 전체=3.5, 행평균=[2,5]             |
| `np.median()`     | 중앙값    | `np.median(b)`         | `np.median(b, axis=0)`         | 전체=3.5, 열=[2.5,3.5,4.5]       |
| `np.std()`        | 표준편차   | `np.std(b)`            | `np.std(b, axis=1)`            | 전체=1.707…, 행=[0.816,0.816]    |
| `np.var()`        | 분산     | `np.var(b)`            | `np.var(b, axis=0)`            | 전체=2.916…, 열=[2.25,2.25,2.25] |
| `np.min()`        | 최솟값    | `np.min(b)`            | `np.min(b, axis=1)`            | 전체=1, 행=[1,4]                 |
| `np.max()`        | 최댓값    | `np.max(b)`            | `np.max(b, axis=0)`            | 전체=6, 열=[4,5,6]               |
| `np.argmin()`     | 최소값 위치 | `np.argmin(b)`         | `np.argmin(b, axis=1)`         | 전체 index=0, 행=[0,0]           |
| `np.argmax()`     | 최대값 위치 | `np.argmax(b)`         | `np.argmax(b, axis=0)`         | 전체 index=5, 열=[1,1,1]         |
| `np.percentile()` | 백분위수   | `np.percentile(b, 50)` | `np.percentile(b, 50, axis=0)` | 전체=3.5                        |
| `np.cumsum()`     | 누적합    | `np.cumsum(b)`         | `np.cumsum(b, axis=1)`         | 행누적=[[1,3,6],[4,9,15]]        |
| `np.diff()`       | 차분     | `np.diff(b)`           | `np.diff(b, axis=0)`           | 행기준=[3,3,3]                   |
| `np.sort()`       | 정렬     | `np.sort(b)`           | `np.sort(b, axis=0)`           | 열정렬=[[1,2,3],[4,5,6]]         |
