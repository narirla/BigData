#  NumPy `ndarray` 생성하는 방법

- NumPy에서 배열(`ndarray`)을 만드는 방법은 매우 다양하며, 사용 목적에 따라 적절한 생성 함수를 선택할 수 있다. 
---

| 기능       | 함수               | 예시                          |
| -------- | ---------------- | --------------------------- |
| 리스트 → 배열 | `np.array`       | `np.array([1,2])`           |
| 범위       | `np.arange`      | `np.arange(5)`              |
| 균등 간격    | `np.linspace`    | `np.linspace(0,1,5)`        |
| 로그 간격    | `np.logspace`    | `np.logspace(1,3,3)`        |
| 0 배열     | `np.zeros`       | `np.zeros((3,3))`           |
| 1 배열     | `np.ones`        | `np.ones((2,2))`            |
| 특정값      | `np.full`        | `np.full((2,2),7)`          |
| 단위행렬     | `np.eye`         | `np.eye(4)`                 |
| 대각선      | `np.diag`        | `np.diag([1,2])`            |
| 랜덤       | `np.random.rand` | `np.random.rand(2,3)`       |
| shape 동일 | `np.ones_like`   | `np.ones_like(x)`           |
| reshape  | `reshape`        | `np.arange(6).reshape(2,3)` |
| 빈 배열     | `np.empty`       | `np.empty((3,3))`           |


### 1. 리스트/튜플로부터 배열 생성 — `np.array()`

```python
import numpy as np

a = np.array([1, 2, 3])
b = np.array([[1, 2], [3, 4]])


###  2. 연속된 숫자 생성 — np.arange()

```python
np.arange(0, 10, 2)   # 0~8, step=2

### 3. 균등 간격 숫자 생성 — np.linspace()

```python
np.linspace(0, 1, 5)  # 0과 1 사이를 5개 구간으로 나눔

### 4. 로그 스케일 간격 생성 — np.logspace()

np.logspace(1, 3, 3)  # 10^1 ~ 10^3 사이 로그 스케일

### 5. 모든 값이 0인 배열 생성 — np.zeros()
np.zeros((3, 4))

### 6. 모든 값이 1인 배열 생성 — np.ones()
np.ones((2, 3))

### 7. 특정 값으로 채운 배열 생성 — np.full()
np.full((2, 2), 7)   # 모든 요소가 7

### 8. 단위 행렬 생성 — np.eye()
np.eye(4)  # 4x4 identity matrix

### 9. 대각선 배열 생성 — np.diag()
np.diag([1, 2, 3])

### 10. 무작위 배열 생성 — np.random.randn(), np.random.rand(), np.random.randint()
np.random.rand(2, 3)       # 0~1 사이 균등 분포
np.random.randn(2, 3)      # 표준 정규 분포
np.random.randint(0, 10, (2,2))  # 0~9 난수

### 11. 기존 배열과 동일 shape으로 생성 — np.zeros_like(), np.ones_like(), np.full_like()
x = np.array([[1,2,3]])
np.zeros_like(x)
np.ones_like(x)
np.full_like(x, 9)

### 12. reshape하여 생성 — np.reshape()
np.arange(12).reshape(3, 4)

### 13. 빈 배열(초기값 미정) 생성 — np.empty()
np.empty((2, 3))   # 초기화되지 않은 메모리

### 14. 랜덤 정규분포 — np.random.normal()
np.random.normal(loc=0, scale=1, size=(2,3))

### 15. 균등분포 — np.random.uniform()
np.random.uniform(0, 1, (3,3))

#  로그 스케일(log scale)을 왜 사용할까? — 쉽게 설명

로그 스케일(log scale)은 **숫자의 차이가 너무 클 때, 이를 한눈에 보기 쉽게 만들기 위해 사용하는 축(scale)** 입니다.

---

##  1. 값의 범위가 너무 넓을 때 보기 좋게 줄여준다

예를 들어 숫자들이 이렇게 있다면:

- 1  
- 10  
- 100  
- 1,000  
- 10,000  
- 100,000  

그래프에 그대로 그리면 **작은 값(1, 10, 100)은 거의 바닥에 붙어서 보이지 않음**  
→ 큰 값에 비해 너무 작기 때문!

 로그 스케일을 사용하면 이 값들이 **동일한 비율**로 보이게 됨  
(10배 → 10배 → 10배 증가가 일정 간격으로)

---

##  2. “비율 차이”를 볼 때 유리하다

일반 스케일: 1 → 100 은 99 차이  
로그 스케일: 1 → 100 은 “100배 증가”로 표현  

즉, **배수 증가**, **성장률**, **비율 변화**를 볼 때 로그가 훨씬 명확함.

---

##  3. 과학·공학 데이터는 원래 규모가 큼  
예:
- 소리의 세기(dB)
- 지진의 크기(리히터 규모)
- 빛의 밝기
- 박테리아 증가
- 금융(복리 성장)

이런 데이터들은 “10배씩 증가”하는 일이 흔해서  
로그 스케일이 더 자연스럽고 읽기 편하다.

---

##  예시 비교

### 🔹 일반 스케일
1, 10, 100, 1000, 10000 을 그래프에 표시하면  
→ 1, 10은 거의 안 보임  
→ 10000만 튀어나옴

###  로그 스케일
1, 10, 100, 1000, 10000 값이  
→ 같은 간격으로 보임  
→ 전체 흐름을 한눈에 볼 수 있음

---

#  한 문장 요약
**로그 스케일은 값의 크기 차이가 매우 클 때, 배수(비율) 기반으로 데이터를 더 보기 좋게 표현하기 위한 방식이다.**

---



#  로그 스케일 vs 일반 스케일 비교 그래프 

````python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'  # 한글 폰트 설정
# 예제 데이터 (범위가 크게 차이남)
x = np.arange(1, 6)
y = np.array([1, 10, 100, 1000, 10000])

plt.figure(figsize=(12, 5))

# -------------------------------
# 1. 일반 스케일 그래프
# -------------------------------
plt.subplot(1, 2, 1)
plt.plot(x, y)
plt.title("일반 스케일 (Linear Scale)")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)

# -------------------------------
# 2. 로그 스케일 그래프
# -------------------------------
plt.subplot(1, 2, 2)
plt.plot(x, y)
plt.yscale('log')     # y축 로그 스케일 설정
plt.title("로그 스케일 (Log Scale)")
plt.xlabel("x")
plt.ylabel("log(y)")
plt.grid(True)

plt.tight_layout()
plt.show()
````

#  NumPy의 `dtype`(데이터 타입)으로 `ndarray` 생성하기  
 **dtype을 지정하여 배열을 만드는 방법**

---

#  1. 주요 NumPy 데이터 타입(dtypes) 표

| dtype | 설명 | 예시 |
|------|------|------|
| `np.int8` | 8비트 정수 | -128 ~ 127 |
| `np.int16` | 16비트 정수 | -32768 ~ 32767 |
| `np.int32` | 32비트 정수 (기본 정수) | 큰 정수 처리 가능 |
| `np.int64` | 64비트 정수 | 매우 큰 정수 처리 |
| `np.uint8` | 부호 없는 8비트 정수 | 0~255 (이미지 처리에서 자주 사용) |
| `np.float16` | 16비트 실수 | 메모리 절약, 낮은 정밀도 |
| `np.float32` | 32비트 실수 | 딥러닝에서 많이 사용 |
| `np.float64` | 64비트 실수 (기본 float) | 높은 정밀도 |
| `np.bool_` | 불리언 타입 | True / False |
| `np.str_` | 문자열 타입 | Python str 확장 |
| `np.object_` | 파이썬 객체 타입 | 다양한 객체 저장 가능 |

---

3. dtype을 사용하여 ndarray 생성하는 방법 요약표

| 생성 함수                        | dtype 지정 | 예시                                      |
| ---------------------------- | -------- | --------------------------------------- |
| `np.array()`                 | 가능       | `np.array([1,2], dtype=np.float32)`     |
| `np.zeros()`                 | 가능       | `np.zeros((3,3), dtype=np.uint8)`       |
| `np.ones()`                  | 가능       | `np.ones(5, dtype=np.int16)`            |
| `np.full()`                  | 가능       | `np.full((2,2), 3, dtype=np.float64)`   |
| `np.arange()`                | 가능       | `np.arange(5, dtype=np.float32)`        |
| `np.linspace()`              | 가능       | `np.linspace(0,10,5, dtype=np.float32)` |
| `np.empty()`                 | 가능       | `np.empty((3,3), dtype=np.int32)`       |
| `np.identity()` / `np.eye()` | 가능       | `np.eye(3, dtype=np.float32)`           |
| `astype()`                   | 변환       | `arr.astype(np.float32)`                |


```python
 
 4. dtype 변환 예제 3가지

✔ 정수 → 실수 변환

✔ 실수 → 정수 변환
