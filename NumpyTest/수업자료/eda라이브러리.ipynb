# 📊 EDA + 딥러닝 + Streamlit 배포 파이썬 라이브러리

---

#  1. **EDA 라이브러리**

## 1) NumPy
- **과학 계산 기반 라이브러리**
- 고속 배열 연산, 벡터/행렬 계산
- pandas·scikit-learn 등 다른 라이브러리들의 기반

### 주요 기능
- ndarray 생성/연산
- 수학, 통계 함수(mean, std, sum)
- 난수 생성

---

## 2) pandas
- 표 형태 데이터(DataFrame) 처리 핵심 라이브러리
- 결측치 처리·그룹화·정렬·머지 등 데이터 핸들링 최강

### 주요 기능
- CSV/Excel/JSON 불러오기
- 결측치 처리
- groupby(), merge(), pivot_table()
- 시계열 분석 도구

---

## 3) Matplotlib
- **파이썬 기본 시각화 도구**
- 그래프 커스터마이징 자유도 높음

### 주요 기능
- 선/막대/산점도/히스토그램
- 2D 시각화 대부분 가능

---

## 4) Seaborn
- Matplotlib 기반의 **통계 시각화 라이브러리**
- 예쁘고 직관적인 그래프 제공

### 주요 기능
- 히트맵(상관관계)
- 카운트플롯
- 바이올린 플롯
- 회귀선(regplot)

---

## 5) Plotly
- **인터랙티브 그래프** 제공
- 웹/대시보드에서 사용 가능

### 주요 기능
- 줌/호버 가능한 그래프
- 3D 플롯
- 지도 기반 시각화

---

## 6) SciPy
- 고급 수학·통계·최적화 기능

### 주요 기능
- t-test, 정규성 검정 등 통계 분석
- 최적화 알고리즘
- 신호 처리

---

## 7) Scikit-learn
- 머신러닝/전처리/기본 분석에 사용

### 주요 기능
- train_test_split()
- StandardScaler()
- PCA 차원축소
- 기본 모델로 데이터 패턴 확인

---

## 8) Missingno
- **결측치 시각화 전용**

### 주요 기능
- 결측 막대 그래프
- 매트릭스 결측 구조
- 결측 히트맵

---

## 9) ydata-profiling (구 pandas-profiling) / Sweetviz
- 자동 EDA 리포트 생성
- 데이터 요약을 빠르게 한 번에 보고 싶을 때 최적

---

#  2. **딥러닝 라이브러리**

## 1) TensorFlow / Keras
- 가장 널리 사용되는 딥러닝 프레임워크
- Keras: 직관적 고수준 API 제공

### 주요 기능
- 신경망 모델 생성, 학습, 평가
- CNN, RNN, LSTM 등 다양한 아키텍처
- GPU 학습 지원
- 모델 저장/로드

---

## 2) PyTorch
- 연구/실험용으로 가장 많이 쓰는 딥러닝 프레임워크
- 직관적인 텐서 연산, 동적 그래프

### 주요 기능
- 모델 정의가 유연함(nn.Module)
- torchvision으로 이미지 데이터 쉽게 처리
- torchtext로 텍스트 처리

---

## 3) TensorBoard
- 딥러닝 학습 과정 시각화 도구

### 주요 기능
- loss/accuracy 곡선 확인
- weight 히스토그램
- 그래프 구조 텐서 시각화

---

## 4) OpenCV
- 이미지 전처리/증강에 많이 사용

### 주요 기능
- 이미지 읽기/자르기/회전
- 색상 변환(BGR ↔ RGB)
- 엣지 검출, 필터링

---

## 5) Albumentations
- 딥러닝 학습용 이미지 증강 라이브러리
- 빠르고 강력함

---

#  3. **Streamlit (웹 배포) 라이브러리**

## 1) Streamlit
- 파이썬 코드 몇 줄로 **웹 앱 배포 가능**
- 데이터 분석/딥러닝/AI 결과를 웹 UI로 쉽게 보여줌

### 주요 기능
- 입력폼(text_input, file_uploader)
- 그래프 표시(Matplotlib/Plotly/pandas)
- 모델 결과 출력
- 캐싱(st.cache_data, st.cache_resource)

---

## 2) streamlit-lottie
- Lottie 애니메이션을 웹에 표시 가능

---

## 3) streamlit-option-menu
- 상단/좌측 메뉴 만들 때 사용

---

## 4) Streamlit + 모델 배포 방식
- 모델(.h5/.pt/.pkl) 불러와서 결과 예측  
- 이미지 업로드 후 예측  
- 차트 및 분석 결과 웹 페이지로 제공
