#  EDA(Exploratory Data Analysis, 탐색적 데이터 분석)

EDA(탐색적 데이터 분석)는 **데이터의 전반적인 구조, 패턴, 이상치, 분포 등을 직관적으로 탐색**하여  
데이터를 이해하고 향후 모델링 방향을 잡기 위한 필수 과정.

```text
유명한 통계학자 로널드 코스(Ronald H. Coase)의 격언인 **"데이터를 충분히 고문한다면, (원하는) 자백을 받아낼 수 있을 것이다 (If you torture the data long enough, it will confess)"**

데이터 분석, 특히 탐색적 데이터 분석(EDA)의 목적은 데이터 자체에 내재된 패턴과 정보를 객관적으로 발견하는 것. 분석가의 선입견이나 가정에 맞춰 데이터를 조작해서는 안 된다는 점을 강조.
```

##  EDA의 목적

- 데이터의 기본 구조 파악  
- 결측치, 이상치 탐지  
- 변수 간 관계 파악  
- 데이터 분포 및 패턴 확인  
- 모델링 전 전처리 방향 설정  
---


<img src='/image/eda.png'>



1. 데이터 이해: 변수 유형, 데이터 분포, 결측값 유무 등을 파악합니다.

2. 데이터 정제 방향 설정: 데이터 정제(Cleaning) 및 변환(Transformation)에 필요한 단계를 결정합니다.

3. 패턴 식별: 변수 간의 관계(상관관계), 추세, 계절성 등의 숨겨진 구조를 찾아냅니다.

4. 이상값(Outliers) 감지: 모델의 성능에 부정적인 영향을 미칠 수 있는 특이한 데이터를 식별하고 처리합니다.

5. 분석: 통계, 시각화

6. 데이터 마이닝(머신러닝 딥러닝): 방대한 데이터속에서 숨겨진 규칙 패턴 등을 찾아내어 예측하거나 의사 결정에 활용하는 것을 목적으로 한다

# 이변수, 다변수란

EDA(탐색적 데이터 분석)에서 **단변수 / 이변수 / 다변수 분석**은 분석 범위를 기준으로 데이터를 나누어 살펴보는 방법.

---

## 🔹 1. 단변수(Univariate) 분석
**하나의 변수만** 단독으로 분석하는 것  
→ 분포, 평균, 중앙값, 최빈값 등을 확인

예:  
- 키(height) 분포  
- 성별(gender) 빈도수  

---

## 🔹 2. 이변수(Bivariate) 분석
**두 개의 변수 간 관계를 분석하는 것**

### ✔ 예시
- 키(height)와 몸무게(weight) 관계  
- 광고비(ad_spend)와 매출(sales)의 상관관계  
- 성별(gender)과 구매여부(purchase)의 관계  

### ✔ 주로 사용하는 방법
| 목적 | 사용 방법 |
|------|-----------|
| 두 수치형 변수 관계 | scatter plot, 상관계수 |
| 범주형 ↔ 수치형 관계 | boxplot, barplot |
| 두 범주형 관계 | 교차표(crosstab), chi-square test |

---

## 🔹 3. 다변수(Multivariate) 분석
**세 개 이상의 변수** 관계를 동시에 분석하는 것

### ✔ 예시
- 나이(age) • 소득(income) • 소비(amount) 관계  
- 높이(height) • 몸무게(weight) • 건강지수(health_score)  
- 시간(time) • 지역(region) • 매출(sales)  

### ✔ 주요 분석 방법
| 목적 | 사용 기법 |
|------|-----------|
| 여러 변수 간 상관성 | heatmap, pairplot |
| 그룹별 통계 비교 | groupby, pivot_table |
| 예측 모델 기반 탐색 | 회귀분석, 의사결정나무 |
| 차원 축소 | PCA(Principal Component Analysis) |

---

##  요약
- **단변수**: 변수 하나만 본다  
- **이변수**: 두 변수 간 관계를 본다  
- **다변수**: 세 개 이상의 변수 관계를 본다
