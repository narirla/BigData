````python
import numpy as np

birthArr = np.loadtxt('data/births.csv', delimiter=',', dtype=np.int64, skiprows=1)
````

1. 남아 수가 평균보다 많은 연도만 출력하시오


````python
birthArr[np.where(birthArr[:,1]>=np.average(birthArr[:,1]))][:,0]
````


2. 남아출생이 가장많은 top5
를 구하시요(년도와 남아수)

````python
top5_index = np.argsort(birthArr[:,1])[:-6:-1]
birthArr[top5_index,:-1]
````

3. 남아율(%) 을 구하고 각년도별
남아수가 차지하는비율컬럼을 추가하시요

````python
boy_rate = birthArr[:,[1]]/(birthArr[:,[1]]+birthArr[:,[2]])*100
np.hstack((birthArr, boy_rate))
````

4. 2000년도 이후데이터 에 대해
출생량 을 구하고
남아수가 1800000 이상이면 '많음'아니면 '적음'
이라고 보여 주시요
```
년도 	남아수 출생량
----------------------
2000   1233   '많음'
2001   1234   '적음'

````python
after_century = birthArr[np.where(birthArr[:,0]>=2000)][:,[0,1]]
boy_many = np.where(after_century[:,[1]]>=1800000,'많음','적음')
np.hstack( (after_century, boy_many) )
````


5.1800년대,1900년대,2000년대 
각 남아수,여야수 평균을 구하시요

````python
np.select( [birthArr[:,0]>=2000, birthArr[:,0]>=1900, birthArr[:,0]>=1800] , [np.average(birthArr[:,[1,2]]), np.average(birthArr[:,[1,2]]), np.average(birthArr[:,[1,2]])])
````

````python
after_2000 = birthArr[np.where(birthArr[:,0]>=2000)]
x = np.average(after_2000[:, [1,2]], axis=0)

after_1900 = birthArr[np.where( (birthArr[:,0]>=1900)&(birthArr[:,0]<2000) )]
y = np.average(after_1900[:, [1,2]], axis=0)

after_1800 = birthArr[np.where( (birthArr[:,0]>=1800)&(birthArr[:,0]<1900) )]
z = np.average(after_1800[:, [1,2]], axis=0)

print(x, y, z)
````

6. 연도별로 누가 더 많이 태어났는지(남아 vs 여아) 비교하시오
각 행에서 남아 > 여아 이면 'Boy', 아니면 'Girl'을 출력.
출력결과
```
년도 결과
----------------
1800 boy
1801 boy
....
2100 girl

````python
np.hstack( (birthArr[:,[0]], np.where(birthArr[:,[1]]>birthArr[:,[2]], 'Boy', 'Girl')) )
````


7. 총 출생아 수 기준 가장 많은 해 년도를 구하시요

````python
max_birth = np.max(np.sum(birthArr[:,[1,2]], axis=1))

birthArr[np.where(np.sum(birthArr[:,[1,2]], axis=1) == max_birth)][:,0]
````


8. 여아 수가 남아 수보다 많은 연도만 추출하시오

````python
birthArr[np.where(birthArr[:,1]<birthArr[:,2])][:,0]
````


9. 남아 증가량(전년 대비 변화)을 계산하시오

````python
birthArr.shape[0]

for i in range(0, birthArr.shape[0]-1):
    print(birthArr[i+1,0], birthArr[i+1,1]-birthArr[i,1])
````
