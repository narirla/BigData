12. 국어점수의 범위(0~50,50~70,70~100)별 돗수를 구하시요 
( 범위, 0~50,50~70,70~100)

````python
ran = [0,50,70,100]
np.histogram(kor,ran)
````

````python
bins = [0, 50, 70, 100]
freq, edges = np.histogram(kor, bins=bins)

print('범위 \t갯수 \n-------------')
for a, b, c in zip(edges[1:], edges[:-1], freq):
    print(f'{b}~{a}\t {c}')
````
