````python
import numpy as np
````

````python
names = np.array( ['홍길동','이순신','임꺽정','정도전','이황', '이이', '김철수', '김영희', '김세종','박문수'] )
kor = np.array( [60,55,70,30,20,90,70,20,90,15 ] )
kor
````

1. 국어점수가 가장큰값을 구하시오. (아래와 같이 출력)  
김철수 90  
김세종 90  

````python
res = np.where(kor==np.max(kor))

for a, b in zip(names[res], kor[res]):
    print(a, b)
````

2. 국어점수가 80점 이상인 값을 구하시오.

````python
kor[kor>=80]
````

3. 국어점수가 50점이상 80이하인 값을 구하시요

````python
kor[(80>=kor) & (kor>=50)]
````

4. 국어점수의 평균을 구하시요.

````python
np.average(kor)
````

5. 국어점수의 총합을 구하시요.

````python
np.sum(kor)
````

6. 국어점수의 표준편차를 구하시요

````python
np.std(kor)
````

7. 국어점수 top5를 구하시요

````python
np.sort(kor)[-1:-6:-1]
````

8. 국어점수가 40점이하인 데이터를 삭제하시요(실데이터 적용X)

````python
kor_delete = np.delete(kor, np.where(kor<=40))
print(kor_delete)
````

9. 국어점수가 50점이상인 데이터에 가산점 10%를 
미만인경우는 20%를 부여하시요

````python
np.where(kor>=50, kor*1.1, kor*1.2)
````

10. 국어점수가 70점이상이면 '합격', 아니면 '불합격' 으로
아래와 같이 출력하시요
```
점수 합격여부
------------------
60 '불합격'
55 '불합격'
...

````python
pass_fail = np.where(kor>=70, '합격', '불합격')

print('점수 \t 합격여부 \n----------------')
for a, b in zip(kor, pass_fail):
    print(f'{a} \t {b}')
````

11.각 데이터 갯수를 구하시요
```
값  갯수
------------
60 1
55 1
70 2

````python
values, counts = np.unique(kor, return_counts=True)

print('값 \t갯수 \n-------------')
for a, b in zip(values, counts):
    print(f'{a}\t {b}')
````

12. 국어점수의 범위(0~50,50~70,70~100)별 갯수를 구하시요 
( 범위, 0~50,50~70,70~100)

````python
bins = [0, 50, 70, 100]
freq, edges = np.histogram(kor, bins=bins)

print('범위 \t갯수 \n-------------')
for a, b, c in zip(edges[1:], edges[:-1], freq):
    print(f'{b}~{a}\t {c}')
````
