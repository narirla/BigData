```text
10. 국어점수가 70점이상이면 '합격', 아니면 '불합격' 으로
아래와 같이 출력하시오

점수 합격여부
------------------
60 '불합격'
55 '불합격'
...

````python
pass_fail = np.where(kor>=70,'합격','불합격')

print('점수', '합격여부', sep="\t")
print('-'*20)
for a, b in zip(kor, pass_fail):
    print(f'{a} \t {b}')
````

```text
11.각 데이터 갯수를 구하시오
값  갯수
------------
60 1
55 1
70 2

````python
values, counts = np.unique(kor, return_counts=True)

print('값', '갯수', sep="\t")
print('-'*15)
for a,b in zip(values,counts):
    print(f'{a}\t{b}')
````

```text
12. 국어점수의 범위(0~50,50~70,70~100)별 돗수를 구하시오 
( 범위, 0~50,50~70,70~100)


````python
bins = [0,50,70,100]
freq, edges = np.histogram(kor, bins=bins)

print('범위', '갯수', sep="\t")
print('-'*20)
for a,b,c in zip(edges[1:], edges[:-1], freq):
    print(f'{b}~{a}\t {c}')
````
