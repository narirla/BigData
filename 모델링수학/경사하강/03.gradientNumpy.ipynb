```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# x = np.array( [1,2,3] )
# y = np.array( [1,2,3] )
# w = -1
def mse(x, y, w):
    hx = w*x # -1*[1,2,3] = [-1,-2,-3]
    # [-1,-2,-3] - [1,2,3] = [-2, -4, -6] = [4,16,36]/3
    c = np.sum( (hx - y)**2 )/len(x)  # 요소별 연산을 하기 때문에 반복문 돌릴 필요 X
    return c
```

```python
def gradient( x, y , w):
    c = np.sum( (w*x-y)*x*2  )
    return c/len(x)
```

```python
x_data = np.array( [1,2,3] )
y_data = np.array( [1,2,3] )
```

```python
mse( x_data, y_data, -1)
```

```python
w = 10
for epoch in range(200):
    c = mse( x_data, y_data, w)
    g = gradient( x_data,y_data, w)
    print( epoch+1, "w=", w,"cost=", c, '기울기:', g)
    w = w-0.1*g
print("w=", w)
```
