```python
import numpy as np

cars = pd.read_csv('cars.csv', encoding='euc-kr')
cars
```

```python
x = cars['speed'].values
y = cars['dist'].values
```

1. 회귀 계수값을 구하시요( w, b)

```python
w = 10
b = 10
epochs= 5000 # 반복 횟수
learning_rate = 0.001
n = len(x)
```

```python
for i in range(epochs):
    hx = w*x + b
    cost = np.sum( (hx-y)**2)/n
    gradienW = np.sum((w*x+b-y )*2*x )/n
    gradienB = np.sum((w*x+b-y )*2 )/n
    w = w-learning_rate*gradienW
    b = b-learning_rate*gradienB
    print('cost:',cost,'w:',w,'b:',b)

print('최종w', w, '최종b', b)
```

2. 자동차속도가 10인경우 예측값을 구하시요

```python
w*10 + b
```

3. 자동차속도X축, 제동거리 y축 scatter를 그리시요

```python
import matplotlib.pyplot as plt

plt.scatter(x, y)       
plt.xlabel("Speed")      
plt.ylabel("Dist")       
plt.show()
```

4. 자동차속도에 따른 예측선을 scatter위에 그리시요.

```python
plt.scatter(x,y)
pred = w*x + b
plt.plot(x, pred, color='green')
plt.show()
```
