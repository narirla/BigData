```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cars = pd.read_csv('cars.csv', encoding='euc-kr')
cars
```

```python
x = cars['speed'].values
y = cars['dist'].values
```

```python
w = np.random.randn()
b = 0
epochs = 5000
learning_rate = 0.001
n = len( x)
```

```python
# 01. 회귀 계수값을 구하시요( w, b)

loss = []
for i in range( epochs):
    hx = w*x + b
    cost = np.sum( (hx-y)**2)/n
    loss.append( cost)
    gradienW = np.sum((w*x+b-y )*2*x )/n
    gradienB = np.sum((w*x+b-y )*2 )/n
    w = w-learning_rate*gradienW
    b = b-learning_rate*gradienB
    # print('cost:',cost,'w:',w,'b:',b)

print('최종w', w, '최종b', b)
```

```python
plt.plot( loss )
plt.show()
```

```python
# 02. 자동차속도가 10인경우 예측값을 구하시요
speed_10 = 10
pred_10 = speed_10 * w + b
print("속도 10일때 제동거리:", pred_10)
```

```python
# 03. 자동차속도X축, 제동거리 y축 scatter를 그리시요
# 04. 자동차속도에 따른 예측선을 scatter위에 그리시요.

pred = x*w + b
plt.scatter( x, y)
plt.plot( x, pred, color='red')
plt.show()
```
