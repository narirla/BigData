#  MySQL 저장 프로시저(Stored Procedure) 기초 문법 정리

---

## 📘 1. 저장 프로시저(Stored Procedure)란?

> **저장 프로시저(Stored Procedure)** 는 자주 사용하는 SQL 구문들을 미리 저장해두고,  
> 필요할 때마다 호출해서 실행할 수 있는 **데이터베이스 내의 함수(프로그램)** 입니다.

✅ **장점**
- SQL 코드 재사용 가능
- 보안성 향상 (직접 쿼리 노출 방지)
- 트랜잭션 관리 용이
- 복잡한 로직을 서버 측에서 처리 → 클라이언트 부하 감소

---

##  2. 기본 문법 구조

```sql
DELIMITER $$

CREATE PROCEDURE 프로시저이름 ( [IN|OUT|INOUT] 매개변수명 데이터형, ... )
BEGIN
    -- SQL 문장들
END $$

DELIMITER ;


| 구분        | 설명                         |
| --------- | -------------------------- |
| **IN**    | 입력용 매개변수 (호출 시 값 전달만 가능)   |
| **OUT**   | 출력용 매개변수 (프로시저 내에서 결과를 반환) |
| **INOUT** | 입력 및 출력 모두 가능              |


```sql

CREATE PROCEDURE show_all_users()
BEGIN
    SELECT * FROM users;
END


```sql
호출
CALL show_all_users();

```sql

CREATE PROCEDURE get_user_by_id(IN user_id INT)
BEGIN
    SELECT * FROM users WHERE id = user_id;
END



CALL get_user_by_id(3);

```sql


CREATE PROCEDURE get_user_name(IN user_id INT, OUT user_name VARCHAR(50))
BEGIN
    SELECT name INTO user_name FROM users WHERE id = user_id;
END

CALL get_user_name(3, @name);
SELECT @name;



``` sql
-- 여러개 결과를 담으려면
CREATE PROCEDURE get_user_name(IN user_id INT, OUT user_name VARCHAR(200))
BEGIN
    SELECT GROUP_CONCAT(name)
    INTO user_name
    FROM users
    WHERE id = user_id;
END;


### 제어문

```sql
IF 조건식 THEN
    SQL문;
ELSEIF 조건식 THEN
    SQL문;
ELSE
    SQL문;
END IF;


```sql
CASE 변수
    WHEN 값1 THEN SQL문;
    WHEN 값2 THEN SQL문;
    ELSE SQL문;
END CASE;


### 반복문

```sql
WHILE 조건식 DO
    SQL문;
END WHILE;



```sql

CREATE PROCEDURE test_if(IN n INT, OUT msg VARCHAR(50))
BEGIN
    IF n > 0 THEN
        SET msg = '양수입니다';
    ELSEIF n < 0 THEN
        SET msg = '음수입니다';
    ELSE
        SET msg = '0입니다';
    END IF;
END 


CALL test_if(5, @result);
SELECT @result;


```sql

CREATE PROCEDURE test_if1(IN n INT)
BEGIN
    DECLARE msg VARCHAR(50);  -- 변수 선언

    IF n > 0 THEN
        SET msg = '양수입니다';
    ELSEIF n < 0 THEN
        SET msg = '음수입니다';
    ELSE
        SET msg = '0입니다';
    END IF;

    SELECT msg AS 결과;   -- 결과 출력
END


CALL test_if1(20);

```sql

CREATE PROCEDURE get_table_data(IN p_table VARCHAR(20))
BEGIN
    -- 조건에 따라 SQL 생성
    IF p_table = 'emp' THEN
        SELECT * FROM emp;
    ELSEIF p_table = 'student' THEN
        SELECT * FROM student;
    END IF;

END


```sql

CREATE PROCEDURE get_dept_name_by_ename(IN p_ename VARCHAR(50))
BEGIN
    DECLARE v_deptno INT;

    -- 1) 사원 이름으로 deptno 조회
    SELECT deptno INTO v_deptno
    FROM emp
    WHERE ename = p_ename;

    -- 2) deptno 로 부서명 조회
    SELECT *
    FROM dept
    WHERE deptno = v_deptno;

END


call get_dept_name_by_ename( 'JAMES');


```sql

CREATE FUNCTION add_two_numbers(a INT, b INT)
RETURNS INT

BEGIN
    RETURN a + b;
END 


SELECT add_two_numbers(5, 7);
select add_two_numbers(sal, 1) from emp;


```sql

CREATE FUNCTION get_half(p_date DATE)
RETURNS INT
BEGIN
    DECLARE h INT;

    -- MONTH() 함수로 월 구한 뒤 6월 이하면 1, 아니면 2
    IF MONTH(p_date) <= 6 THEN
        SET h = 1;
    ELSE
        SET h = 2;
    END IF;

    RETURN h;
END 


select get_half(now());
select get_half( hiredate ) from emp;


```sql

-- 프로시저 목록조회
SHOW PROCEDURE STATUS;
-- 프로시저 내용 확인
SHOW CREATE PROCEDURE 프로시저이름;
-- 프로시저 삭제
DROP PROCEDURE 프로시저이름;

call insertproc('테스트',20,'2015-10-11');

select * from student;

call select_example( 20);


call get_user_name(20, @name);
select @name;

CALL test_if(-5, @result);
SELECT @result;

CALL test_if1(20);

call get_dept_name_by_ename( 'JAMES');


call get_table_data('student');

SELECT add_two_numbers(5, 7);

select add_two_numbers(sal, 1) from emp;

select get_half(now());
select get_half( hiredate ) from emp;
