#### 2. 주민등록연앙인구

```python
# 멀티 인덱스 제거
df_1 = pd.read_csv("C:/final_project/data/raw/시군구_성_연령_5세_별_주민등록연앙인구_20260126141321.csv", header=1)
df_2 = pd.read_csv("C:/final_project/data/raw/시군구_성_연령_5세_별_주민등록연앙인구_2023__20260126141435.csv", header=1)
```

```python
df_1["행정구역(시군구)별"] = df_1["행정구역(시군구)별"].replace({
    "강원도": "강원특별자치도",
    "전라북도": "전북특별자치도"
})  # 값 변경
df_1
```

```python
df_2
```

```python
df_all = pd.concat([df_1, df_2], ignore_index=True)
df_all.rename(columns={
    "시점" : "연도", 
    "행정구역(시군구)별" : "행정구역", 
    "계" : "연앙인구수(명)"
    }, inplace=True)    # 컬럼명 변경
df_all["연앙인구수(천명)"] = df_all["연앙인구수(명)"] / 1000    # 천명 당 연앙인구수 컬럼 추가
df_all
```

```python
df_all.to_csv("C:/final_project/data/processed/주민등록연앙인구_2015-2024.csv", index=False, encoding="utf-8-sig")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
df_z = pd.read_csv('C:/final_project/busan-ai-agent/analysis/notebooks/data/busan_youth_zscore_only.csv')
df_z.head()
```

```python
x_col = "상용월평균급여(원)_z"
y_col = "청년_30_39_증감률(%)_z"

plt.figure(figsize=(7, 5))
sns.regplot(data=df_z, x=x_col, y=y_col, scatter_kws={"alpha" : 0.7}, line_kws={"color": "red"}, ci=None)
plt.title("평균 급여 vs 30–39세 청년 인구 증감률")
plt.xlabel("평균 급여")
plt.ylabel("30–39세 청년 인구 증감률")
plt.show()

# === 해석 방법 ===
# 점 구름 중심이 대각선 -> 선형
# 추세선이 거의 수평 -> 영향 약함
# 완전 곡선 -> 선형 X
```

```python
# lowess 곡선 추가 (비선형 감지)
fig, axes = plt.subplots(1, 2, figsize=(14, 5), sharey=True)

sns.regplot(data=df_z, x=x_col, y="청년_19_29_증감률(%)_z", lowess=True, scatter_kws={"alpha": 0.6}, ci=None, ax=axes[0])
axes[0].set_title("19–29세")

sns.regplot(data=df_z, x=x_col, y="청년_30_39_증감률(%)_z", lowess=True, scatter_kws={"alpha": 0.6}, ci=None, ax=axes[1])
axes[1].set_title("30–39세")

plt.suptitle("평균 급여와 청년 인구 증감률 (연령대 비교)")
plt.show()

# === 해석 방법 ===
# lowess가 직선과 비슷 -> 선형 충분
# S자 / 꺾임 -> 비선형 의심
# 특정 구간만 반응 -> 임계점 존재
```

```python
df_z["region_type"] = "비수도권"
df_z.loc[df_z["행정구역"].isin(["서울특별시", "경기도", "인천광역시"]), "region_type"] = "수도권"
df_z.loc[df_z["행정구역"] == "세종특별자치시", "region_type"] = "세종"


plt.figure(figsize=(7, 5))
sns.scatterplot(data=df_z, x=x_col, y=y_col, hue="region_type", palette={"수도권":"red", "세종":"blue", "비수도권":"gray"}, alpha=0.7)
sns.regplot(data=df_z[df_z["region_type"] == "비수도권"], x=x_col, y=y_col, scatter=False, color="black", ci=None)
plt.title("수도권 vs 세종 vs 비수도권")
plt.xlabel("평균 급여")
plt.ylabel("30–39세 청년 인구 증감률")
plt.show()
```

```python
plt.figure(figsize=(10, 7))

sns.scatterplot(data=df_z, x=x_col, y=y_col, hue="행정구역", palette="tab20", s=100, alpha=0.8)
sns.regplot(data=df_z, x=x_col, y=y_col, scatter=False, color="black",ci=None)

for _, row in df_z.iterrows():
    plt.text(row[x_col], row[y_col], row["행정구역"], fontsize=8, alpha=0.9)

plt.title("행정구역별 평균 급여 vs 30–39세 청년 인구 증감률")
plt.xlabel("평균 급여")
plt.ylabel("30–39세 청년 인구 증감률")
plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()
```
