2015~2019년까지 세종시 데이터 미분류로 인한 결측치 보충 작업

```python
# 1. 채워야 할 컬럼 목록 정의 (숫자 데이터만)
target_cols = ['상용월평균급여(원)', '상용월급여액(원)', '상용정액급여(원)', '상용초과급여(원)', '상용특별급여(원)']

# 2. 보정 비율 계산 및 결측치 채우기 함수
def fill_missing_sejong(sejongdf, cols):
    df_filled = sejongdf.copy()
    
    # 충남 데이터 미리 뽑아두기 (속도 향상)
    chungnam_df = sejongdf[sejongdf['행정구역'] == '충청남도'].set_index('연도')
    
    for col in cols:
        # (1) 2020년 이후 데이터로 '세종 vs 충남' 비율 계산
        mean_chungnam = sejongdf[(sejongdf['행정구역'] == '충청남도') & (sejongdf['연도'] >= 2020)][col].mean()
        mean_sejong = sejongdf[(sejongdf['행정구역'] == '세종특별자치시') & (sejongdf['연도'] >= 2020)][col].mean()
        
        ratio = mean_sejong / mean_chungnam
        print(f"[{col}] 보정 비율: {ratio:.4f}")  # 비율 확인용 출력

        # (2) 세종시의 빈 값 찾아서 채우기
        # 조건: 세종시이면서 + 해당 컬럼이 비어있는 경우
        mask = (df_filled['행정구역'] == '세종특별자치시') & (df_filled[col].isna())
        
        # 빈 곳의 '연도'를 가져와서 -> 충남의 해당 연도 값 * 비율 적용
        # map 함수를 쓰면 반복문 없이 한 번에 처리가 가능합니다
        df_filled.loc[mask, col] = df_filled.loc[mask, '연도'].map(
            lambda y: chungnam_df.loc[y, col] * ratio if y in chungnam_df.index else None
        )
        
    return df_filled

# 실행
df = round(fill_missing_sejong(sejongdf, target_cols),1)

# 결과 확인 (2015년 데이터가 잘 들어갔는지)
print(df[df['행정구역'] == '세종특별자치시'].head())
```
