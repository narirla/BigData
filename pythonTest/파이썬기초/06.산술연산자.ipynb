### 지수 연산자

```python
print(2**2)
print(2**3)
```

```python
a = 5
rst = a ** 2
print(rst)
```

### 곱하기 연산자

```python
a = 5
rst = a * 2
print(rst)
```

### 나누기 연산자

```python
a = 5
rst = a / 2
print(rst)
```

### 나머지 연산자

```python
a = 5
rst = a % 3
print(rst)
```

### 몫 연산자

```python
a = 5
rst = a // 3
print(rst)
```

### 더하기 연산자

```python
a = 5
rst = a + 3
print(rst)
```

### 빼기 연산자

```python
a = 5
rst = a - 3
print(rst)
```

```python
rst = 3+2**2*2-1  #우선 순위가 좌에서 우로 감
print(rst)
```
