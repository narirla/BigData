1. 1cm 가 0.393701 인치인 경우 7cm가 몇인치 인지 구하시오
2. 반지름이 5이고 높이가 10인 원기둥의 부피를 구하시오
(원기둥의부피: 원의 면적 X 높이)
3. 폭이 10이고 높이가 5인 사각형의 면적을 구하시오
4. 정수가 5인 경우 짝수인지 홀수인지 구하시오
5. fruit = ["사과","포도","홍시"]에서 "홍시"요소가 있는지 확인하시오.
결과 값 '있다' 또는 '없다'

```python
# 1
cm = 7
inch = 0.393701
result = cm*inch
print(result)
```

```python
# 2
r = 5
pi = 3.14
h = 10
v = pi*r**2*h
print(v)
```

```python
# 3
w = 10
h = 5
result = w*h
print(result)
```

```python
# 4
n = 5
rst = '짝수' if n%2 == 0 else '홀수'
print(rst)
```

```python
# 5
fruit = ["사과","포도","홍시"]
rst = '있다' if '홍시' in fruit else '없다'
print(rst)
```
