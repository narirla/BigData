### 딕셔너리 멤버 함수

```python
d = {'aa':10 ,'bb':20, 'cc':30}
print(d.keys())
print(d.values())
print(d.items())
```

```python
d = {'aa':10 ,'bb':20, 'cc':30}
print(d['aa'])
print(d.get('aa'))
#print(d['ff'])
print(d.get('ff',100)) 
```

```python
d = {'aa':10 ,'bb':20, 'cc':30}
d.pop('aa')
print(d)
```

```python
d = {'aa':10 ,'bb':20, 'cc':30}
d['aa'] = 100 # 키가 있으면 수정
d['dd'] = 200 # 키가 없으면 추가
print(d)
```

#### 세트 멤버함수

```python
s = {10,20,30,40,50}
s.add(60)
s.remove(20)
print(s)
```

```python
s = {10,20,30,40,50,20,20} #중복 데이터 자동제거
print(s) 
```

```python
s1 = { 10,20,30 }
s2 = { 20,30,40,50 }
print( s1.intersection(s2) ) # 교집합
print( s1 & s2 )
print( s1.union(s2) ) #합집합
print( s1 | s2 )
print(s1.difference(s2) ) #차집합
print( s1-s2 )
print( s2.difference(s1) )
print ( s1 - s2)
```
