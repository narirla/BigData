관계연산자(산술연산자 보다 우선 순위 낮다) : 연산 결과는 bool 타입
- >  (크다)
- >= (크거나 같다)
- <  (작다)
- <= (작거나 같다)
- == (같다)
- != (같지 않다)

```python
a = -5
rst = a > 0
print(rst)
```

```python
a = -5 
rst = a >= 0
print(rst)
```

```python
a = 3
rst = a == 0
print(rst)
```

```python
a = 3
rst = a != 0
print(rst)
```

```python
a = 3
# rst = 0 <a <10
rst = 0 <= a <= 10
print(rst)
```
