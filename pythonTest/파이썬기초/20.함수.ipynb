```python
def hap(a,b):
    rst = a+b
    return rst

def gop(a,b):
    return a*b

def minus(a,b):
    return a-b

def circle(r):
    return 3.14*r**2

# def ispass(jumsu):
#     if jumsu>=70:
#         return '합격'
#     else:
#         return '불합격'

def ispass(jumsu):
    return '합격' if jumsu>=70 else '불합격'

def intList(num):
    return [ n for n in range(1,num+1) ]
```

```python
rst = hap(10,20)
print(rst)
```

```python
rst = gop(10,20)
print(rst)
```

```python
rst = minus(20,10)
print(rst)
```

```python
rst = circle(10)
print(rst)
```

1. cm 를 인자로 받아 인치를 반환하는 함수를 만드시오.(cm = 0.3937inch)
2. 반지름과 높이를 인자로 받아 원기둥의 부피를 반환하는 함수를 만드시오.

```python
def cmToInch(cm):
    return cm*0.3937

def cylinder(r,h):
    return 3.14*r**2*h
```

```python
rst = cmToInch(10)
print(rst)
```

```python
rst = cylinder(5,10)
print(rst)
```
