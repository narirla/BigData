```python
# 입력한 점수가 70점 이상이면 '합격' 아니면 '불합격' 출력하시오

jumsu = input('점수:')
jumsu = int(jumsu)
if jumsu >= 70:
    print('합격')

else:
    print('불합격')  
```

퀴즈 하나의 점수를 입력받아 
90~100 'A'
80~89 'B'
70~79 'C'
60~69 'D'
나머지 'F'


```python
jumsu = input('점수:')
jumsu = int(jumsu)
if 90 <= jumsu <=100 :
    print('A')
elif 80 <= jumsu <90:
    print('B')
elif 70 <= jumsu <80:
    print('C')
elif 60 <= jumsu <70:
    print('D')
else:
    print('F')  
```
