### 퀴즈: 세금 3.3%를 제한 실수령액 리스트를 구하시오

```python
salary = [1000,2000,3000,4000,5000]
rst = [n*(1-0.033) for n in salary]
print(rst)
```

```python
# if 문이 들어가면 두번째 n으로 들어감
rst = [ n for n in range(1,6) if n>=3]
print(rst)
```

## 퀴즈 1~10까지 짝수 리스트를 만드시오

```python
rst = [n for n in range(1,11) if n%2==0]
print(rst)
```

## 퀴즈: '이' 글자가 있는 이름에 대한 리스트를 구하시오

```python
names = ['김철수','홍길동','이순신','이황','이이']
rst = [ n for n in names if '이' in n ]
print(rst)
```

```python
# zip: 동일한 인덱스끼리 튜플로 만들어줌
a = ['aa','bb','cc']
b = [10,20,30]
print(list(zip(a,b)))
print(dict(zip(a,b)))
```

```python
for n in zip(a,b):
    print(n)
```

```python
# enumerate: 인덱스랑 값을 튜플로 만들어줌
for n in enumerate(a):
    print(n)
```

```python
for i,v in enumerate(a):
    print(i,v)
```
