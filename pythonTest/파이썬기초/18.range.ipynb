## 리스트를 만들어주는 함수
- range(시작값,끝값,증가치)
- 시작값 <= 리스트 < 끝값

```python
list( range(1,10,1) ) #1 <= 리스트< 10
```

```python
list( range(1,10,2) )
```

```python
list( range(1,10) )
```

```python
list( range(10) ) #0<=리스트<10
```

```python
for n in range(1,5):
    print(n)
```

### break : 반복문 강제 탈출

```python
for n in range(1,11):
    if n == 4:
        break
    print(n)
```

### continue를 만나게 되면 print 부분이 안찍힘

```python
for n in range(1,11):
    if n == 4:
        continue
    print(n)
```

```python
## 1~10까지의 정수 중 3의 배수를 제외하고 출력하시오(단 continue 사용)
for n in range(1,11):
    if n % 3== 0 :
        continue
    print(n)

```
