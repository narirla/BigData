### 문자열, 리스트, 튜플, 세트, 딕셔너리

```python
s1 = "abc\
def"
s2 = 'abc\
def'
s3 = """abc
    def"""
s4 = '''abc
    def'''
print(s1)
print(type(s1))
print(s2)
print(type(s2))
print(s3)
print(type(s3))
print(s4)
print(type(s4))
```

```python
s1 = "abc\ndef"
s2 = 'abc\ndef'
print(s1)
print(s2)
```

### 2.리스트 타입 [  ]: 변경 가능(mutable): 추가, 수정, 삭제 ok

```python
# a = 10 [10, 1.1, 'abc'] 거의 사용 X 
myList = [10, 20, 30, 40, 50]
myList1 = [1.1, 2.2, 3.3]
myList2 = ['aa','bb','cc']
print( myList)
print( type(myList))
print( myList1)
print( myList2)
```

### 3. 튜플 (  ): 변경불가( immutable): 추가, 수정, 삭제 X
