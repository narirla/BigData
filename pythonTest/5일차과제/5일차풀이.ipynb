````python
class MyObj:
    count = 0
    def __init__(self):
        MyObj.count +=1

    @staticmethod
    def getCount():
        return MyObj.count

Obj1 = MyObj()
Obj2 = MyObj()
Obj3 = MyObj()

# print(f"생성된 객체 수 :{MyObj.count}")
print(f"생성된 객체 수 :{MyObj.getCount() }")
````

````python
#2. 차를 상속받는 자전차 클래스 생성

class 차:
    def __init__(self, 바퀴, 가격):
        self.바퀴 = 바퀴
        self.가격 = 가격

class 자전차(차):
    def __init__(self, 바퀴, 가격, 구동계):
        super().__init__(바퀴, 가격)
        self.구동계 = 구동계

bicycle = 자전차(2, 100, "시마노")
print(bicycle.바퀴) 
print(bicycle.가격) 
print(bicycle.구동계)
````

3. 아래 코드가 동작하도록  Book 클래스를 완성하시요. 
class Book: 
    def __init__(self, title, author): 
        self.title = title 
        self.author = author 
        
book = Book("데미안", "헤르만 헤세") 
print(book)   
# 책제목:데미안 저자: 헤르만 헤세

````python
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author

    def __repr__(self):
        return f"책제목:{self.title} 저자:{self.author}"

book = Book("데미안", "헤르만헤세")
print(book) # book.__repr__() , book.__str__()
````
