class 차:
    def __init__(self, 바퀴, 가격):
        self.바퀴 = 바퀴
        self.가격 = 가격
bicycle = 자전차(2, 100, "시마노")
print(bicycle.바퀴) # 2

print(bicycle.구동계) # 시마노

````python
class 차:
    def __init__(self, 바퀴, 가격):
        self.바퀴 = 바퀴
        self.가격 = 가격


class 자전차(차):
    def __init__(self, 바퀴, 가격, 구동계):
        super().__init__(바퀴, 가격)
        self.구동계 = 구동계

    def show(self):
        print('바퀴',self.바퀴)        
        print('가격',self.가격)        
        print('구동계',self.구동계)
                
bicyle = 자전차(2,100, "시마노")
print(bicyle.바퀴)
print(bicyle.가격)
print(bicyle.구동계)
````

class Test:

    def show(self):

obj1 = Test( 10,20)

````python
# 그림을 그려볼 것(메모리)
class Test:
    def __init__(self, a,b):
        self.a = a
        self.b = b

    def setData(self, x,y):
        self.a = x
        self.b = y
        
    def show(self):
        print( self.a, self.b)

    def __add__(self, other):
        return  (self.a + other.a, self.b + other.b)

obj1 = Test( 10,20)
obj2 = Test(100,200)
rst = obj1 + obj2
print( rst )
````

5. 두 학생의 점수가 같으면 ==이 True가 되도록 ,한 학생의 점수가 더 작으면 <이 True가
되도록 구현하세요.
class Student:
    def __init__(self, name, score):
        self.name = name
        self.score = score

# __eq__, __lt__ 구현
s1 = Student("민수", 85)
s2 = Student("지연", 90)
s3 = Student("철수", 85)
print(s1 == s3) # True
print(s1 < s2) # True

````python
class Student:
    def __init__(self, name, score):
        self.name = name
        self.score = score

    def __eq__(self, other):
        if self.score == other.score:
            return True
        else:
            return False
        
    def __lt__(self, other):
        if self.score < other.score:
            return True
        else:
            return False

s1 = Student("민수", 85)
s2 = Student("지연", 90)
s3 = Student("철수", 85)
print(s1 == s3) # True
print(s1 < s2) # True
````
