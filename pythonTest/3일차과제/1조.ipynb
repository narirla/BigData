### 1. 점수를 인자로 받아 해당 학점을 구하는 함수를 만드시요.
점수   학점
81~100 A
61~80  B
41~60  C


```python
def 학점(jumsu):
    if  81 <= jumsu <= 100:
        return 'A'
    elif 61 <= jumsu <= 80:
        return 'B'
    elif 41 <= jumsu <= 60:
        return 'c'
    else:
        return 'F' 
```

### 2. 키와 몸무게를 인자로 받아 비만도를 구하는 함수를 만드시요
표준체중(kg)=(신장(cm)-100)×0.85
비만도(%)=현재체중/표준체중(%)×100
비만도가90%이하
저체중,
90초과~110%
정상,
110초과~120%
과체중,
120%초과
비만

```python
def 비만도(h, w):

    표준체중 = (h - 100)*0.85
    비만도 = 현재체중/표준체중*100
    
    if 비만도 <= 90:
        return '저체중'
    elif 90<비만도<=110:
        return '정상'
    elif 110<비만도<=120:
        return '과체중'
    else:
        return '비만도'
```

### 4. 2개의 정수를 인자로 받아 절대값의 합을 반환하는 함수를 만드시요

```python
def 절대값(a,b):
    a = -a if a < 0 else a
    b = -b if b < 0 else b
    return a + b

print(절대값(5, -3))
```

### 6. 년도를 인자로 받아 윤년여부를 반환하는 함수를 만드시요.
윤년의 조건
1-1) 4로 나눠 떨어지지만 100으로 나눠 떨어지지 않아야 한다 또는
1-2) 400 으로 나눠 떨어지면 윤년임

```python
def 윤년(year):
    return '윤년' if year%4 == 0 and year%100!=0 or year%400==0 else '윤년아님'
```

```python
print(윤년(2024))
```
