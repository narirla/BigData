1. 점수를 인자로 받아 해당 학점을 구하는 함수를 만드시요. 
점수 
학점 
81~100 A 
61~80 
B 
41~60 
C
나머지 'F'

````python
def grade(score) :
    if 81 <= score <= 100 :
        return 'A'
    elif 61 <= score <= 80 :
        return 'B'
    elif 41 <= score <= 60 :
        return 'C'
    else :
        return 'F'
````

````python
grade( 30)
````

````python
# 2. 키와 몸무게를 인자로 받아 비만도를 구하는 함수

def bmi(height, weight):
    standard_weight = (height - 100) * 0.85
    obesity = (weight / standard_weight) * 100
    if obesity <= 90:
        return "저체중"     
    elif obesity <= 110:
        return "정상"
    elif obesity <= 120:
        return "과체중"
    else:
        return "비만"

print( bmi(180, 70) )
````

````python
# 04. 2개의 정수를 인자로 받아 절대값의 합을 반환하는 함수를 만드시오.
def myabs( a:int, b:int )->int :
    a = -a if a<0 else a
    b = -b if b<0 else b
    return a+b

print(myabs(-3,5))
````

````python
# 6. 년도를 인자로 부여받아 윤년여부를 반환하는 함수
def leap_year(year):
    if year % 4 == 0 and year % 100 != 0 or year % 400 == 0:
        return True
    else:
        return False

print(leap_year(2024))
````

8. 상품가격과 지불액을 인자로 받아 거스름돈을 최소화 하도록 잔돈의 갯수를 반환하는 함수를 만드시요.

````python
def exchange(price,pay) :
    exchange = pay - price
    n500 = exchange // 500
    exchange = exchange % 500
    n100 = exchange // 100
    exchange = exchange % 100
    n50 = exchange // 50
    exchange = exchange % 50
    n10 = exchange // 10
    return n500, n100, n50, n10
    # return f'500원:{n500},100원:{n100},50원:{n50},10원:{n10}'
rst = exchange( 1870, 2000)
print( f'500원짜리:{rst[0]},100원짜리:{rst[1]},50원:{rst[2]},10원:{rst[3]}' )
````

9. 하나의 정수를 인자로 받아 해당 정수의 약수를 반환하는 함수를 만드시요
예) 10 -> [1,2,5,10]

````python
def yaksu( num):
    return [n for n in range( 1, num+1) if num%n==0 ]

yaksu(100)
````

````python
def yaksu(a:int) :
    s = []
    for n in range(1, a+1) :
        if a%n==0 :
            s.append(n)
    return s

print(yaksu(100))
````
