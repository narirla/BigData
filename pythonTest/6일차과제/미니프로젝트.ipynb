```python
class 성적관리:
    def __init__(self, 이름, 국어, 영어, 수학):
        self.이름 = 이름
        self.국어 = 국어
        self.영어 = 영어
        self.수학 = 수학

    def show(self):
        print(f'이름 = {self.이름}')
        print(f'국어 = {self.국어}')
        print(f'영어 = {self.영어}')
        print(f'수학 = {self.수학}')

입력 = 성적관리('홍길동','70','80','90')
print(입력)
```
