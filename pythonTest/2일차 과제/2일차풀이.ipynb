# 11. 키와 몸무게를 입력받아 비만도를 구하고 결과를 출력하시오.

````python
신장 = float( input('신장 : (신장은 100 이상입니다.)') )
현재체중 = float( input('몸무게 : ') )
# 표준체중(kg)=(신장(cm)-100)×0.85
# 비만도(%)=현재체중/표준체중(%)×100
표준체중 = (신장-100) * 0.85
비만도 = 현재체중/표준체중*100

if 비만도<=90 :
    print('저체중')
elif 90<비만도<=110 :
    print('정상')
elif 110<비만도<=120 :
    print('과제중')
else :
    print('비만')
````

7. 상품가격과 지불액을 입력받아 거스름돈을 최소화 하도록 잔돈의 갯수를 구하시요
(잔돈, 500, 100, 50, 10 )
예) 상품가격 1320원 지불액 2000원 인경우
잔돈은 680원 500원 1개 100원1개 50원 1개 10원3개


````python
product = 1320
pay = 2000
exchange = pay-product

n500 = exchange//500
exchange = exchange%500

n100 = exchange//100
exchange = exchange%100

n50 = exchange//50
exchange = exchange%50

n10 = exchange//10

print( f'500원갯수:{n500} 100원갯수:{n100} 50원갯수:{n50} 10원갯수:{n10}')
````

````python
#17. 상품가격과 지불액을 입력하여 거스름돈을 최소화하시오.
price = int(input("상품 가격을 입력하세요: "))
money = int(input("지불한 금액을 입력하세요: "))

change = money - price

for coin in [500, 100, 50, 10]:
    count = change // coin
    change %= coin 
    print(f"{coin}원 {count}개", end=' ')
````
