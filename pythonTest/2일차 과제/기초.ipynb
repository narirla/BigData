##문제7 for 문을 사용해서 리스트에 저장된 값을 다음과 같이 출력하라.


````python
리스트 = ["김밥", "라면", "튀김"] #변수명을 함수명으로 쓰면 다음 함수가 오류가 뜸
for n in 리스트:
    print('오늘의 메뉴:', n)
````

##문제9
for 문을 사용해서 3 의 배수만을 출력하라.
리스트 = [3, 100, 23, 44,9]


````python
list = [3, 100, 23, 44,9]
for n in list:
    if n % 3 == 0:
        print(n)
````
