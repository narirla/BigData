1. 아래 처럼 동작 하도록 myprint함수를 구현하시요
mylist = [10,20,30]
myprint( mylist, lambda v: print(v) )
결과)
10
20
30
myprint( mylist, lambda v: print(v,end=' ') )
결과) 10 20 30


```python
def myprint(data, key):
    for n in data:
        key(n)

mylist = [10, 20, 30]

myprint(mylist, lambda v:print(v))

myprint( mylist, lambda v:print(v,end=' ') )
```

삼성 = Stock("삼성전자", "005930", 15.79, 1.33, 2.83)
print(삼성.배당수익률) #2.83출력
삼성.show() #종목 ~ 배당수익률 모두 출력

```python
class Stock:
    def __init__(self, 종목명, 종목코드, PER, PBR, 배당수익률):
        self.종목명 = 종목명
        self.종목코드 = 종목코드
        self.PER = PER
        self.PBR = PBR
        self.배당수익률 = 배당수익률

    def show(self):
        print(self.종목명, self.종목코드, self.PER,self.PBR,self.배당수익률 )


삼성 = Stock("삼성전자", "005930", 15.79, 1.33, 2.83)
삼성.show()

print(삼성.배당수익률)
```

3. 아래의 표를 참조하여 3종목에 대해 객체를 생성하고 각각 show함수를 통해 내
용을 출력하시요.
종목명 종목코드 PER PBR 배당수익률
삼성전자 005930 15.79 1.33 2.83
현대차 005380 8.70 0.35 4.27
LG 전자 066570 317.34 0.69 1.37

```python
class Stock:
    def __init__(self):
        self.종목명 = '삼성전자'
        self.종목코드 = '005930'
        self.PER = 15.79
        self.PBR = 1.33
        self.배당수익률 = 2.83
    
    def setData(self, a,b,c,d,e):
        self.종목명 = a
        self.종목코드 = b
        self.PER = c
        self.PBR = d
        self.배당수익률 = e
    
    def show(self):
        print(self.종목명, self.종목코드, self.PER,self.PBR,self.배당수익률 )


삼성전자 = Stock("삼성전자", "005930", 15.79, 1.33, 2.83)
삼성전자.show()

현대차 = Stock('현대차','005380' ,8.70, 0.35, 4.27)
현대차.show()

LG전자 = Stock('LG 전자', '066570', 317.34, 0.69, 1.37)
LG전자.show()
```

4.
Account 클래스를 만든후 객체를 생성하여 저장된 정보를 출력하는 display_info()
메서드를 이용하여 아래 처럼 출력하시요.
은행이름: SC 은행
예금주: 파이썬
계좌번호: 111-11-111111
잔고: 10,000 원
추가)format 내장함수 이용시 thousand 표시됨
num = 123456789
f = format(num, ',')
print(f) # 123,456,789

```python
class Account():
    def __init__(self):
        self.은행이름 = "SC은행"
        self.예금주 = '파이썬'
        self.계좌번호 = '111-11-111111'
        self.잔고 = 10000

    def setData(self, a,b,c,d):
        self.은행이름 = a
        self.예금주 = b
        self.계좌번호 = c
        self.잔고 = d

    def display_info(self):
        print("은행이름 :",self.은행이름)
        print("예금주 :",self.예금주)
        print("계좌번호 :",self.계좌번호)
        print("잔고 :", format(self.잔고,','),"원")

은행 = Account()
은행.display_info()
 
        
```
