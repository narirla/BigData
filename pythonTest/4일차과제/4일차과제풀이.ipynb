````python
# 04. Account 클래스를 만든 후 객체를 생성하여 저장된 정보를 출력하는 display_info() 메서드를 이용하여 아래처럼 출력하시오.

class Account:
    def __init__(self):
        self.은행이름 = 0
        self.예금주 = 0
        self.계좌번호 = 0
        self.잔고 = 0
    
    def setDara(self, 은행이름, 예금주, 계좌번호, 잔고):
        self.은행이름 = 은행이름
        self.예금주 = 예금주
        self.계좌번호 = 계좌번호
        self.잔고 = 잔고
    
    def display_info(self):
        print('은행이름 :', self.은행이름)
        print('예금주 :', self.예금주)
        print('계좌번호 :', self.계좌번호)
        print('잔고 :', format(self.잔고, ','),'원')

scB = Account()
scB.setDara('sc은행', '파이썬', '111-11-111111', 10000)
scB.display_info()
````
