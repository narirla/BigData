1. 아래 처럼 동작 하도록 myprint함수를 구현하시요 
mylist = [10,20,30] 
myprint( mylist, lambda v: print(v) ) 
결과) 
10 
20 
30 
myprint( mylist, lambda v: print(v,end=' ') ) 
결과) 10 20 30 

````python
def myprint(data, key):
    for n in data:
        key(n)

mylist = [10, 20, 30]
myprint(mylist, lambda v : print(v))
myprint(mylist, lambda v : print(v,end=' '))
````

2. 아래 멤버를 가지는 Stock 클래스를 정의하시요 
삼성 = Stock("삼성전자", "005930", 15.79, 1.33, 2.83)  
print(삼성.배당수익률) #2.83출력 
삼성.show() #종목 ~ 배당수익률 모두 출력 
항목 정보 
종목명 삼성전자 
종목코드 005930 
PER 15.79 
PBR 1.33 
배당수익률 2.83

````python
class Stock:
    def __init__(self, 종목명, 종목코드, PER, PBR, 배당수익률):
        self.종목명 = 종목명
        self.종목코드 = 종목코드
        self.PER = PER
        self.PBR = PBR
        self.배당수익률 = 배당수익률

    def show(self):
        print(self.종목명, self.종목코드, self.PER, self.PER, self.배당수익률)

삼성 = Stock("삼성전자", "005930", 15.79, 1.33, 2.83)
print(삼성.배당수익률)
삼성.show()
````

4. 
Account 클래스를 만든후 객체를 생성하여 저장된 정보를 출력하는 display_info() 
메서드를 이용하여 아래 처럼 출력하시요. 
은행이름: SC은행 
예금주: 파이썬 
계좌번호: 111-11-111111 
잔고: 10,000원 
추가)format 내장함수 이용시 thousand 표시됨 
num = 123456789 
f = format(num, ',') 
print(f)  # 123,456,789

````python
class Account:
    def __init__(self):
        self.은행이름 = 'SC은행'
        self.예금주 = '파이썬'
        self.계죄번호 = '111-11-111111'
        self.잔고 = 10000

    def display_info(self):
        print(f"은행이름: {self.은행이름}\n예금주: {self.예금주}\n계죄번호: {self.계죄번호}\n잔고: {format(self.잔고, ',')}원")

SC은행 = Account()
SC은행.display_info()
````
