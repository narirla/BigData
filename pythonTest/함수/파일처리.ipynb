### 쓰기모드

```python
# fp = open( "c:\\my\\test.txt", 'w') 
# fp = open( r"c:\my\test.txt", 'w') #제어문자 무시
fp = open( r"test.txt", 'w') #만약에 text.txt 파일이 있으면 새로 만든다
print(type(fp))
fp.write('hello\n')
fp.write('korea')
fp.close()
```

```python
with open( r"test.txt", 'w') as fp:
    fp.write('hello1\n')
    fp.write('hi korea')
    print(fp.closed)

print(fp.closed)
```

```python
with open( r"test.txt", 'w') as fp:
    fp.write('abcdefghi')
```

```python
with open( r"test.txt", 'w') as fp:
    fp.write('abc\n')
    fp.write('def\n')
    fp.write('ghi\n')
```

```python
with open('test.txt', 'r') as fp:
    # txt = fp.read()
    txt = fp.readline()
    print(txt.strip())
    txt = fp.readline()
    print(txt.strip())
```

```python
with open('test.txt', 'r') as fp:
    # txt = fp.read()
    txt = fp.readline()
    print(txt)
    txt = fp.readline()
    print(txt)
```

```python
with open('test.txt', 'r') as fp: #제일 많이 씀
    for r in fp:
        print(r.strip())
```

### csv 파일 기록

```python
import csv
fp = open('my.csv' ,'w', newline='')
wr = csv.writer(fp)
# print(type(wr))
wr.writerow( ['aa', 10, 20] )
wr.writerow( ['bb', 10, 20] )
wr.writerow( ['cc', 10, 20] )
fp.close()
```

```python
fp = open('my.csv', 'r')
rd = csv.reader(fp)
for a,b,c in rd:
    print(a,b,c)
fp.close()
```
