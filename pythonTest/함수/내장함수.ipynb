## 내장함수 import 필요없음

| 함수                          | 설명               | 예시                                         |
| --------------------------- | ---------------- | ------------------------------------------ |
| `print()`                   | 콘솔에 출력           | `print("Hello")` → `Hello`                 |
| `input()`                   | 사용자 입력 받기        | `name = input("이름: ")`                     |
| `len()`                     | 길이(문자열, 리스트 등)   | `len([1,2,3])` → `3`                       |
| `type()`                    | 자료형 확인           | `type(3.14)` → `<class 'float'>`           |
| `int()`, `float()`, `str()` | 형 변환             | `int("10")` → `10`                         |
| `abs()`                     | 절댓값              | `abs(-5)` → `5`                            |
| `max()`, `min()`            | 최대/최소값           | `max([1,5,3])` → `5`                       |
| `sum()`                     | 합계 계산            | `sum([1,2,3])` → `6`                       |
| `round()`                   | 반올림              | `round(3.14159, 2)` → `3.14`               |
| `sorted()`                  | 정렬된 리스트 반환       | `sorted([3,1,2])` → `[1,2,3]`              |
| `reversed()`                | 역순 반복자 반환        | `list(reversed([1,2,3]))` → `[3,2,1]`      |
| `enumerate()`               | 인덱스와 값을 함께 반복    | `for i, v in enumerate(['a','b']):`        |
| `zip()`                     | 여러 시퀀스를 묶어서 반복   | `list(zip([1,2],[3,4]))` → `[(1,3),(2,4)]` |
| `map()`                     | 함수를 각 원소에 적용     | `map(str, [1,2,3])` → `['1','2','3']`      |
| `filter()`                  | 조건에 맞는 원소만 선택    | `filter(lambda x:x>2,[1,2,3,4])` → `[3,4]` |
| `any()`                     | 하나라도 참인지         | `any([0,1,0])` → `True`                    |
| `all()`                     | 모두 참인지           | `all([1,2,3])` → `True`                    |
| `range()`                   | 숫자 범위 생성         | `range(5)` → `[0,1,2,3,4]`                 |
| `id()`                      | 객체의 고유 주소 반환     | `id(42)`                                   |
| `isinstance()`              | 객체 타입 검사         | `isinstance(3, int)` → `True`              |
| `dir()`                     | 객체의 속성/메서드 보기    | `dir(str)`                                 |
| `help()`                    | 함수/모듈 도움말 보기     | `help(len)`                                |
| `eval()`                    | 문자열을 파이썬 코드로 실행  | `eval("3+4")` → `7`                        |
| `exec()`                    | 여러 줄 코드 실행       | `exec("for i in range(3): print(i)")`      |
| `globals()` / `locals()`    | 전역/지역 변수 딕셔너리 반환 | 변수 디버깅 시 유용                                |
| `chr()`                     | 코드값 → 문자         | `chr(65)` → `'A'`                          |
| `ord()`                     | 문자 → 코드값         | `ord('A')` → `65`                          |
| `bin()`, `oct()`, `hex()`   | 2/8/16진수 문자열 반환  | `bin(10)` → `'0b1010'`                     |
| `divmod()`                  | 몫과 나머지 동시에 반환    | `divmod(7,3)` → `(2,1)`                    |
| `pow()`                     | 거듭제곱             | `pow(2,3)` → `8`                           |
| `sorted()`                  | 정렬된 리스트 반환       | `sorted([5,3,1])` → `[1,3,5]`              |

```python
mylist =[22,11,33]
print(sorted(mylist))

for i,v in enumerate(mylist): #인덱스와 값을 반환
    print(i,v)
```

```python
a = ['aa', 'bb', 'cc']
b = [1,2,3]
c = [10,20,30]
for n in zip(a,b,c): # 동일한 인덱스끼리 튜플로 묵어줌
    print(n)
```

```python
mylist = [10,20,30]
m = map(lambda v:v+2, mylist)
list(m)
```

```python
def fn(v):
    return v + 3
mylist = [10,20,30]
m = map(fn, mylist)
list(m)
```

```python
mylist = ['10','20','30']
m = map(int, mylist)
list(m)
```

```python
mylist = ['10','20','30']
c = [int(n) for n in mylist]
print(c)
m = map(int, mylist)
list(m)
```

```python
mylist = [10,20,30,40,50]
f = filter(lambda v:v>=30, mylist)
list(f)
```
