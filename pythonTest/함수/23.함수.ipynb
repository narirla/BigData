###디폴트 아규먼트

```python
def fn(a=1,b=2):
    print("a=",a)
    print("b=",b)
```

```python
fn(10,20)
```

```python
fn()
```

```python
fn(100)
```

```python
fn(11)
```

```python
fn(b=11) #명시적 호출 (아규먼트)
```

### 여러개 리턴

```python
def fn1():
    return 10,20
```

```python
rst = fn1()
print(rst)
```

```python
a, b = fn1()
print(a,b)
```

```python
def shape(r, h,w):
    return 3.14*r**2*h, h*w
```

```python
rst = shape(3, 10,20)
print(rst)
print(rst[0])
print(rst[1])
```

### 가변인자: 함수 호출 시 아규먼트 제한 없음
- def fn(a)

```python
def fn3( *args):
    print(args)
```

```python
fn3(10,20,30,40,50)
fn3("aa","bb","cc")
```

```python
def circles(*rs):
    # rs = (1,2,3,4,5)
    for r in rs:
        print(3.14*r**2)
```

```python
circles(1,2,3,4,5)
```

```python
def fn4( *args, a=0):
    print('args=' , args)
    print('a=' , a)
```

```python
# fn4(1,2,3,4,5) 모두 가변인자가 받아서 a가 받을게 없어서 오류가 남
```

```python
fn4(1,2,3,4,5)
```

```python
fn4(1,2,3,4,a=5)
```

### 타입명시함수: 3.5버전 이후부터
def fn5(a:인자타입, b:인자타입)-> 반환타입: 

```python
# def hap(a,b):
def hap(a:object, b:object) -> object:
    return a+b
```

```python
rst = hap(10.1, 20.2)
print(round(rst,1))
```

```python
rst = hap('aaa','bbb')
print(rst)
```

```python
# def fn5(a,b):
def fn5(a:int, b:int)->int: 
    return a+b
```

```python
rst = fn5(10,20)
print(rst)
```
