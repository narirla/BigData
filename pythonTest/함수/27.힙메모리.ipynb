### call by reference(shallow copy: 주소만 복사)

```python
my = [10,20,30]
def fn(aa):
    aa[0] = 100
    print('id aa=', id(aa))
    print('aa=', aa)

print('myid', id(my))
fn(my)
print('my=', my)
```

### return

```python
def fn():
    mylist = [10,20,30]
    return mylist

rst = fn()
print(rst)
```

```python
mylist = [10,20,30]
aa = mylist ##shallow copy
print( id(aa), id(mylist))
aa[0] = 100
print(mylist)
```

### deep copy

```python
mylist = [10,20,30]
# aa = [10,20,30] 
aa = mylist.copy() # deep copy
print( id(aa), id(mylist))
aa[0] = 100
print(mylist)
```

### reference count (참조계수 기법): 파이썬 메모리 관리

```python
mylist = [10,20,30]
mylist = "abc" #리스트 객체는 소멸(why? 참조하는 변수가 0)
print(mylist)
```

```python
mylist = [10,20,30]
aa = mylist #리스트에 대한 reference count 2
mylist = "abc" #리스트에 대한 reference count 1
print(mylist)
print(aa)
print(id(mylist),id(aa))
```
