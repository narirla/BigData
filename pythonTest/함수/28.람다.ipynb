## 데이터를 필터링하는 함수

```python
def myfilter( data ):
    return [n for n in data if n>= 30]

def myfilter1( data ):
    return [n for n in data if n<= 20]

def myfilter2( data ):
    return [n for n in data if n== 20 or n==40]
```

```python
mylist = [10,20,30,40,50]
# rst = myfilter(mylist)
rst = myfilter2(mylist)
print(rst)
```

### 일급함수 기능: open closed 기법

```python
def myfilter(data, key):
    return [ n for n in data if key(n)]
```

```python
def fn(v):
    return v>= 30

def fn1(v):
    return v<= 20

mylist = [10,20,30,40,50]
# rst = myfilter(mylist, fn)
rst = myfilter(mylist, fn1)
print(rst)
```

- 함수일반화에 성공(일급함수)
- 필요: 임시 함수 필요함(람다함수)

### 람다함수 문법
- lambda 인자,...:리턴값
- :(return 의미를 포함하고 있기떄문 return키워드 사용X)

```python
# def hap(a,b):
#     return a+b
hap = lambda a,b:a+b
rst = hap(10,20)
print(rst)
```

```python
def myfilter(data, key):
    return [ n for n in data if key(n)]

mylist = [10,20,30,40,50]
# rst = myfilter(mylist, lambda v:v>=30)
rst = myfilter(mylist, lambda v:v<=20)
print(rst)
```

```python
def mymap(data, key):
    return [key(n) for n in data]

mylist = [10,20,30,40,50]
rst = mymap(mylist, lambda v:v+2)
print(rst)
rst = mymap(mylist, lambda v:v+5)
print(rst)
```
