### 일급 함수: 리턴, 인자(아규먼트), 할당, 함수 내에 함수를 선언(데코레이터)

### 할당 : 함수를 변수로 받을 수 있음

```python
def fn():
    print( 'fn call')

aa = fn
print( 'fn',id(fn), 'aa', id(aa))
aa()
```

## 아규먼트 : 함수를 매개변수로 전달하는 것, 함수 호출 시 전달되는 실제 값

```python
def fn():
    print( 'fn call')

def afn(aa):
    print('aa', id(aa))
    aa()

print('fn', id(fn))
afn(fn)
```
