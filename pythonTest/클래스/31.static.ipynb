```python
class Test:
    st = 100 #static 멤버(객체 생성 X), 클래스 변수
    def __init__(self): 
        self.a = 0 # instance변수 (반드시 객체 생성필요)
        self.b = 0 # instance변수 (반드시 객체 생성필요)

    def setData(self, x,y): 
        self.a = x
        self.b = y

    def show(self): # instance 멤버함수
        # Test.st = 10 Ok 
        print(self.a, self.b)

    @staticmethod #aa가 셀프가 되는 것을 방지
    def stFn(aa): #self가 없는 함수(static 멤버함수), 인스턴트 멤버 사용불가
        # self.a = 0 X
        Test.st = aa

    @staticmethod
    def getSt():
        return Test.st


print( Test.st)
obj = Test() # 객체 생성: 메모리 할당이 됨
# obj.__init__(odj)
obj.setData(10,20) #obj.setData(self, 10,20)
obj.show()
# obj.stFn() #obj.stFn(obj) #obj인자를 받을 수 없다 -> self가 없기때문
Test.stFn(10)
print(Test.st)
print(Test.getSt())
```

### 정리: 
- 클래스 내부에 선어되있고 멤버함수 외부에 선언된 변수를 static변수
- self거 없는 멤버함수는 staic 멤버함수
- 클래스명.멤버함수()로 호출한다

- 인스턴스멤버 데이터는 객체 생성시 메모리가 별도로 할당되기 때문에 일반적으로 데이터 용도
- static멤버는 연산 및 처리용도로 사용된다.

```python
class Student:
    def __init__(self, name, age, addr):
        self.name = name
        self.age = age
        self.addr = addr
    
    def show(self):
        print('이름', self.name)
        print('나이', self.age)
        print('주소', self.addr)
```

```python
std1 = Student('홍길동', 20, '서울시')
# std1.__init__(std1,'홍길동', 20, '서울시')
std1.show()
std2 = Student('임꺽정', 30, '부산시')
# std2.__init__(std2,'임꺽정', 30, '부산시')
std2.show()
```

```python
mylist = []
mylist.append( Student('홍길동',20,'서울시'))
mylist.append( Student('임꺽정',30,'부산시'))

for n in mylist:
    n.show() #n.show(n)
    print('-----------------')
```

```python

def circle_area(self, r):
    return 3.14*r**2

def cylinder(self, r,h):
    return 3.14*r**2*h
```

```python
class Circle:
    def circle_area(self, r):
        return 3.14*r**2
    
    def cylinder(self, r,h):
        return 3.14*r**2*h
```

```python
c = Circle()
print(c.circle_area(5))
print(c.cylinder(3,10))
```

```python
class Circle:
    @staticmethod
    def circle_area( r):
        return 3.14*r**2
    
    @staticmethod
    def cylinder( r,h):
        return 3.14*r**2*h
```

```python
print(Circle.circle_area(5))
print(Circle.cylinder(3,10))
```
