```python
pizza_slice = 9
person = 0
if person == 0:
    print('사람 수는 0이면 안됩니다.')
elif person<0:
    print('사람 수는 양수이여야 합니다.')
else:
    per_pizza_slice = pizza_slice/person
    print(per_pizza_slice)
```

## if 문이 많아지면 유지보수가 어렵다.

```python
pizza_slice = 9
person = -3

try:
    if person <0:
        raise Exception('person이 음수면 안됩니다.')
    per_pizza_slice = pizza_slice/person
    print(per_pizza_slice)
except Exception as err:    #모든 예외를 모두 처리(상위클래스)
    print('에러발생',err)
```

```python
n = input('숫자입력')
try:
    rst = int(n) + 10
    print(rst)
except ValueError as err:
    print('에러발생(반드시 숫자만 입력)', err)
```
