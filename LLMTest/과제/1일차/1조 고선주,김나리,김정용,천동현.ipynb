````python
import os
import openai
import pygame
from openai import OpenAI
from IPython.display import Audio

apiKey = os.getenv("OPENAI_API_KEY")
openai.api_key = apiKey
````

````python
client = OpenAI()
conversation = [{"role": "system", "content": "You are a helpful teacher"}]


def gpt_conv(user_input):
    conversation.append({"role": "user", "content": user_input})
    completion = client.chat.completions.create(model="gpt-4o", messages=conversation)
    assistant_response = completion.choices[0].message.content
    conversation.append({"role": "assistant", "content": assistant_response})
    return assistant_response


def text_to_audio(text, filename):
    audio_response = client.audio.speech.create(
        model="tts-1",
        input=text,
        voice="nova",
    )
    with open(filename, "wb") as fp:
        fp.write(audio_response.content)
    pygame.mixer.init()
    pygame.mixer.music.load(filename)
    pygame.mixer.music.play()


text = gpt_conv("인공지능에 대해 알려줄래")
text_to_audio(text, "about_ai.mp3")

while True:
    question = input("계속하시겠습니까(y,n)?: ")
    if question.lower() == "y":
        text = gpt_conv("좀 더 간단히 알려줄래")
        text_to_audio(text, "about_ai_summary.mp3")
    elif question.lower() == "n":
        pygame.mixer.quit()
        print("종료")
        break
    else:
        print("다시 입력")
        continue
````
