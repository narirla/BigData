```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import XMLOutputParser # 출력 양식 파서
```

### XMLOutputParser

```python
xml_out = XMLOutputParser()
format_instructions = xml_out.get_format_instructions()
print( format_instructions)
```

```python
template = "10개의 팀을 보여줘 {subject}.\n{format_instructions}"
prompt = PromptTemplate(
    input_variables=["subject"],
    template=template,
    partial_variables={"format_instructions": format_instructions}
)
print( prompt.format( subject="한국의 야구팀은" ) )
```

```python
gpt = ChatOpenAI( model_name="gpt-3.5-turbo", temperature=0)
rst = gpt.invoke( prompt.format(subject="한국의 야구팀은") )
print( rst.content ) # 한국 야구 10팀을 XML 형식으로 출력
```

```python
xml_out.parse(rst.content)  # XML 파싱, 딕셔너리 타입으로 변경
```

### JsonOutputParser

```python
from langchain_core.output_parsers import JsonOutputParser
```

```python
parser = JsonOutputParser()
format_instructions = parser.get_format_instructions()
print(format_instructions)
```

```python
template = "당신의 이름은 {name}이고, 나이는 {age}을 다음형식으로 출력:\n\n{format_instructions}"
prompt = PromptTemplate(
    input_variables=["name", "age"],
    template=template,
    partial_variables={"format_instructions": format_instructions}
)
print(prompt.format(name="홍길동", age=30))
```

```python
gpt = ChatOpenAI( model_name="gpt-3.5-turbo", temperature=0)
rst = gpt.invoke( prompt.format(name="홍길동", age=30) )
print( rst.content )
```

```python
parser.parse(rst.content)  # JSON 파싱
```

```python
from pydantic import BaseModel, Field
```

```python
class Country(BaseModel):
    continet: str = Field(description="사용자가 물어본 나라가 속한 대륙")
    population: int = Field(description="사용자가 물어본 나라의 인구")  

parser = JsonOutputParser(pydantic_object=Country)
format_instructions = parser.get_format_instructions()
print(format_instructions)
```

```python
template = "answer the question.\n{question}\n\n{format_instructions}"
prompt = PromptTemplate(   template=template,
    input_variables=["question"],   
    partial_variables={"format_instructions": format_instructions}
)

print( prompt.format( question="아르헨티나는 어떤 나라야?" ) )
```

```python
gpt = ChatOpenAI( model_name="gpt-3.5-turbo", temperature=0)
rst = gpt.invoke( prompt.format( question="아르헨티나는 어떤 나라야?" ) )
print( rst.content )
```
