```python
from langchain_openai import ChatOpenAI
from langchain_core.messages import SystemMessage, HumanMessage, AIMessage
```

```python
gpt = ChatOpenAI(model='gpt-3.5-turbo', temperature=0)
rst = gpt.invoke('인공지능에 대해 간단히 설명해줘')
print(type(rst))
print(rst)
print(rst.content)
```

```python
messages = [ SystemMessage(content='You are a helpful assistant.'), # 역할 부여
            HumanMessage(content='인공지능에 대해 간단히 알려줄래')]
gpt = ChatOpenAI(model='gpt-3.5-turbo', temperature=0)
rst = gpt.invoke(messages)
print(rst.content)
```

## 대화식 채팅

```python
messages = [ SystemMessage(content='You are a helpful assistant.')]
gpt = ChatOpenAI(model='gpt-3.5-turbo', temperature=0)
messages.append(HumanMessage(content='인공지능에 대해 간단히 알려줄래'))

rst = gpt.invoke(messages)
print(rst.content)
print('-'*20)
messages.append(AIMessage(content=rst.content))

messages.append(HumanMessage(content='더 간단히 알려줄래'))
rst = gpt.invoke(messages)
print(rst.content)
```

## stream

```python
gpt = ChatOpenAI(model='gpt-3.5-turbo', temperature=0)
rst = gpt.stream('인공지능에 대해 5개로 요약해서 알려줄래')
# print(rst)
for chunk in rst:
    print(chunk.content, end='')
```

## generate (질문 여러 개 가능)

```python
messages_batch = [
    [HumanMessage(content='인공지능에 대해 간단히 알려줄래')],
    [HumanMessage(content='파이썬에 대해 알려줄래')]
]

gpt = ChatOpenAI(model='gpt-3.5-turbo', temperature=0)
rst = gpt.generate(messages_batch, n=1)
print(rst.generations[0][0].text)
# print(rst.generations[0][1].text)
print('=====')
print(rst.generations[1][0].text)
# print(rst.generations[1][1].text) n=2일 경우
```

```python
messages_batch = [
    [HumanMessage(content='빈센트 반 고흐에 대해 간단히 알려줄래')]
]

gpt = ChatOpenAI(model='gpt-3.5-turbo', temperature=0.2)
rst = gpt.generate(messages_batch, n=2)
print(rst.generations[0][0].text)
print('=====')
print(rst.generations[0][1].text)

# print(rst.generations[1][0].text)
# print(rst.generations[1][1].text) #n=2일 경우
```
