```python
import streamlit as st
from PIL import Image
from openai import OpenAI
import base64

client = OpenAI()

def encode_image(image_path):
    with open(image_path, "rb") as f:
        return base64.b64encode(f.read()).decode("utf-8")

def image_exp( image_file ):
    image_base64 = encode_image( image_file )  # 여기에 본인의 로컬 이미지 경로
    data_url = f"data:image/jpeg;base64,{image_base64}"
    response = client.chat.completions.create(
        model="gpt-4o",
        messages=[{
            "role": "user",
            "content": [
                {"type": "text", "text": "이 이미지에 대해 설명해줘"},
                {"type": "image_url", "image_url": {"url": data_url}},
            ],
        }]
    )
    return response.choices[0].message.content

st.title('이미지 업로드')
upfile = st.file_uploader('이미지 업로드',type=['jpg','png','jpeg'] )

if upfile is not None:
    image = Image.open( upfile)
    st.image(image)

    strPath = f'uploadfile\{upfile.name}'
    image.save(strPath)
    with st.spinner():
        exp = image_exp( strPath)
        st.write( exp )
    
```

```python
import streamlit as st

st.title('텍스트 업로드')
upfile = st.file_uploader('텍스트 업로드', type=['txt'] )

#서버로 파일전송완료
if upfile is not None:
    content = upfile.read().decode('utf-8')
    st.write( content )
    txtPath = f'uploadfile\{upfile.name}'
    with open(txtPath,'w') as fp:
        fp.write( content)
```

```python
import streamlit as st

if 'txt' not in st.session_state:
    st.session_state['txt'] = ""
prompt = st.text_input( '프롬프트', placeholder="입력후 enter")
if prompt:
    # txt = txt + prompt # txt +=prompt
    st.session_state['txt'] +=prompt
    st.write( st.session_state['txt'] )
```

```python
import streamlit as st
from openai import OpenAI
import openai
import os

if "client" not in st.session_state:
    api_key = os.getenv('OPENAI_API_KEY')
    openai.api_key = api_key
    st.session_state['client']  = OpenAI()

def GPT(prompt):
    completion = st.session_state['client'].chat.completions.create(
        model='gpt-3.5-turbo', 
        messages=[{'role':'user','content':prompt}], stream=True )
    
    for c in completion:
        if c.choices[0].delta.content:
            yield c.choices[0].delta.content

st.title( '🤖 GPT 챗봇')
st.markdown('---')

# 안에 영역 생성
with st.form('myform'): 
    prompt = st.text_input('👩‍🦰프롬프트', placeholder="입력")
    submit = st.form_submit_button('확인')
    if submit:
        myemp = st.empty()
        full_text =""
        for  chunk in  GPT(prompt):
            full_text += chunk
            myemp.write(full_text)
```
