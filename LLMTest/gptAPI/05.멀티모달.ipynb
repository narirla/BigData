```python
import os
import openai 
from openai import OpenAI
from IPython.display import Audio, display
apiKey = os.getenv('OPENAI_API_KEY')
openai.api_key = apiKey
```

### whisper : 음성 -> 텍스트

```python
client = OpenAI()

fp = open('audio/베트남여행.mp3', 'rb')

transcription = client.audio.transcriptions.create( file=fp, model='whisper-1')
# print(transcription)
print(transcription.text)

completion = client.chat.completions.create(model='gpt-3.5-turbo',
                                            messages=[{'role':'user','content':f'Please summarize the following:{transcription.text}'}])
print('-'*50)
print(completion.choices[0].message.content)
```

```python
client = OpenAI()

def transcribe_audio(file_path):
    fp = open(file_path, 'rb')
    transcription = client.audio.transcriptions.create( file=fp, model='whisper-1')
    return transcription.text
def transcribe_and_summarize(trans_text):
    completion = client.chat.completions.create(model='gpt-3.5-turbo',
                                                messages=[{'role':'user','content':f'Please summarize the following:{trans_text}'}])

    return completion.choices[0].message.content

text = transcribe_audio('audio/베트남여행.mp3')
summarize_text = transcribe_and_summarize(text)
print('요약 내용 : ', summarize_text)
```

### tts-1 : 텍스트 -> 음성으로 변환

```python
client = OpenAI()
response = client.audio.speech.create(input='안녕하세요. 오늘 날씨가 참 덥네요',
                                    model='tts-1', voice="nova")

with open('out.mp3', 'wb') as fp:
    fp.write(response.content)
```

```python
audio = Audio(filename='out.mp3', autoplay=True)
display(audio)
```

### dalle-e : 이미지 생성

```python
import requests
client = OpenAI()
response = client.images.generate(prompt="a white siamese cat", model='dall-e-3',
                                size='1024x1024', quality="standard", n=1)
image_url = response.data[0].url # 생성된 이미지의 URL 추출
image_data = requests.get(image_url).content # URL로부터 이미지 바이너리 데이터 다운로드

# 다운로드한 이미지를 파일로 저장
with open('genimage.jpg','wb') as fp:
    fp.write(image_data)
```
