````python
import os
import openai 
from openai import OpenAI
apiKey = os.getenv('OPENAI_API_KEY')
openai.api_key = apiKey
````

````python
client = OpenAI()

conv_history = [{'role':'system', 'content':'You are a helpful teacher'}]

conv_history.append( {'role':'user', 'content':'인공지능에 대해 간단히 알려줄래'} )

completion = client.chat.completions.create(model='gpt-3.5-turbo',
                                            messages=conv_history)
chatReply = completion.choices[0].message.content
print(chatReply)
conv_history.append({'role':'assistant', 'content':chatReply})
conv_history.append({'role':'user','content':'더 간단하게 요약해줘'})

completion = client.chat.completions.create(model='gpt-3.5-turbo',
                                            messages=conv_history)
print('='*60)
print(completion.choices[0].message.content)
````

````python
client = OpenAI()

conv_history=[{'role':'system','content':'You are a helpful teacher'}]

def conv_with_gpt(user_input):
    conv_history.append({'role':'assistant', 'content':user_input})
    completion = client.chat.completions.create(model='gpt-3.5-turbo',
                                                messages=conv_history)
    
    chatReply = completion.choices[0].message.content
    conv_history.append({'role':'assistant','content':chatReply})
    return chatReply

response = conv_with_gpt('인공지능에 대해 알려줄래')
print(response)
print('-'*20)
response = conv_with_gpt('더 간단히 요약해줘')
print(response)
````
