```python
import os
import openai 
from openai import OpenAI
apiKey = os.getenv('OPENAI_API_KEY')
openai.api_key = apiKey
```

- system role : 역할 부여
- user role : 사용자 질문( 프롬프트 )
- assistant role : gpt 답변( 컴플리션 )

```python
client = OpenAI()

completion = client.chat.completions.create(model="gpt-3.5-turbo",
    messages=[
        {'role':'system', 'content':'You are a helpful teacher'}, # 역할 부여
        {'role':'user',
            'content':'Are ther other measure than time complexity for an algorithm?'},
        {'role':'assistant', # gpt 답변 
            'content':'Yes, there are other measures besides time complexity\
            for an algorithm, such as space complexity.' }, 
        {'role':'user','content':'What is it?'}] )

print( completion.choices[0].message.content)
```
