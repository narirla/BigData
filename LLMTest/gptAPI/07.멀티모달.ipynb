````python
from openai import OpenAI
import os
api_key = os.getenv("OPENAI_API_KEY")
OpenAI.api_key = api_key
client = OpenAI()
### whisper 모델 : 음성을 텍스트로 변환해줌
fp = open("./data/audio/베트남여행.mp3", "rb") # r(read) b: 사운드, 영상, 이미지 등의 바이너리 형태
transcription = client.audio.transcriptions.create(file=fp, model="whisper-1")
completion = client.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=[
        {"role": "user", "content": f"{transcription.text} 를 요약해줘"}
    ]
)
print(completion.choices[0].message.content)
````

````python
client = OpenAI()

def transcribe_audio( file_path):
    fp = open( file_path, 'rb')
    trascription = client.audio.transcriptions.create(file=fp, 
                                                      model='whisper-1')
    fp.close()
    return trascription.text

def transcribe_and_summarize( trascription_text ):
    response = client.chat.completions.create(model='gpt-3.5-turbo',
                messages=[
                    {'role':'system',
                        'content':'You are helpful assistant that summarizes text'},
                    {'role':'user',
                    'content':f'다음 내용을 간단히 요약해줘:\n{trascription_text}'}
                ])
    return response.choices[0].message.content


trascription_text = transcribe_audio( 'audio/베트남여행.mp3')
print('오디오로부터 추출된 텍스트:', trascription_text)

summarize_text = transcribe_and_summarize( trascription_text )
print( '내용:', summarize_text)
````

### tts-1 텍스트 --> 음성

````python
client = OpenAI()

response = client.audio.speech.create(input='안녕하세요 오늘 날씨가 참 덥네요',
                                      model="tts-1" , voice='nova' )

with open( 'out.mp3', 'wb') as fp:
    fp.write( response.content )
````

### 5분 퀴즈: 
질문:키보드입력
gpt 의 답변을 음성으로 들려주세요: 

voice_gpt.py 파일에 코드 작성

````python
from openai import OpenAI
import openai
import os
from playsound import playsound

api_key = os.getenv('OPENAI_API_KEY')
openai.api_key = api_key
client = OpenAI()

prompt = input('prompt: ')

response = client.chat.completions.create(
    model='gpt-3.5-turbo',
    messages=[
        {'role': 'user', 'content': prompt}
    ]
)
resp = response.choices[0].message.content
print( 'gpt',resp )

response_to_voice = client.audio.speech.create(input=resp, model='tts-1',voice='nova')
with open('out.mp3', 'wb') as fp: #'wb(write binary) 모드로 파일을 열어야함'
    fp.write(response_to_voice.content)

playsound('out.mp3')
````
