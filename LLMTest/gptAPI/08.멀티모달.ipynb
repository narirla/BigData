### 음성녹음 wav 파일로 저장

```python
import sounddevice as sd
from scipy.io.wavfile import write

try:
    fs = 16000 # 샘플링속도
    duration = 5 # 녹음시간 5
    print( '녹음시작')
    audio_data = sd.rec( int(duration*fs) , samplerate=fs, channels=1 )
    sd.wait()
    print( '녹음완료')
    write( 'rec.wav', fs, audio_data)
except Exception as err:
    print( '에러발생', err)
```
