### Stream 인자

```python
client = OpenAI()
completion = client.chat.completions.create(model='gpt-3.5-turbo',
                                            messages=[{'role':'user','content':'인공지능에 대해 5개로 요약해줘. 각 요약에는 번호를 붙여줘'}],
                                            stream=True) # stream=True 실시간 답변 출력
for c in completion:
    # print(c) -> 한 글자씩 끊어서 출력
    if c.choices[0].delta.content: # if를 줘서 마지막 문장에 붙는 None 미출력
        print(c.choices[0].delta.content, end='')
```

### n : 답변의 수를 지정

```python
client = OpenAI()
completion = client.chat.completions.create(model='gpt-3.5-turbo',
                                            messages=[{'role':'user','content':'빈센트 반 고흐에 대해 알려줄래'}],
                                            n=2) # n=2 -> 답변 두개 출력
# completion
for choice in completion.choices:
    print(choice.message.content)
    print('-'*30)
```

### max_tokens: 최대 응답 토큰의 수(특정 글자 이내로 알려줘)

```python
client = OpenAI()
completion = client.chat.completions.create(model='gpt-3.5-turbo',
                                            messages=[{'role':'user','content':'빈센트 반 고흐에 대해 알려줄래'}],
                                            n = 2, max_tokens=200) # n=2 -> 답변 두개 출력, max_tokens=200 -> 200 글자 이상 넘어가면 출력 멈춤
# completion
for choice in completion.choices:
    print(choice.message.content)
    print('-'*30)
```

### temperature
- 허용 값 : 0 ~ 2
- 값이 클 수록 창의성(자유도)이(/가) 높아진다.
- 0 보수적 응답(확실한 내용을 중심으로 답변) -> 사실에 근거한 답변을 받으려면 temperature=0 을 주면 됨

```python
client = OpenAI()
completion = client.chat.completions.create(model='gpt-3.5-turbo',
                                            messages=[{'role':'user','content':'빈센트 반 고흐에 대해 알려줄래'}],
                                            temperature=0) # temperature = 0 -> 창작 금지, fact 만 답변
# completion
for choice in completion.choices:
    print(choice.message.content)
    print('-'*30)
```
