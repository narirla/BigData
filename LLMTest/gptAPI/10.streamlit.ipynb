**Streamlit(스트림릿)**은 데이터 사이언티스트와 머신러닝 엔지니어가 **Python 코드만으로 빠르고 아름다운 웹 애플리케이션을 만들 수 있게 해주는 오픈소스 프레임워크**입니다.

HTML, CSS, JavaScript 같은 프론트엔드 지식이 없어도 데이터 분석 결과를 시각화하거나 모델 데모 페이지를 구축하는 데 최적화되어 있습니다.

---

## 1. Streamlit의 주요 특징

* **파이썬 전용:** 백엔드와 프론트엔드를 구분할 필요 없이 오직 Python으로만 작성합니다.
* **빠른 개발 속도:** 위젯(버튼, 슬라이더 등)을 변수처럼 선언하여 즉시 UI를 구성할 수 있습니다.
* **실시간 반영 (Hot Reloading):** 코드를 수정하고 저장하면 브라우저에 즉시 업데이트됩니다.
* **강력한 시각화 지원:** Matplotlib, Plotly, Altair, Pandas 등 주요 데이터 라이브러리와 완벽하게 호환됩니다.
* **배포 용이성:** 'Streamlit Community Cloud'를 통해 GitHub 저장소와 연결하면 클릭 몇 번으로 무료 배포가 가능합니다.

---

## 2. 기본 구성 요소 (핵심 함수)

스트림릿에서 자주 사용하는 기능들은 다음과 같습니다.

| 구분 | 함수 예시 | 용도 |
| --- | --- | --- |
| **텍스트** | `st.title()`, `st.header()`, `st.write()` | 제목, 본문, 마크다운 출력 |
| **데이터** | `st.dataframe()`, `st.table()` | 인터랙티브 표 또는 정적 표 출력 |
| **차트** | `st.line_chart()`, `st.pyplot()` | 선 그래프 및 다양한 시각화 도구 출력 |
| **입력 위젯** | `st.button()`, `st.slider()`, `st.text_input()` | 사용자로부터 입력을 받는 도구 |
| **레이아웃** | `st.sidebar`, `st.columns()` | 사이드바 생성 및 화면 분할 |

---


## 3. 장점과 단점

### 장점

* 학습 곡선이 매우 낮음 (파이썬만 알면 바로 시작 가능)
* 프로토타입 제작에 압도적으로 유리함
* 디자인이 기본적으로 깔끔하고 현대적임

### 단점

* 복잡하고 고도로 커스터마이징된 UI 구현은 어려움
* 페이지가 상호작용할 때마다 코드 전체가 처음부터 다시 실행됨 (캐싱(`@st.cache_data`)으로 보완 가능)


### streamlit
- webserver

- 풍부한 plugin
- 반드시 py 파일로 작성

````python
import streamlit as st

with st.form('myform'):
    name = st.text_input('이름')
    age = st.number_input('나이',value=0, step=1, min_value=0, max_value=100)
    birth = st.date_input('생일')
    submit = st.form_submit_button('확인')
    if submit:
        s = f'이름:{name} 나이:{age} 생일:{birth}'
        st.write( s )

st.write('테스트')
````

````python
import streamlit as st
import os
import openai 
from openai import OpenAI
apiKey = os.getenv('OPENAI_API_KEY')
openai.api_key = apiKey

client = OpenAI()

with st.form('myform'):
    prompt = st.text_input('프롬프트: ')
    submit = st.form_submit_button('요청')
    if submit:
        s = f'프롬프트: {prompt}'
        with st.spinner('잠시 기다려 주세요...'):
            completion = client.chat.completions.create(model='gpt-4o',
                                                    messages=[{'role':'user','content':prompt}])
        st.write( completion.choices[0].message.content )
````

````python
import streamlit as st
from openai import OpenAI
import openai
import os

# client  = OpenAI()
if "client" not in st.session_state:
    api_key = os.getenv('OPENAI_API_KEY')
    openai.api_key = api_key
    st.session_state['client']  = OpenAI()

def GPT(prompt):
    response = st.session_state['client'].chat.completions.create(
        model='gpt-3.5-turbo', 
        messages=[{'role':'user','content':prompt}]
    )
    return response.choices[0].message.content

st.title( '🤖 GPT 챗봇')
st.markdown('---')

# 안에 영역 생성
with st.form('myform'): 
    prompt = st.text_input('👩‍🦰프롬프트', placeholder="입력")
    submit = st.form_submit_button('확인')
    if submit:
        with st.spinner():
            rst = GPT(prompt)
            s = f'🤖 GPT 답변 : {rst}'
            st.write(s)
````
