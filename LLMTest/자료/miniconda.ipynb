# NVIDIA 그래픽 카드 드라이버 설치


## 1. NVIDIA 드라이버 다운로드 링크

가장 권장되는 방법은 NVIDIA 공식 홈페이지에서 본인의 그래픽 카드 모델에 맞는 드라이버를 직접 선택하는 것입니다.

* **NVIDIA 공식 드라이버 다운로드:** [https://www.nvidia.com/download/index.aspx](https://www.nvidia.com/download/index.aspx)
* **Product Type:** GeForce (일반 사용자용)
* **Product Series:** 본인의 그래픽 카드 시리즈 (예: RTX 40 Series)
* **Operating System:** Windows 10/11 또는 Linux 64-bit 선택


## 2. 설치 후 확인 명령어

드라이버가 정상적으로 설치되었는지 확인하는 방법입니다.

### **Windows (명령 프롬프트 또는 PowerShell)**

Windows에서는 명령어를 통해 드라이버 버전과 GPU 상태를 즉시 확인할 수 있습니다.

1. `Win + R` 키를 누르고 **cmd**를 입력하여 실행합니다.
2. 다음 명령어를 입력합니다.
```bash
nvidia-smi

```
* 이 명령어를 입력했을 때 GPU 이름, 드라이버 버전, CUDA 버전 등이 포함된 표가 나타나면 성공입니다.



### **Linux (Ubuntu 등)**

터미널을 열고 아래 명령어 중 하나를 사용하세요.

* **상세 상태 확인:**
```bash
nvidia-smi

```


* **드라이버 버전만 간단히 확인:**
```bash
modinfo nvidia | grep ^version

```

# Miniconda

**Miniconda**는 파이썬(Python) 개발 환경을 효율적으로 관리하기 위한 **최소한의 설치 패키지**입니다.

복잡한 설정을 피하고, 프로젝트마다 필요한 도구만 골라 담고 싶을 때 사용합니다. 핵심 내용은 다음과 같습니다.



### 1. Miniconda의 핵심 구성

Miniconda는 딱 3가지만 포함하고 있어 매우 가볍습니다.

* **Conda:** 패키지 및 가상환경 관리 도구 (핵심 엔진)
* **Python:** 기초적인 파이썬 언어
* **최소한의 기본 패키지:** 실행에 꼭 필요한 몇 가지 도구

### 2. 왜 사용하는가? (주요 장점)

* **가상환경 분리:** 프로젝트 A는 파이썬 3.8, 프로젝트 B는 3.11이 필요할 때, 서로 충돌하지 않게 방마다 따로 설치하는 물리적 공간을 만들어 줍니다.
* **패키지 관리:** `pip`보다 강력하게 라이브러리 간의 의존성(버전 꼬임)을 해결해 줍니다. 특히 데이터 분석이나 AI 관련 라이브러리 설치에 특화되어 있습니다.
* **가벼움:** 'Anaconda(아나콘다)'라는 전체 패키지는 용량이 매우 크지만(수 GB), Miniconda는 수백 MB 정도로 가볍게 시작할 수 있습니다.

### 3. 아나콘다(Anaconda) vs 미니콘다(Miniconda)

| 구분 | Anaconda | Miniconda |
| --- | --- | --- |
| **용량** | 매우 큼 (약 3GB+) | 작음 (약 400MB) |
| **포함 내용** | 1,500개 이상의 데이터 과학 패키지 포함 | Conda, Python, 기본 도구만 포함 |
| **추천 대상** | 입문자, 모든 도구가 한 번에 필요한 분 | **필요한 것만 골라 설치하고 싶은 개발자** |


### 4. 자주 쓰는 명령어 예시

설치 후 터미널(또는 CMD)에서 아래 명령어만 알아도 바로 시작할 수 있습니다.

* **가상환경 만들기:** `conda create -n 내이름 python=3.10`
* **가상환경 들어가기:** `conda activate 내이름`
* **패키지 설치하기:** `conda install numpy`
