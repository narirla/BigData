```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate
```

```python
product = '컴퓨터'
s = f"{product} 제품 홍보 문구 작성해줘"
print(s)
```

```python
# 특정 부분만 바뀌고 나머진 고정, 바뀌는 부분만 변수로 지정
template = "{product} 제품 홍보 문구 작성해줘"
prompt = PromptTemplate(input_variables=['product'], template=template)
txt = prompt.format(product='컴퓨터')
txt
```

```python
# 프롬프트만 지정하면 자동으로 변수 인식
template = "{product} 제품 홍보 문구 작성해줘"
prompt = PromptTemplate.from_template(template=template)
prompt.format(product='컴퓨터')
```

```python
template = "{product} 제품 홍보 문구 {desc}"
prompt = PromptTemplate(input_variables=['product', 'desc'], template=template)
txt = prompt.format(product='컴퓨터', desc='만들어줘')
txt
```

```python
# api와 연결
template = "{product} 제품 홍보 문구 작성해줘"
prompt = PromptTemplate(input_variables=['product'], template=template)
gpt = ChatOpenAI(model='gpt-3.5-turbo')
rst = gpt.invoke(prompt.format(product='컴퓨터'))
print(rst.content)
```

```python
template = "오늘은 {day}입니다. {name}님, 좋은 하루 되세요!"
prompt = PromptTemplate(input_variables=['name'],
                        partial_variables={'day':'화요일'}, # 특정 고정 값을 줄 수 있음
                        template=template)
prompt.format(name='홍길동')
```
