# 🔧 Hugging Face `pipeline()` 함수 주요 인자 정리표

`pipeline()` 함수는 Hugging Face Transformers에서 다양한 태스크를 쉽게 실행할 수 있도록  
제공되는 고수준(high-level) API.

---

## ✅ `pipeline()` 함수 인자 설명표

| 인자명              | 자료형           | 기본값      | 설명                                                                                   |
|---------------------|------------------|-------------|----------------------------------------------------------------------------------------|
| `task`              | `str`            | **필수**     | 수행할 태스크 명칭 (예: `"text-classification"`, `"text-generation"`, `"ner"` 등)     |
| `model`             | `str` 또는 모델 객체 | 자동 선택   | 사용할 모델 이름 (Hugging Face Hub의 경로) 또는 사전 로드된 모델 객체                   |
| `tokenizer`         | `str` 또는 객체     | 모델 자동 매칭 | 사용할 토크나이저 (`AutoTokenizer`)                                                      |
| `framework`         | `'pt'`, `'tf'`    | 자동 감지    | 사용할 딥러닝 프레임워크 (`pt` = PyTorch, `tf` = TensorFlow)                            |
| `device`            | `int`            | `-1` (CPU)  | 실행할 장치: `-1` = CPU, `0` = 첫 번째 GPU (CUDA:0), `1` = 두 번째 GPU 등                |
| `revision`          | `str`            | `"main"`    | 모델의 특정 버전(브랜치, 태그 또는 커밋)을 지정할 수 있음                              |
| `use_fast`          | `bool`           | `True`      | fast tokenizer 사용 여부 (`True`일 경우 rust 기반 빠른 토크나이저 사용)                  |
| `config`            | `str` 또는 객체    | 모델 자동 로딩 | 모델 구성 설정 (`AutoConfig`)                                                            |
| `feature_extractor` | 객체 또는 이름     | 자동 감지    | 이미지/음성 태스크에서 입력 전처리에 사용                                                |
| `device_map`        | `str`, `dict`    | 없음         | 여러 GPU에 자동 분산 로딩 (`"auto"` 권장)                                               |
| `trust_remote_code` | `bool`           | `False`      | 사용자 정의 모델 로딩 허용 여부 (보안 주의 필요)                                         |
| `model_kwargs`      | `dict`           | 없음         | 모델 초기화 시 추가 인자 전달용 (`from_pretrained()`의 kwargs)                         |

---

## 🧪 예시

```python
from transformers import pipeline

# 기본 감성 분석 (CPU 실행)
clf = pipeline("sentiment-analysis")
print(clf("I love Hugging Face!"))

# GPU에서 텍스트 생성 (LLaMA 사용)
gen = pipeline("text-generation", model="meta-llama/Llama-3-8B-Instruct", device=0)
print(gen("Once upon a time", max_new_tokens=30))
