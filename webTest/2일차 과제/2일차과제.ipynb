```python
from bs4 import BeautifulSoup
import requests
import warnings
warnings.simplefilter('ignore')
```

```python
url = 'http://www.moviechart.co.kr/rank/realtime/index/image'
respone = requests.get(url)
html = BeautifulSoup(respone.content, 'html.parser')
html
```

```python
for movie in html.select('.movieBox-item'):
    print(movie.select('.rank.realtime_rank23')[0].text.strip())
    print(movie.select('.movie-title > h3 > a')[0].text.strip())
    print(movie.select('li.ticketing')[0].text)
    print(movie.select('li.movie-launch')[0].text)
    print('=========================================')

    
```

```python
moviechart = []
for movie in html.select('.movieBox-item'):
    
    ranked = movie.select_one('.rank.realtime_rank23').text.strip()
    movietitle = movie.select_one('.movie-title > h3 > a').text.strip()
    raw_ticketing = movie.select_one('li.ticketing').text.split(':')[1].strip()
    ticketing = float(raw_ticketing.replace('%',''))
    movielaunch = movie.select_one('li.movie-launch').text.split(':')[1].strip()

    moviechart.append((ranked,movietitle,ticketing,movielaunch))
    print(ranked,movietitle,ticketing,movielaunch)
```

```python
import pymysql
try:
    sql = 'insert into moviechart (ranked,movietitle,ticketing,movielaunch) values(%s,%s,%s,%s)'

    conn = pymysql.connect(host='localhost', 
                        user='root', password='1234', database='test',
    charset='utf8mb4' )
    cur = conn.cursor()
    cur.executemany(sql, moviechart)
    conn.commit()
    conn.close()
except Exception as err:
    print('에러', err)

```
