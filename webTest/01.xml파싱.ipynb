````python
kpop.find_all('song')
````

````python
for song in kpop.find_all('song'):
    print(song)
    print('============================')
````

````python
for song in kpop.find_all('song'):
    print(song['album']) #key 값 형식으로 가져옴
    print(song.singer.string)
    print(song.title.string)
    print('============================')
````

### 퀴즈: food.xml의 음식명, 가격을 파싱에서 출력하시오.

````python
fp = open('data/food.xml')
bmenu = BeautifulSoup(fp,"html.parser")
bmenu
````

````python
bmenu
````

````python
bmenu.find_all('food')
````

````python
for food in bmenu.find_all('food'):
    print(food.find('name').string) # .name을 사용하면 속성과 출력함으로 find()로 묶어줄것
    print(food.price.string)
    print('===============================')
````
