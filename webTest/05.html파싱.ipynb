```python
from bs4 import BeautifulSoup
import requests
import warnings
warnings.simplefilter('ignore')
```

```python
url = 'http://www.menupan.com/restaurant/bestrest/bestrest.asp?pt=rt&areacode=bs205'
reponse = requests.get(url)
html = BeautifulSoup(reponse.content, 'html.parser')
html
```

```python
for a in html.select('.restName > a'):
    print(a.string)
```

```python
for a in html.select('.listType'):
    print(a.string)
```

```python
for a in html.select('.listArea'):
    print(a.string)
```

```python
html.select('.rankingList li')
```

```python
for li in html.select('.rankingList li'):
    print('순위: ',li.select_one('.numTop, .rankNum').string)
    print(li.select_one('.restName > a').string)
    print(li.select_one('.listType').string)
    print(li.select_one('.listArea').string)
    print('===================================')
```

```python
href =[]
for a in html.select('.restName > a'):
    print(a['href'])
    href.append(a['href'])
```

```python
href
```

```python
for h in href:
    baseurl = 'http://www.menupan.com'
    reponse = requests.get(baseurl+h)
    html = BeautifulSoup(reponse.content, 'html.parser')

    print(html.select_one('.name').text.split('[')[0])
    print(html.select_one('.tel1').string)
    print(html.select_one('.add1 > a').string)
    print(html.select_one('.total').string)
    print('=================================================')
```

```python
mydata = []
for h in href:
    baseurl = 'http://www.menupan.com'
    reponse = requests.get(baseurl+h)
    html = BeautifulSoup(reponse.content, 'html.parser')

    foodname = html.select_one('.name').text.split('[')[0].strip()
    tel = html.select_one('.tel1').string
    addr = html.select_one('.add1 > a').string
    rank = float(html.select_one('.total').string)
    mydata.append((foodname, tel, addr, rank))
    print(foodname)
```

```python
mydata
```

```python
import pymysql
try:
    sql ='insert into busanfood values(%s,%s,%s,%s)'

    conn = pymysql.connect(host='localhost', 
                        user='root', password='1234', database='test')
    cur = conn.cursor()
    cur.executemany(sql, mydata)
    conn.commit()
    conn.close()
except Exception as err:
    print('에러', err)

```

```python
import pymysql
try:
    sql ='select * from busanfood order by frank desc'

    conn = pymysql.connect(host='localhost', 
                        user='root', password='1234', database='test')
    cur = conn.cursor()
    cur.execute(sql)
    result = cur.fetchall()
    conn.close()
    for a,b,c,d in result:
        print(f'상호명: {a}, 전화번호: {b}, 주소: {c}, 평점: {d}')
except Exception as err:
    print('에러', err)

```
