````python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
#  비정상 시계열 데이터 생성
np.random.seed(42)
x = np.arange(1, 21) + np.random.normal(0, 1, 20)
print('x=', x)
#  1차 차분 (Differencing)
diff_x = np.diff(x)
print('diff_x=', diff_x)

#  원래 데이터 vs 차분 데이터 시각화
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.plot(x, marker='o')
plt.title(" 원래 시계열 (Non-Stationary)")
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(diff_x, marker='o', color='orange')
plt.title(" 1차 차분 시계열 (Stationary?)")
plt.grid(True)

plt.tight_layout()
plt.show()
````

##  np.diff(x)의 원리

`np.diff(x)`는 시계열 데이터에서 **연속된 원소의 차이**를 구하는 함수로,  
수학적으로는 **1차 차분(differencing)** 을 수행합니다.

---

###  수식 표현

$
\text{diff}_x[t] = x[t+1] - x[t], \quad \text{for } t = 0 \text{ to } N-2
$


---

###  예제

```python
import numpy as np

x = np.array([10, 12, 15, 14])
diff_x = np.diff(x)
print(diff_x)  # [2, 3, -1]


###  예시 수식

주어진 시계열 데이터가 다음과 같을 때:

$
x = [10,\ 12,\ 15,\ 14]
$

1차 차분 결과 `diff_x`는 아래와 같다:

\[
\begin{align*}
\text{diff}_x[0] &= 12 - 10 = 2 \\
\text{diff}_x[1] &= 15 - 12 = 3 \\
\text{diff}_x[2] &= 14 - 15 = -1 \\
\end{align*}
\]

---

결과적으로,

$
\text{diff}_x = [2,\ 3,\ -1]
$


### 차분하면 stationary(정상성)가 되는지 직관적으로 이해하기:

```text

 예: 주가, 매출, 트래픽 등
주가가 100 → 102 → 104 → 106 ... 이렇게 계속 오르면
평균이 시간이 지날수록 증가 → 이것은 비정상 시계열
즉, 값 자체는 계속 커지고 있어서 평균이 일정하지 않음

 그런데 "차분"을 해보면?

시계열:       [100, 102, 104, 106]
1차 차분:       [  2,   2,   2 ]
차분 결과는 늘 2 → 변하지 않음!

이제는 평균도, 분산도 시간에 따라 일정

 그래서 정상성(stationarity) 이 생김

 결론
차분은 변화량(증가나 감소) 만 남겨서 원래의 추세나 계절성 등 시간에 따라 변하는 구조를 제거.
시계열을 stationary(정상적) 으로 만들어주게됨

````python
import numpy as np

x = np.array([10, 12, 15, 14])
diff_x = np.diff(x)
print(diff_x)  # [2, 3, -1]
````
