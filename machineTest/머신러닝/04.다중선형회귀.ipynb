```python
import pandas as pd
import numpy as np

from sklearn.linear_model import LinearRegression, SGDRegressor

import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_csv('data/data-01.csv',header=None)
df.columns = ['q1','q2','mid','final']
df
```

```python
x = df.iloc[:,:-1].values
y = df.iloc[:,[-1]].values
```

```python
model = LinearRegression()
model.fit(x,y)
```

```python
print(model.coef_) # 3x1
print(model.intercept_)
```

q1: 90, q2:90, mid:95

```python
model.predict([[90,90,95]])
```

### 결론 : 머신러닝, 딥러닝 공통
---------------------------------------------
1. w(행, 열): [컬럼의 갯수, 라벨의 갯수], b:[라벨의 갯수]
2. 특성 데이터: 행x열(컬럼의 갯수)
3. 라벨 데이터: 행X열(대부분 1열)
4. 예측 : 행x열(컬럼의 갯수가 동일해야 계산이 가능)

### Quiz LinearRegress 이용 
1. 전기 생산량이 : 4,5 인 경우 전기 사용량을 예측하시오.
2. 실데이터 scatter, 예측은 라인차트

```python
df = pd.read_csv('data/electric.csv', index_col='Unnamed: 0')
df
```

```python
x = df[['전기생산량']].values
y = df[['전기사용량']].values
```

```python
# x = df.iloc[:,:-1].values
# y = df.iloc[:,[-1]].values
```

```python
x.shape
```

```python
y.shape
```

```python
model = LinearRegression()
model.fit(x,y) # w:[1,1] b:[1], hx = matmul(x,w) + b로 일반화 되어 있음 
```

```python
print(model.coef_)# 3x1
print(model.intercept_)
```

```python
model.predict([[4],[5]]) # 2x1 * 1x1
```

```python
pred = model.predict(x)

plt.xlabel('전기생산량')
plt.ylabel('전기사용량')
plt.scatter(x, y)
plt.plot(x, pred, color='red')
plt.show()
```
