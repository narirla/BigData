```python
import pandas as pd
import numpy as np

from sklearn.linear_model import LinearRegression, SGDRegressor

import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
carDF = pd.read_csv('data/cars.csv', index_col='Unnamed: 0')
carDF
```

```python
x = carDF.iloc[:,:-1]
y = carDF.iloc[:,-1]
```

```python
model = SGDRegressor( verbose=1, n_iter_no_change=50)
model.fit(x,y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
pred = model.predict(x)
```

```python
plt.scatter(x,y)
plt.plot(x, pred, 'r--')
plt.show()
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y, pred) # 회귀계수 평가 지표 0~1 사이 값
```

```python
model.score(x,y) # 결정 계수 값
# score는 특정 라벨 데이터를 주면 됨
```
