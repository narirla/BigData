```python
def sigmoid( z ):
    return 1/( 1+math.e**-z)
```

```python
print( sigmoid( -10) )
print( sigmoid(-1))
print( sigmoid(0))
print( sigmoid(1))
print( sigmoid(10))
```

```python
np.linspace( -10,10, 50)
```

```python
for z in np.linspace( -10,10, 50):
    s = sigmoid( z )
    plt.plot( z, s, 'ro')
    print(z,s)
plt.show()
```
