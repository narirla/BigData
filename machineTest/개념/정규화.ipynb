정규화는 데이터의 스케일을 조정하여 모델 학습 시 안정성과 성능을 높이는 데 사용.

---

##  scikit-learn (sklearn) 정규화 종류와 공식

### 1. Min-Max Scaling (최소-최대 스케일링)

특성(feature) 값을 0과 1 사이의 범위로 변환합니다. 데이터의 분포는 유지되지만, 이상치(Outlier)에 민감합니다.

* **sklearn 클래스:** `sklearn.preprocessing.MinMaxScaler`
* **공식:**

    $$X_{\text{norm}} = \frac{X - X_{\text{min}}}{X_{\text{max}} - X_{\text{min}}}$$

    * $X$: 원본 데이터 포인트
    * $X_{\text{min}}$: 특성의 **최솟값**
    * $X_{\text{max}}$: 특성의 **최댓값**
    * $X_{\text{norm}}$: 정규화된 값

### 2. Standard Scaling (표준화)

데이터의 평균을 0, 표준편차를 1로 조정합니다. 정규 분포(Gaussian distribution)를 따르는 데이터에 특히 유용하며, 이상치의 영향을 덜 받습니다. 이는 정규화(Normalization)보다는 **표준화(Standardization)**라고 부르는 것이 더 정확합니다.

* **sklearn 클래스:** `sklearn.preprocessing.StandardScaler`
* **공식:**

    $$z = \frac{X - \mu}{\sigma}$$

    * $X$: 원본 데이터 포인트
    * $\mu$: 특성의 **평균** (Mean)
    * $\sigma$: 특성의 **표준 편차** (Standard Deviation)
    * $z$: 표준화된 값 (Z-score)

### 3. Robust Scaling (로버스트 스케일링)

중앙값(Median)과 사분위 범위(Interquartile Range, IQR)를 사용하여 스케일링합니다. Min-Max Scaling이나 Standard Scaling보다 **이상치에 훨씬 덜 민감**합니다.

* **sklearn 클래스:** `sklearn.preprocessing.RobustScaler`
* **공식:**

    $$X_{\text{robust}} = \frac{X - Q_2}{Q_3 - Q_1}$$

    * $X$: 원본 데이터 포인트
    * $Q_1$: 특성의 **1사분위 수** (25th percentile)
    * $Q_2$: 특성의 **중앙값** (Median, 50th percentile)
    * $Q_3$: 특성의 **3사분위 수** (75th percentile)
    * $Q_3 - Q_1$: 사분위 범위 (IQR)
    * $X_{\text{robust}}$: 로버스트 스케일링된 값

### 4. Normalizer (L2 정규화)

**데이터의 개별 행(샘플)**에 대해 적용되며, 각 샘플의 벡터 길이를 1로 만듭니다. 주로 텍스트 분석이나 군집화(Clustering)에서 유클리드 거리(Euclidean distance) 기반의 유사도를 측정할 때 유용합니다.

* **sklearn 클래스:** `sklearn.preprocessing.Normalizer`
* **공식 (L2 Norm):**

    $$X_{\text{norm}} = \frac{X}{\|X\|_2}$$

    여기서 $\|X\|_2$는 L2 노름(Euclidean norm)이며 다음과 같습니다:

    $$\|X\|_2 = \sqrt{\sum_{i=1}^{n} x_i^2}$$

    * $X$: 원본 데이터 샘플 (벡터)
    * $x_i$: 벡터 $X$의 $i$번째 원소
    * $\|X\|_2$: L2 노름 (벡터의 길이)
    * $X_{\text{norm}}$: 정규화된 벡터

### 📚 요약 표

| 정규화/표준화 기법 | Scikit-learn 클래스 | 사용되는 통계량 | 주요 특징 |
| :--- | :--- | :--- | :--- |
| **Min-Max Scaling** | `MinMaxScaler` | $\text{Min}, \text{Max}$ | 데이터 범위를 $[0, 1]$로 조정. 이상치에 민감. |
| **Standard Scaling** | `StandardScaler` | $\mu$ (평균), $\sigma$ (표준편차) | 평균 $0$, 표준편차 $1$로 조정. 이상치의 영향 적음. |
| **Robust Scaling** | `RobustScaler` | $Q_2$ (중앙값), $IQR$ ($Q_3 - Q_1$) | 이상치에 **매우 강함**. |
| **Normalizer (L2)** | `Normalizer` | L2 Norm ($\|X\|_2$) | 개별 샘플(행)의 벡터 길이를 $1$로 조정. |
