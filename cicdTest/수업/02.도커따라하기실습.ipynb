#  기초실습( 이미지 다운로드 실행)

![nginx](image/nginx.png)


```bash
- nginx: 웹서버
docker pull nginx 
docker image ls


docker run --name webserver -d -p 80:80 nginx

docker run -d nginx 실행하면 
- Nginx 컨테이너를 백그라운드(-d)에서 실행.
- 호스트와 포트 매핑이 없어 외부에서 접근 불가.
- 내부적으로는 Nginx가 80 포트에서 실행되지만, 호스트에서 직접 접근할 방법이 없음.

docker run -d -p 80:80 nginx

Nginx 컨테이너를 백그라운드(-d)에서 실행.
-p 80:80을 사용하여 호스트(Windows/Linux/Mac)의 80번 포트를 컨테이너의 80번 포트에 연결.
외부에서 http://localhost:80으로 접속 가능.
웹서버실행 확인: http://localhost:80
컨테이너 목록 보기: docker ps


- 컨테이저 종료 : 컴퓨터 전원끄기
docker stop webserver
웹서버실행 확인: http://localhost:80 (페이지 열리지 않음 확인)



---

<img src='image/port.png'>

<img src='image/port1.png'>
