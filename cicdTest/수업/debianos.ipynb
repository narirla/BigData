### 데비안

 **데비안(Debian) 리눅스의 최신 개발 버전인 13버전(코드네임: Trixie)** 기반.

데비안은 리눅스 생태계에서 '뿌리'와 같은 역할을 하는 매우 중요한 운영체제. 

---

### 1. 데비안 13 (Trixie)의 특징

* **코드네임 'Trixie':** 데비안은 전통적으로 영화 *토이 스토리(Toy Story)*의 캐릭터 이름을 코드네임으로 사용. '트릭시'는 작중에 등장하는 파란색 트리케라톱스 공룡 장난감 이름.
* **Testing (Trixie):** 비교적 최신 소프트웨어를 포함하고 있으며, 차기 안정판이 될 버전.

* **안정성과 신뢰성:** 데비안은 "안정성"에 거의 집착할 정도로 엄격한 테스트를 거친다. 그래서 서버 운영이나 도커 베이스 이미지로 전 세계적으로 가장 많이 사용.

### 2. 컨테이너에 사용되는 이유

도커에서 `debian:13`이나 `node:latest`, `python:3.12` 같은 이미지를 사용했을 때 이 OS가 포함되는 경우가 많다.

* **가장 표준적인 환경:** 우분투(Ubuntu)도 이 데비안을 기초로 만들어졌다. 즉, 데비안에서 돌아가는 프로그램은 웬만하면 우분투에서도 잘 돌아간다.
* **패키지 매니저:** `apt` 또는 `apt-get`을 사용하여 필요한 프로그램을 설치.

### 3. 주요 명령어 가이드

이 OS 환경에서 자주 쓰게 될 명령어들은 다음과 같습니다.

| 작업 | 명령어 |
| --- | --- |
| **패키지 목록 업데이트** | `apt update` |
| **새 프로그램 설치** | `apt install <패키지명>` |
| **설치된 패키지 삭제** | `apt remove <패키지명>` |
| **버전 상세 확인** | `cat /etc/debian_version` |
