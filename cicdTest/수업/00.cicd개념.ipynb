## CI/CD

<img src='https://www.testaify.com/hubfs/CI-CD_Pipeline_2-01.webp'>

CI/CD는 현대적인 소프트웨어 개발 프로세스의 핵심으로, **'지속적 통합(Continuous Integration)'**과 **'지속적 서비스 제공/배포(Continuous Delivery/Deployment)'**를 결합한 개념입니다.

쉽게 말해, 개발자가 작성한 코드를 자동으로 검사하고 빌드해서 사용자에게 전달하는 **'자동화된 파이프라인'**이라고 이해하시면 됩니다.

---

## 1. CI (Continuous Integration): 지속적 통합

CI는 여러 명의 개발자가 작업한 코드가 정기적으로 공통 저장소(Git 등)에 통합되는 것을 의미합니다.

* **주요 목표:** 코드 변경 사항을 자주 통합하여 "빌드 오류"나 "코드 충돌"을 조기에 발견하는 것.
* **핵심 단계:**
1. **Code:** 개발자가 코드를 수정하고 저장소에 푸시(Push).
2. **Build:** 서버에서 코드를 컴파일하고 실행 가능한 형태로 만듦.
3. **Test:** 작성된 단위 테스트 등을 통해 코드에 결함이 없는지 자동 확인.



> **결과:** 개발자는 코드 품질을 유지하면서 더 빠르게 개발에 집중할 수 있습니다.

---

## 2. CD (Continuous Delivery & Deployment): 지속적 제공 및 배포

CD는 CI 과정을 거친 코드를 실제 서비스 환경에 반영하는 과정입니다. 여기서 **Delivery**와 **Deployment**는 약간의 차이가 있습니다.

### 지속적 제공 (Continuous Delivery)

* 빌드/테스트가 완료된 코드를 배포 가능한 상태로 유지하는 것.
* **실제 배포 여부는 '사람'이 결정**하여 수동으로 클릭할 때 배포됩니다.

### 지속적 배포 (Continuous Deployment)

* 테스트를 통과한 코드를 **자동으로 실제 운영 환경에 배포**하는 것.
* 사람의 개입 없이 고객에게 즉시 기능이 제공됩니다.

---

## 3. CI/CD를 사용하는 이유 (장점)

| 구분 | 설명 |
| --- | --- |
| **속도 향상** | 수동 작업이 줄어들어 배포 주기가 획기적으로 단축됩니다. |
| **리스크 감소** | 작은 단위로 자주 배포하므로, 문제가 생겨도 원인을 찾기 쉽습니다. |
| **품질 유지** | 자동화된 테스트가 모든 코드 변경을 감시하려 버그를 사전에 차단합니다. |
| **비용 절감** | 개발자가 배포 업무에 쏟는 시간과 에너지를 개발에 더 집중할 수 있습니다. |

---

## 4. 대표적인 CI/CD 도구

현재 실무에서 가장 많이 쓰이는 도구들은 다음과 같습니다.

* **Jenkins:** 가장 오래되고 강력한 오픈소스 도구 (자유도가 높음).
* **GitHub Actions:** GitHub 저장소와 연동이 매우 쉽고 별도의 서버 설정이 거의 필요 없음.
* **GitLab CI/CD:** GitLab 내부에 통합되어 있어 관리가 편리함.
* **CircleCI / Travis CI:** 클라우드 기반의 빠른 설정이 장점인 서비스형 도구.
