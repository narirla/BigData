```text
dbeaver 접속
jdbc:mysql://localhost:3306/your_database_name?allowPublicKeyRetrieval=true

MySQL Workbench 접속 설정 방법
워크벤치를 열고 **"Setup New Connection"**
Connection Name: 자유롭게 입력 (예: Docker-MySQL)
Hostname: 127.0.0.1 또는 localhost
Port: 3306 (Docker 실행 시 -p 앞 숫자에 쓴 번호)
Username: root
Password: Store in Vault... 버튼을 눌러 설정한 비밀번호 입력
```
