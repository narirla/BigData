# 🐳 도커 컨테이너 로그 조회 명령어

## 1. 특정 컨테이너의 로그 조회
```bash
docker logs <컨테이너_이름 또는 컨테이너_ID>

docker logs my-container
docker logs 123abc456def


2. 실시간 로그 출력 (로그 스트리밍)
docker logs -f <컨테이너_이름 또는 컨테이너_ID>
docker logs -f my-container

3. 최근 로그의 특정 라인만 조회
docker logs --tail <라인_수> <컨테이너_이름 또는 컨테이너_ID>
docker logs --tail 100 my-container

4. 타임스탬프 포함하여 로그 조회
docker logs --timestamps <컨테이너_이름 또는 컨테이너_ID>

5. 로그 조회 시 에러 출력 (stderr만 조회)
docker logs --stderr <컨테이너_이름 또는 컨테이너_ID>

6. 로그 조회 시 표준 출력만 조회
docker logs --stdout <컨테이너_이름 또는 컨테이너_ID>

7.이전 제외 현재부터 로그확인
docker logs --tail 0 -f 9c
