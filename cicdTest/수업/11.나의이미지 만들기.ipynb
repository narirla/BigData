# Dockerfile 명령어 옵션 설명

| **명령어**         | **설명**                                                                                     | **예시**                                                                                 |
|---------------------|----------------------------------------------------------------------------------------------|------------------------------------------------------------------------------------------|
| `FROM`              | Docker 이미지의 기반이 될 이미지를 지정합니다.                                               | `FROM python:3.10-slim`                                                                 |
| `LABEL`             | 메타데이터를 추가합니다. 예를 들어, 작성자, 버전 등 정보를 기록할 수 있습니다.               | `LABEL maintainer="your-name@example.com"`                                               |
| `RUN`               | 컨테이너 빌드 중에 실행할 명령어를 지정합니다. 보통 패키지 설치, 컴파일 등을 수행합니다.      | `RUN apt-get update && apt-get install -y curl`                                           |
| `COPY`              | 파일이나 디렉토리를 로컬 시스템에서 Docker 이미지 내의 지정된 경로로 복사합니다.            | `COPY ./app.py /app/`                                                                   |
| `ADD`               | `COPY`와 비슷하지만, 추가적으로 압축 파일을 자동으로 압축 해제하거나 URL에서 파일을 가져옵니다. | `ADD myapp.tar.gz /app/`                                                                 |
| `WORKDIR`           | 후속 명령어들이 실행될 작업 디렉토리를 설정합니다.                                            | `WORKDIR /app`                                                                            |
| `CMD`               | 컨테이너가 시작될 때 기본적으로 실행될 명령어를 지정합니다. 하나의 `CMD`만 사용할 수 있습니다. | `CMD ["python", "app.py"]`                                                              |
| `ENTRYPOINT`        | 컨테이너 실행 시 반드시 실행될 명령어를 지정합니다. `CMD`와 결합해서 사용할 수 있습니다.      | `ENTRYPOINT ["python"]`                                                                   |
| `EXPOSE`            | 컨테이너가 외부와 통신하기 위해 열어야 할 포트를 지정합니다.                                   | `EXPOSE 80`                                                                              |
| `ENV`               | 환경 변수 값을 설정합니다. 컨테이너 내에서 환경 변수를 사용할 수 있습니다.                    | `ENV APP_HOME /app`                                                                      |
| `VOLUME`            | 컨테이너와 호스트 시스템 간에 공유할 디렉토리(볼륨)를 지정합니다.                             | `VOLUME ["/data"]`                                                                       |
| `USER`              | 컨테이너 내에서 명령을 실행할 사용자를 설정합니다.                                            | `USER appuser`                                                                           |
| `ARG`               | 빌드 시간에 사용할 수 있는 변수를 정의합니다.                                                 | `ARG VERSION=1.0`                                                                        |
| `SHELL`             | 사용할 셸을 변경할 수 있습니다. 기본 셸을 변경할 때 사용합니다.                               | `SHELL ["/bin/bash", "-c"]`                                                              |

## 주요 사항:
- `FROM`은 **Dockerfile**에서 반드시 첫 번째로 작성해야 하는 명령어입니다.
- `CMD`와 `ENTRYPOINT`는 **컨테이너 시작 시 실행할 명령어**를 설정하지만, 서로 다르게 동작합니다:
  - `CMD`는 기본값을 설정하며, `docker run` 명령어에서 다른 값을 넘기면 덮어씁니다.
  - `ENTRYPOINT`는 항상 실행되며, `CMD`는 `ENTRYPOINT`의 추가 인자 역할을 합니다.


```bash

# 1. Python 이미지를 기반으로 설정 (Python 3.10 사용)
FROM python:3.10-slim

# 2. 작업 디렉토리 생성
WORKDIR /app

# 3. 현재 디렉토리의 파일을 컨테이너의 /app 디렉토리로 복사
COPY app.py .

# 4. 파이썬 스크립트 실행
CMD ["python", "app.py"]

```bash

docker build -t your-image-name .
docker images

docker run my-python # 실행하기


Docker는 이미지를 로컬 디스크에 저장하지만, 해당 이미지는 일반적인 
파일로 접근할 수 있는 형태가 아니고, Docker의 내부 구조에 의해 관리.


docker save 명령어를 사용하여 이미지를 .tar 파일로 저장할 수 있다.
docker save -o my-python-image.tar my-python


docker load -i my-python-image.tar



### docker-hub 에 이미지 올리기 

# Docker Hub 이미지 푸시 및 설정 절차

## 1. Docker Hub 계정 생성
먼저, Docker Hub 계정이 없으면 생성해야 합니다.

- [Docker Hub 웹사이트](https://hub.docker.com/)에 접속하여 계정을 만듭니다.
- 계정 생성 후 로그인합니다.

## 2. Docker Hub에서 레포지토리 생성
이미지를 푸시할 리포지토리를 Docker Hub에서 생성해야 합니다.

1. **로그인 후** Docker Hub 웹사이트에서 **"Repositories"** 탭으로 이동합니다.
2. **"Create Repository"** 버튼을 클릭하여 새 리포지토리를 생성합니다.
3. 리포지토리 이름, 설명, 공개/비공개 설정 등을 선택합니다.
   - **Public**: 누구나 이미지를 볼 수 있습니다.
   - **Private**: 비공개로 설정하여 해당 계정만 이미지를 볼 수 있습니다.

## 3. Docker 로그인
Docker Hub에서 이미지를 푸시하기 전에 Docker CLI를 통해 로그인해야 합니다.

- 터미널에서 아래 명령어를 실행하여 Docker에 로그인합니다:

```bash
docker login
# docker logout # 업로드후 로그아웃


Docker Hub 계정의 사용자명과 비밀번호를 입력합니다. 로그인 후 성공적으로 Docker Hub에 연결됩니다.

4. 이미지 태그 지정
docker tag my-python <docker-username>/<repository-name>:<tag>
docker tag my-python your-username/my-python:latest

5. 이미지 푸시
docker push your-username/my-python:latest

6. Docker Hub에서 이미지 관리
Docker Hub에서 푸시한 이미지를 관리할 수 있습니다.

- 이미지에 대한 설명을 추가하거나 태그를 업데이트할 수 있습니다.
- 버전 관리를 통해 latest, v1, v2 등 여러 버전으로 관리할 수 있습니다.
- Private 설정을 통해 이미지 접근을 제한할 수 있습니다

```bash
docker pull your-username/my-python:latest
docker pull vision22/my-python:latest

---

### 도커허브외 이미지 배포방법

 Docker Hub 대신 **사설 저장소(Private Registry)**를 구축하거나 직접 파일생성.

가장 대표적인 3가지 방법.

---

### 1. 직접 구축: Docker Registry

도커 자체에서 제공하는 `registry` 이미지를 사용하여 우리 회사만의 로컬 서버를 만드는 방법.

* **설치 방법:** 회사 내 서버(또는 내 PC)에서 다음 명령어를 실행.
```bash
docker run -d -p 5000:5000 --name local-registry registry:2

```

* **사용 방법:** 
1. 이미지 태그: `docker tag 내이미지 localhost:5000/내이미지:v1`
2. 업로드: `docker push localhost:5000/내이미지:v1`

* **장점:** 무료이며 설정이 매우 빠릅니다.
* **단점:** 보안 설정(SSL, 로그인 등)을 직접 추가해야 하며, 서버가 꺼지면 관리가 어렵다.

---

### 2. 전문 소프트웨어 사용: Harbor (가장 많이 사용)

회사 규모가 조금 있다면 오픈소스 프로젝트인 **Harbor**를 많이 사용.

* **특징:** * **권한 관리:** 팀원별로 읽기/쓰기 권한을 줄 수 있다.
* **보안 스캔:** 이미지에 취약점이 있는지 자동으로 검사.
* **UI 제공:** 웹 브라우저에서 업로드된 이미지를 편하게 관리할 수 있다.

* **장점:** 기업용 저장소로 손색없는 강력한 기능을 제공합니다.

---

### 3. 클라우드 비공개 저장소 (AWS, Azure 등)

회사가 클라우드를 이미 사용 중이라면 해당 서비스의 비공개 컨테이너 저장소를 쓰는 것이 가장 안전하고 관리가 편합니다.

* **AWS:** ECR (Elastic Container Registry)
* **Google Cloud:** Artifact Registry
* **Azure:** Container Registry
* **특징:** 외부 접속을 차단하고 회사 IP에서만 접근 가능하도록 설정할 수 있습니다.

---

### 4. 물리적 전달 (인터넷이 안 되는 경우)

보안상 폐쇄망이라 인터넷 자체가 안 된다면, 이미지를 **파일(`.tar`)**로 추출해서 USB 등으로 옮길 수 있습니다.

1. **이미지를 파일로 저장:**
```bash
docker save -o my_image.tar my-pandas-app:v1

```

2. **다른 컴퓨터에서 이미지 불러오기:**
```bash
docker load -i my_image.tar

```

Docker 이미지 태그 뒤에 붙는 `slim`, `alpine` 등은 이미지의 **기반이 되는 운영체제(OS)와 포함된 패키지의 범위**를 결정. 각각 용도와 장단점이 뚜렷하므로 상황에 맞춰 선택하는 것이 중요.


### Docker 이미지 태그별 비교

| 태그 종류 | 기반 OS | 특징 | 장점 | 단점 |
| --- | --- | --- | --- | --- |
| **Default** (예: `3.10`) | Debian (Full) | 가장 일반적인 이미지. 수많은 빌드 도구와 라이브러리 포함. | 호환성이 가장 좋고, 웬만한 라이브러리는 바로 빌드됨. | 이미지 크기가 매우 큼 (보통 수백 MB). |
| **-slim** | Debian (Slim) | 실행에 꼭 필요한 최소 패키지만 포함. 빌드 도구 등은 제외됨. | **안정성**과 **작은 크기**의 균형이 가장 좋음. | 컴파일이 필요한 패키지 설치 시 별도의 도구 설치가 필요함. |
| **-alpine** | Alpine Linux | 보안과 경량화에 특화된 초소형 OS 기반. | **용량이 압도적으로 작음** (수십 MB 단위). 보안 취약점 노출 적음. | 일반적인 Linux 표준(glibc) 대신 `musl`을 사용하여 일부 라이브러리와 호환성 문제 발생 가능. |
| **-windowsservercore** | Windows Server | Windows 환경 전용 이미지. | Windows 전용 종속성이 있는 앱 실행 가능. | 용량이 매우 크고 Linux 호환 안 됨. |

---

### 어떤 것을 선택해야 할까요?

1. **개발 초기 또는 복잡한 라이브러리 사용 시:** 일반 **Default** 이미지
* `pandas`, `numpy` 등 C 컴파일러가 필요한 라이브러리를 쓸 때 설정을 최소화하고 싶을 때 적합.


2. **운영(Production) 환경 배포 시:** **`-slim`** (가장 권장)
* 보통 `slim` 버전으로 시작해서 필요한 패키지만 `apt-get`으로 추가하는 방식이 가장 안정적이고 효율적입니다.


3. **마이크로서비스 및 극단적 경량화:** **`-alpine`**
* 용량을 최소화해야 하거나 보안이 극도로 중요할 때 사용합니다. 다만, Python의 경우 `alpine`에서 빌드 속도가 오히려 느려지거나 호환성 이슈가 잦아 주의가 필요합니다.
