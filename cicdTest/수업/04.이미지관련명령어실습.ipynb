이미지 목록보기
```bash
docker images 
docker image ls
docker images -q # 또는 --quiet : 이미지 아이디컬럼만 보여줌
```

이미지 삭제
```bash

docker rmi nginx:stable-perl  #이미지명
또는
docker rmi <이미지아이디:일부만입력됨>

docker images 로 삭제 확인
docker rmi nginx
docker rmi b52 ( 컨테이너 사용중 중단된 이미지는 에러발생)

Error response from daemon: conflict: unable to delete b52e0b094bc0 (must be forced) 
- image is being used by stopped container 5fd1f929e193
```

------------------------
```bash
docker rmi -f b52
docker imgages rm -f b52
```

컨테이너 처리
```bash
docker ps #실행 중인 모든 컨테이너 목록 보기
docker ps -q # 실행 중인 컨테이너 ID만 보기
docker ps -a #모든 컨테이너 목록 보기 (실행 중이지 않은 컨테이너 포함)

docker stop webserver #컨테이너에서 현재 실행중이미지는 컨테이너 중단후 
docker rm 컨테이너아이디 # 컨테이너 삭제

docker rm -f 컨테이너 아이디  #중지후 삭제 한번에


docker stop webserver

docker rmi $(docker images -q) #컨테이너 사용하지 않는 모든 이미지 삭제

docker rmi -f $(docker images -q)  #컨테이너 중단 포함함
또는
docker rmi -f `docker images -q`  #컨테이너 중단 포함함



`$( )` 문법:

 리눅스/유닉스 셸(Shell)에서 사용하는 **명령어 치환(Command Substitution)** 기능.

---

### 1. 의미와 작동 방식

`$(명령어)`는 **"괄호 안에 있는 명령어를 먼저 실행하고, 그 결과값(출력물)을 그 자리에 텍스트로 치환하라"**는 뜻.

작동 순서

1. 먼저 `docker images -q`를 실행하여 **이미지 ID 리스트**를 뽑아냅니다.
2. 그 결과값(예: `fb01117203ff ba6acccedd29`)이 `$( )` 자리에 그대로 들어갑니다.
3. 최종적으로 `docker rmi -f fb01117203ff ba6acccedd29`라는 명령어가 완성되어 실행됩니다.

---

### 2.  예시

이미지가 3개 있다면 .

* **사람이 직접 입력할 때:**

`docker rmi -f id1 id2 id3` (일일이 ID를 복사해서 붙여넣어야 함)

* **`$( )`를 사용할 때:**
`docker rmi -f $(docker images -q)` (도커가 알아서 ID를 뽑아다 채워줌)

---

### 3. 유사한 문법 (백틱)

오래된 스크립트에서는 `$( )` 대신 **백틱(Backticks, ```)** 을 사용가능.

```bash
docker rmi -f `docker images -q`
```
두 방식은 기능적으로 동일

---

### 4. 자주 쓰이는 응용 패턴

도커 관리 시 이 문법을 활용하면 굉장히 편리.

* **모든 컨테이너 중지하기:** `docker stop $(docker ps -aq)`
* **특정 이름이 포함된 이미지 삭제하기:** `docker rmi $(docker images | grep "test" | awk '{print $3}')`
