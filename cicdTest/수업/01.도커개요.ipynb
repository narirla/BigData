```bash
FROM python:3.10
COPY . /app
RUN pip install -r requirements.txt
CMD ["python", "app.py"]

#결론:Dockerfile을 통해 누구나 동일한 환경 구성 가능

 ![컨테이너](image/docker_container.png)

## 도커의 구성

<img src='https://storage.googleapis.com/static.fastcampus.co.kr/prod/uploads/202309/190146-1143/%EB%8F%84%EC%BB%A4-%EA%B5%AC%EC%A1%B0.png'>


- 이미지(Image): 어플리케이션을 실행하기 위한 모든 파일과 설정을 담고 있는 템플릿.

- 컨테이너(Container): 이미지를 기반으로 생성되어 실제로 어플리케이션을 실행하는 인스턴스입니다.

- 도커 허브(Docker Hub): 도커 이미지를 저장하고 공유할 수 있는 중앙 레포지토리입니다.(https://hub.docker.com/)


구분의  가장 쉬운 비유는 **'요리 레시피'**와 **'실제 요리'**, 그리고 **'레시피 공유 사이트'**로 생각하는 것.

###  도커 핵심 개념 비교

| 구분 | **도커 이미지 (Image)** | **도커 컨테이너 (Container)** | **도커 허브 (Docker Hub)** |
| --- | --- | --- | --- |
| **비유** | **설계도 / 레시피** | **실제 제품 / 완성된 요리** | **앱스토어 / 레시피 도서관** |
| **상태** | 실행되지 않은 정적인 파일 상태 | 이미지가 실행되어 살아있는 상태 | 이미지들을 저장하고 공유하는 공간 |
| **특징** | 한 번 만들어지면 변하지 않음 (Read-only) | 이미지를 기반으로 생성되며 데이터 수정 가능 | 누구나 이미지를 올리거나 다운로드 가능 |
| **역할** | 환경 설정, 소스 코드 등을 하나로 압축 | 실제 서비스가 돌아가는 독립적인 환경 | 필요한 환경(DB, 웹서버 등)을 검색해 가져옴 |


- 도커 엔진(Docker Engine): 그 요리를 직접 만들고 관리하는 **'주방장'이자 '식당 시스템 전체'** 라고 할 수 있다.
사용자가 "이 이미지를 실행해줘!"라고 명령을 내리면, 그 명령을 받아서 실제로 컨테이너를 생성하고 실행하는 핵심 소프트웨어.


---

###  흐름으로 이해하기 (Workflow)

도커를 사용하는 과정은 보통 다음과 같은 순서로 이루어집니다.

1. **Build (빌드):** 내가 만든 프로그램을 **도커 이미지**로 만듭니다.
2. **Push (푸시):** 만든 이미지를 **도커 허브**에 올려서 보관합니다.
3. **Pull (풀):** 필요할 때 어디서든 도커 허브에서 **이미지**를 다운로드합니다.
4. **Run (실행):** 이미지를 실행시켜서 **컨테이너**라는 독립된 방을 만듭니다.

###  한 줄 요약

* **이미지:** 프로그램을 실행하기 위한 모든 것이 담긴 **"설치 파일"**
* **컨테이너:** 그 설치 파일을 실행해서 돌아가고 있는 **"실제 프로그램"**
* **도커 허브:** 전 세계 사람들이 만들어 놓은 설치 파일이 모인 **"클라우드 저장소"**


<img src='https://img1.daumcdn.net/thumb/R1280x0/?scode=mtistory2&fname=https%3A%2F%2Fblog.kakaocdn.net%2Fdna%2FFNAzM%2FbtsB1tLAprc%2FAAAAAAAAAAAAAAAAAAAAAKvWG61g0olc2LerYPjk5ed2o-MxWtCTUF4ApiGsgtTl%2Fimg.png%3Fcredential%3DyqXZFxpELC7KVnFOS48ylbz2pIh7yKj8%26expires%3D1767193199%26allow_ip%3D%26allow_referer%3D%26signature%3Dm2SFISUUNIMMyNzJnyi6vbRbKw8%253D'>

### 1. 도커엔진과  도커 호스트

* **도커 호스트 (Docker Host):** 도커 엔진이 설치되어 실행되고 있는 **실제 컴퓨터(서버)** 그 자체를 말합니다. (하드웨어 + OS 포함)
* **도커 엔진 (Docker Engine):** 그 컴퓨터 안에서 컨테이너를 관리하는 **핵심 소프트웨어(프로그램)**입니다.

---

### 2. 비유로 이해하기 (공장)

| 항목 | 비유 | 설명 |
| --- | --- | --- |
| **도커 호스트** | **공장 부지/건물** | 전기가 들어오고 기계가 놓여있는 실제 장소 (내 노트북이나 서버) |
| **도커 엔진** | **공장 관리 시스템** | 공장 안에서 기계를 돌리고 물건을 생산하도록 지시하는 시스템 |
| **도커 컨테이너** | **제품 (컨테이너 상자)** | 공장에서 실제로 만들어져 작동하는 결과물 |

---

### 3. 관계 설명

1. **호스트는 엔진을 품고 있습니다:** 사용자님의 윈도우 노트북에 도커를 설치했다면, 지금 사용자님의 **노트북**이 '도커 호스트'가 된 것이고, 그 안에서 돌아가는 **프로그램**이 '도커 엔진'입니다.
2. **명령의 흐름:** 사용자(나)가 명령을 내리면 → **도커 호스트** 안에 있는 **도커 엔진**이 그 명령을 알아듣고 → 컨테이너를 만들거나 지웁니다.
3. **상호작용:** 우리가 "도커 호스트에 접속한다"라고 하면 보통 그 컴퓨터에 원격으로 들어가는 것을 의미하고, "도커 엔진을 재시작한다"라고 하면 소프트웨어 서비스를 껐다 켜는 것을 의미합니다.

---

### 4. 윈도우에서의 특이점 (중요!)

사용자님이 윈도우에서 도커를 쓰실 때는 한 단계가 더 숨어 있습니다.

* **실제 호스트:** 사용자님의 윈도우 PC
* **가상 호스트:** WSL2 (리눅스 커널 환경)
* **도커 엔진:** 그 WSL2 안에서 돌아가는 소프트웨어

윈도우는 리눅스가 아니기 때문에, 윈도우 PC가 직접 호스트 역할을 완벽히 수행하지 못하고 **WSL2라는 리눅스 환경을 가상으로 만들어 이를 '도커 호스트'처럼 활용**하게 됩니다.


> **"도커 엔진은 도커 호스트(컴퓨터) 위에서 실행되는 핵심 관리 프로그램이다."**


#  컨테이너의 독립성

---

####  컨테이너의 독립성이란?

- 컨테이너는 **서로 독립적으로 실행**됨  
- 하나의 컨테이너에서 발생한 문제가  
  다른 컨테이너에 영향을 주지 않음  

---

####  각각의 컨테이너는

- **자신만의 파일 시스템** 사용  
- **고유한 네트워크 인터페이스** 보유  (각자의 ip와 port 보유)
- **자체 프로세스 공간**에서 동작  

> 독립적인 환경을 갖는 작은 가상 시스템이라고 볼 수 있음

---

####  예시

```text
웹 서버 컨테이너   : nginx  
DB 서버 컨테이너   : mysql  
캐시 서버 컨테이너 : redis


--- 

## 커널이란?

*운영체제(Windows, Linux, macOS)* 는  **[커널 + 서비스 프로그램 + UI(화면)]** 를 모두 합친 패키지 상품 같은 것, 그중에서 실질적으로 컴퓨터를 움직이게 하는 본체가 바로 커널.

### 1. 커널의 역할 (비유: 레스토랑)

컴퓨터를 하나의 **레스토랑**이라고 가정.

* **하드웨어 (주방 기구):** 프라이팬, 가스레인지, 냉장고 (CPU, 메모리, 디스크)
* **응용 프로그램 (손님):** "스테이크 주문할게요!", "물 좀 주세요!" (웹 브라우저, 게임등 )

* **커널 (주방장):** 손님의 주문을 받고, 어떤 프라이팬을 쓸지, 불은 얼마나 세게 할지 **직접 하드웨어를 조작하는 사람**

> **결론:** 소프트웨어(손님)가 하드웨어(주방 기구)를 직접 건드리면 사고가 날 수 있기 때문에, **가운데서 이를 안전하게 관리하고 연결해 주는 핵심 관리자**가 바로 커널입니다.

---

### 2. 커널이 하는 4가지 주요 업무

| 기능 | 설명 (쉽게 풀이) |
| --- | --- |
| **자원 관리** | CPU가 어떤 프로그램을 먼저 처리할지 순서를 정합니다. (스케줄링) |   
| **메모리 관리** | 각 프로그램이 컴퓨터의 메모리(RAM)를 어디서부터 어디까지 쓸지 나눠준다.     |
| **장치 제어** | 키보드, 마우스, 모니터 등 하드웨어와 소프트웨어가 대화하게 해준다.     |
| **보안 및 격리** | 프로그램이 다른 프로그램의 영역을 침범하지 못하게 벽을(격리) 쳐준다.    |

---

### 3. 왜 도커에서 커널이 중요한가요?

1. **커널은 무겁습니다:** 운영체제에서 가장 만들기도 어렵고 용량도 큽니다.
2. **도커의 전략:** "이미 내 컴퓨터(호스트)에 성능 좋은 **리눅스 커널(주방장)**이 있는데, 컨테이너마다 주방장을 새로 고용할 필요가 있을까? 그냥 한 명의 주방장이 여러 개의 주문(컨테이너)을 처리하게 하자!"
3. **결과:** 컨테이너에는 주방장이 없고 **'레시피(사용자 환경)'**만 들어있어서 아주 가볍고 빠르게 실행되는 것입니다.

###  요약

**커널은 소프트웨어의 명령을 하드웨어로 전달하고 자원을 배분하는 "컴퓨터의 두뇌이자 지휘자"입니다.**

 윈도우에서 도커를 돌리기 위해  **WSL2(Windows Subsystem for Linux)**라는 '리눅스 커널'을 따로 설치필요(도커는 리눅스 커널을 사용함)

 **컨테이너는 자기만의 커널을 가지고 있지 않고, 호스트(내 컴퓨터)에 있는 리눅스 커널을 공유해서 사용**. 이것이 도커가 가볍고 빠른 핵심 이유.

---

### 1. 도커의 '공유' 구조

일반적인 가상 머신(VM)과 도커 컨테이너의 가장 큰 차이점은 **커널을 새로 만드느냐, 빌려 쓰느냐**의 차이.

* **가상 머신 (VM):** 하드웨어 위에 가상 커널(Guest OS)을 통째로 올립니다. 컴퓨터 안에 또 다른 컴퓨터를 통째로 만드는 식이라 무겁습니다.

* **도커 컨테이너:** 호스트 OS의 리눅스 커널을 그대로 빌려 씁니다. 대신 **리눅스 커널의 특수 기능**들을 사용해 마치 독립된 공간인 것처럼 속입니다.


### 2. 어떻게 커널 하나를 나눠 쓰나요? (핵심 기술)

리눅스 커널에는 컨테이너를 가능하게 하는 두 가지 핵심 요소:

1. **네임스페이스 (Namespaces):** "너는 이 방(프로세스, 네트워크 등)만 봐"라고 **시야를 격리**시키는 기술입니다. 컨테이너는 이 기술 덕분에 자신이 독립된 운영체제라고 착각합니다.

2. **Cgroups (Control Groups):** "너는 CPU를 이만큼만 써"라고 **자원 사용량을 제한**하는 기술입니다.


### 3. 왜 리눅스 커널이 꼭 필요한가요?

도커는 원래 **리눅스의 이런 기능들을 활용하기 위해 만들어진 기술**입니다. 그래서 윈도우나 맥에서 도커를 쓸 때도 내부적으로 아주 얇은 리눅스 층(WSL2 등)을 깔아서 그 커널을 빌려 쓰는 것입니다.

---

###  요약표

| 항목 | 가상 머신 (VM) | 도커 컨테이너 (Container) |
| --- | --- | --- |
| **커널 (Kernel)** | **개별 보유** (독립적) | **호스트 커널 공유** (빌려 씀) |
| **자원 효율** | 낮음 (OS 자체가 리소스를 많이 먹음) | 매우 높음 (필요한 앱만 돌림) |
| **실행 속도** | 느림 (부팅 과정 필요) | 매우 빠름 (프로세스 실행 수준) |

> **정리하자면:** 컨테이너 안에는 리눅스 OS의 "껍데기(파일, 라이브러리 등)"는 들어있지만, "심장(커널)"은 호스트의 것을 함께 사용한다고 이해





#  컨테이너 운영체제는 Linux만?

---

####  컨테이너 = 리눅스?

- Docker 컨테이너는 원래 **리눅스 커널 기술(LXC)** 기반  
- 리눅스의 `cgroups`, `namespaces` 등을 활용

> 즉, 전통적인 컨테이너는 **리눅스 기반** 임

---

- **Windows 컨테이너**도 존재함
- Windows Server에서 **Windows 기반 컨테이너 실행** 가능  
- Docker는 **Linux 컨테이너**와 **Windows 컨테이너**를 구분해 실행

---

####  운영체제별 컨테이너 종류

| 운영체제        | 컨테이너 가능 여부 | 비고                         |
|----------------|-------------------|------------------------------|
| **Linux**       |  가능             | 표준적인 Docker 컨테이너 환경 |
| **Windows**     |  가능             | Windows 전용 컨테이너 지원     |
| **macOS**       |  가능 (간접적)     | 실제로는 Linux VM 위에 실행    |

---

####  macOS에서 도커는?

- macOS는 리눅스 커널이 아니기 때문에  
  Docker Desktop이 내부적으로 **Linux VM (예: Alpine)** 실행

> macOS → VM → Linux → Docker 컨테이너

---

####  결론

- 컨테이너는 **리눅스 기반이 기본**  
- 하지만 **Windows 컨테이너도 존재**  
- macOS는 **리눅스 VM을 통한 우회 방식** 사용



# 도커 이미지의 핵심 개념

---

##  이미지(Image)란?

- 애플리케이션, 라이브러리, 설정 등 **모든 구성 요소 포함**( 프로그램을 실행하기 위한 모든 것을 포함)
- 변경 불가능한 **불변(Immutable)** 구조
- 복잡한 설치과정 없이 이미지를 컨테이너에 장착하면 모든 환경구성 완료

---

##  구성 요소

1. **Base Image** (예: `ubuntu`, `node`, `python`)
2. **추가 파일/라이브러리**  
3. **명령어** (CMD, RUN 등)

> 모두 Dockerfile에 정의되어 있음

---

##  이미지 vs 컨테이너

| 항목        | 설명                                   |
|-------------|----------------------------------------|
| **이미지**   | 컨테이너의 설계도 / 정적 / 읽기 전용       |
| **컨테이너** | 이미지로부터 만들어진 실행 인스턴스 / 동적 |

---

##  이미지 재사용

- 한 번 만든 이미지는 여러 컨테이너에서 재사용 가능
- 다양한 환경에서 일관된 실행 보장

```bash
docker run myimage:v1


#  컨테이너와 이미지의 관계

---

##  기본 개념

- **컨테이너**는 **이미지로부터 생성**됨
- 보통 **컨테이너 하나당 이미지 하나** 사용

```bash
docker run myimage:latest



#  Docker 엔진 설치 방법
### https://www.docker.com/ 에서 다운로드후 설치
---

##  운영체제별 Docker 실행 방식

| 운영체제        | Docker 엔진 실행 방식                          | 설명 |
|----------------|-----------------------------------------------|------|
| **Linux**       | 직접 실행 가능 (Native)                       | 리눅스 커널 기반으로 바로 실행 가능 |
| **Windows**     | Docker Desktop + WSL 2 or Windows 컨테이너    | 내부적으로 WSL2 리눅스 커널 사용 |
| **macOS**       | Docker Desktop (내부 Linux VM 사용)           | HyperKit 또는 Apple Virtualization 사용 |

---
