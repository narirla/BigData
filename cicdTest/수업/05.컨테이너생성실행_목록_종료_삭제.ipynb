# 🐳 도커 컨테이너 생성 명령어

## 1. 도커 컨테이너 생성 (실행하지 않고 생성만)
```bash
docker create <옵션> <이미지_이름>
docker create mysql
* create 시 mysql 이미지가 없으면 다운후 컨테이터 실행

2. 컨테이너 이름 지정하여 생성
docker create --name <컨테이너_이름> <이미지_이름>
docker create --name my-container ubuntu

3. 포트 매핑하여 컨테이너 생성
docker create -p <호스트_포트>:<컨테이너_포트> <이미지_이름>
docker create -p 8080:80 nginx

4. -e 옵션을 사용하여 컨테이너 내에서 환경 변수를 설정하여 생성
docker create -e MY_ENV_VAR=value ubuntu

5. 볼륨 마운트하여 컨테이너 생성
docker create -v <호스트_디렉토리>:<컨테이너_디렉토리> <이미지_이름>

6. 커맨드 실행을 위한 컨테이너 생성
docker create <이미지_이름> <명령어>
docker create ubuntu bash


docker create는 컨테이너를 생성만 하고, 실행을 하지 않으므로, 
docker start <컨테이너_이름> 명령어로 컨테이너를 실행해야 한다.




```bash
docker pull nginx
docker create --name my-container -p 8080:80 nginx
docker start my-container

--> docker run --name my-container -p 8080:80 nginx

# 🐳 도커 컨테이너 실행 명령어

## 1. 기존 컨테이너 실행
```bash
docker start <컨테이너_이름 또는 컨테이너_ID:일부만입력됨>
docker start my-container
docker start 123abc456def

2. 백그라운드에서 컨테이너 실행
docker start -d <컨테이너_이름 또는 컨테이너_ID>
docker start -d my-container

3. 특정 명령어 실행하면서 컨테이너 실행
docker start -a <컨테이너_이름 또는 컨테이너_ID>
docker start -a my-container

4. 도커 컨테이너와 상호작용 모드로 실행 (터미널 연결)
docker exec -it <컨테이너_이름 또는 컨테이너_ID> <명령어>

5. 포트 매핑하여 실행 중인 컨테이너 접속
docker exec -it -p <호스트_포트>:<컨테이너_포트> <컨테이너_이름 또는 컨테이너_ID> <명령어>



# 🐳 도커 이미지 실행 명령어

## 1. 도커 이미지 실행 (기본 명령어)
```bash
docker run <이미지_이름>

ex) docker run ubuntu

docker run 명령어는 이미지가 로컬에 없으면 자동으로 다운로드(pull) 하고, 이미지가 있으면 해당 이미지를 기반으로 컨테이너를 생성하고 실행

결론 : 모두처리 ==> 이미지다운(없으면) + 컨테이너생성 + 실행
- docker pull + docker create + docker start

2. 실행 후 컨테이너에 명령어 전달
```bash
docker run <이미지_이름> <명령어>
docker run ubuntu echo "Hello, Docker!"

3. 컨테이너를 백그라운드에서 실행
```bash
docker run -d <이미지_이름>
docker run -d nginx


4. 포트 매핑 (호스트와 컨테이너 간 포트 연결)
```bash

docker run -p <호스트_포트>:<컨테이너_포트> <이미지_이름>
docker run -p 8080:80 nginx

-p 호스트포트:컨테이너포트
http://localhost:80
-p 80:80 → "호스트의 80번 포트를 컨테이너의 80번 포트에 연결"

ex ) docker run -d -p 80:8080 my-custom-nginx
컨테이너 내부의 /etc/nginx/nginx.conf 파일을 수정하여 8080 포트에서 실행되도록 변경해야 함.


5. 이름을 지정하여 컨테이너 실행

```bash

docker run --name <컨테이너_이름> <이미지_이름>
docker run --name my-container ubuntu


6. 상호작용 모드로 실행 (터미널 연결)
```bash
docker run -it <이미지_이름> <명령어>
docker run -it ubuntu bash


7. 볼륨 마운트 (호스트 디렉토리와 컨테이너 디렉토리 연결)
```bash
docker run -v <호스트_디렉토리>:<컨테이너_디렉토리> <이미지_이름>
docker run -v /host/data:/container/data ubuntu


실행된 컨테이너는 docker ps 명령어로 확인할 수 있다.

-d 옵션을 사용할 때, 실행된 컨테이너의 로그를 보려면 docker logs <컨테이너_이름> 명령어를 사용.

# 🐳 실행 중인 도커 컨테이너 목록 확인 명령어

```bash
1. 실행 중인 모든 컨테이너 목록 보기

docker ps

2. 실행 중인 컨테이너 ID만 보기
docker ps -q

3. 모든 컨테이너 목록 보기 (실행 중이지 않은 컨테이너 포함)
docker ps -a

4. 특정 컨테이너의 상태 및 정보 보기
docker ps -f "name=<컨테이너_이름>"
docker ps -f "name=my-container"

5. 컨테이너 목록을 더 자세히 보기
docker ps --format "{{.ID}}: {{.Names}}"
출력예)
123abc456def: nginx-container
789ghi123jkl: my-container

--format 옵션을 사용하여 출력 형식을 커스터마이징할 수 있다.



# 🐳 도커 컨테이너 종료 명령어

## 1. 특정 컨테이너 종료
```bash
docker stop <컨테이너_이름 또는 컨테이너_ID>
docker stop my-container
docker stop 123abc456def

2. 여러 개의 컨테이너 종료
docker stop <컨테이너_이름1> <컨테이너_이름2> ...
docker stop container1 container2

3. 실행 중인 모든 컨테이너 종료
docker stop $(docker ps -q)

4. 강제로 컨테이너 종료
docker kill <컨테이너_이름 또는 컨테이너_ID>
docker kill my-container

docker stop 명령어는 컨테이너에 종료 신호를 보내어 정상 종료를 시도.
docker kill 명령어는 즉시 프로세스를 종료.


# 🐳 도커 컨테이너 삭제 명령어

## 1. 실행 중인 컨테이너 삭제
```bash
docker rm <컨테이너_이름 또는 컨테이너_ID>
docker rm my-container
docker rm 123abc456def

2. 실행 중인 컨테이너를 강제로 삭제
docker rm -f <컨테이너_이름 또는 컨테이너_ID>
docker rm -f my-container

3. 모든 중지된 컨테이너 삭제
docker rm $(docker ps -a -q)
docker rm $(docker ps -qa)

4. 특정 컨테이너 여러 개 삭제
docker rm <컨테이너_이름1> <컨테이너_이름2> ...

5. 사용되지 않는 모든 중지된 컨테이너 삭제 (정리)
docker container prune
