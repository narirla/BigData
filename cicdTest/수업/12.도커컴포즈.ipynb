### Docker Compose

```bash
Docker Compose는 다중 컨테이너 Docker 애플리케이션을 정의하고 관리하는 도구.
하나의 docker-compose.yml 파일을 사용하여 여러 컨테이너를 구성하고 실행할 수 있다.


## 주요 옵션 및 설명

| 옵션                | 설명                                                                                                                                         | 예시                                                            |
|---------------------|----------------------------------------------------------------------------------------------------------------------------------------------|-----------------------------------------------------------------|
| `version`           | Docker Compose 파일의 버전을 지정합니다. 보통 최신 버전을 사용하는 것이 좋습니다. (`'3.8'`, `'3'`, `'2'` 등)                               | `version: '3.8'`    최신도커에서는 자동으로 처리(생략함)                                            |
| `services`          | 여러 컨테이너(서비스)를 정의하는 섹션입니다. 각 서비스는 독립적인 컨테이너로 실행됩니다.                                                     | `services: mysql:`                                              |
| `image`             | 사용할 Docker 이미지를 지정합니다. 이미지를 직접 지정하거나 Docker Hub에서 가져올 수 있습니다.                                              | `image: mysql:8`                                                |
| `build`             | Dockerfile을 사용하여 이미지를 빌드하는 옵션입니다. `context`와 `dockerfile`을 사용하여 빌드 경로를 지정할 수 있습니다.                      | `build: ./app`                                                  |
| `context`           | 빌드를 위한 컨텍스트 디렉토리 경로를 지정합니다. 보통 현재 디렉토리(`.`)로 설정합니다.                                                        | `build: .`                                                       |
| `dockerfile`        | 사용하고자 하는 Dockerfile의 경로를 지정합니다. 기본적으로 `Dockerfile`을 찾지만, 다른 이름을 지정할 수 있습니다.                          | `dockerfile: Dockerfile.dev`                                     |
| `ports`             | 호스트와 컨테이너 간의 포트를 매핑합니다. `host_port:container_port` 형식으로 작성합니다.                                                    | `ports: - "8080:80"`                                            |
| `environment`       | 환경 변수들을 설정하는 옵션입니다. 서비스에서 필요한 환경 변수를 설정합니다.                                                                  | `environment: - MYSQL_ROOT_PASSWORD=root`                        |
| `volumes`           | 호스트와 컨테이너 간의 디렉토리 또는 파일을 매핑합니다. 데이터 지속성을 위해 사용됩니다. `host_path:/container_path` 형식입니다.            | `volumes: - ./data:/var/lib/mysql`                               |
| `networks`          | 서비스를 연결할 네트워크를 정의합니다. 기본적으로 Docker Compose는 각 서비스를 위한 기본 네트워크를 생성하지만, 커스텀 네트워크를 사용할 수 있습니다. | `networks: - my_network`                                         |
| `depends_on`        | 특정 서비스가 다른 서비스에 의존하는 경우, 의존성을 정의합니다. 하나의 서비스가 시작되기 전에 다른 서비스가 실행되도록 설정합니다.             | `depends_on: - db`                                              |
| `restart`           | 컨테이너가 종료될 때 자동으로 재시작할지 여부를 설정합니다.                                                                                   | `restart: always`                                               |
| `command`           | 컨테이너가 시작될 때 실행할 명령어를 지정합니다.                                                                                              | `command: ["python", "app.py"]`                                 |
| `entrypoint`        | 컨테이너의 초기 실행 명령어를 설정합니다. `command`와 다르게 컨테이너가 시작될 때 기본 실행 프로그램을 지정합니다.                          | `entrypoint: ["python", "app.py"]`                              |
| `logging`           | 컨테이너의 로그 수집 방법을 설정합니다. 다양한 드라이버를 선택할 수 있습니다.                                                                 | `logging: driver: json-file`                                     |
| `extra_hosts`       | 호스트 이름을 `/etc/hosts`에 추가하는 옵션입니다.                                                                                             | `extra_hosts: - "somehost:192.168.1.1"`                         |
| `healthcheck`       | 컨테이너의 상태를 모니터링하는 옵션입니다. 주기적인 건강 상태 점검을 설정할 수 있습니다.                                                       | `healthcheck: test: ["CMD", "curl", "-f", "http://localhost"]` |


## 도커컴포즈 실행

```yaml
docker-compose up --build
docker-compose up --build -d
docker-compose build  # 이미지를 수동으로 빌드 
docker-compose up  # 빌드된 이미지를 실행(이미지가있는 경우는 )


```yaml
services:
  db:        # 서비스의 이름. 다른 서비스에서 'db'라는 이름으로 호출할 수 있다.
    image: mysql:5.7     # Docker Hub에서 공식 MySQL 5.7 버전을 내려받아 사용.
    environment:         # 컨테이너 안에서 사용할 환경 변수 설정.
      MYSQL_ROOT_PASSWORD: rootpassword # MySQL 관리자(root) 계정의 비밀번호를 설정.
      MYSQL_DATABASE: testdb         # 컨테이너 실행 시 자동으로 생성할 데이터베이스 이름.
      MYSQL_USER: testuser           # 추가로 생성할 일반 사용자 이름.
      MYSQL_PASSWORD: testpassword   # 위 일반 사용자의 비밀번호.
    ports:                           # 포트 포워딩 설정.
      - "3306:3306"                  # [내 PC 포트]:[컨테이너 포트]. 내 PC에서 3306으로 접속하면 컨테이너로 연결.
    volumes:                         # 데이터 보존 및 파일 연결을 위한 볼륨 설정.
      - D:\dockerTest\dockercompose_python_mysql\mysql-data:/var/lib/mysql 
      # 내 PC의 실제 폴더와 MySQL의 데이터 저장 폴더를 연결하여 컨테이너가 꺼져도 데이터가 보존된다..
      - .\init.sql:/docker-entrypoint-initdb.d/init.sql
      # 내 PC의 init.sql 파일을 컨테이너 내부의 초기화 폴더로 넣는다. DB 실행 시 이 SQL이 자동 실행.

  python-app:            # 파이썬 서비스의 이름입니다.
    build: ./python-app       
    # ./python-app 폴더 안에 있는 Dockerfile을 찾아 이미지를 직접 빌드.
    depends_on:      # 의존성 설정.
      - db       
      # db 서비스가 먼저 실행된 후에 파이썬 앱을 실행하도록 순서를 정한다.
    volumes:        # 코드 수정을 실시간으로 반영하기 위한 설정.
      - D:\dockerTest\dockercompose_python_mysql\python-app:/usr/src/app
      # 내 소스 코드 폴더를 컨테이너 내부 작업 폴더와 동기화합니다. 코드 수정 시 바로 반영.
    working_dir: /usr/src/app     
    # 컨테이너 실행 시 명령어가 실행될 기본 디렉토리.
    command: python app.py    
    # 컨테이너가 시작될 때 실행할 명령어(앱 실행).

volumes:       # 이 파일 전체에서 사용할 '이름이 있는 볼륨'들을 정의.
  mysql-data:     # mysql-data라는 이름의 볼륨을 생성.



 ### 빌드후 단축키

 `v`, `o`, `w`는 **Docker Compose V2**에서 지원하는 **단축키(Interactive Shortcuts)**
 터미널이 켜져 있는 상태에서 키보드의 해당 알파벳을 누르면 다음 기능들이 실행.

---

### 1. `v` (View in Docker Desktop)

* **기능:** 현재 실행 중인 프로젝트를 **Docker Desktop(GUI 프로그램)** 화면으로 바로 띄워줍니다.
* **언제 쓰나요?** 터미널 로그 대신 마우스로 편리하게 컨테이너 상태를 확인하거나, 개별 컨테이너의 CPU/메모리 사용량을 보고 싶을 때 유용합니다.

### 2. `o` (View Config)

* **기능:** 현재 적용된 **최종 Docker Compose 설정 내용**을 터미널에 출력합니다.
* **언제 쓰나요?** 여러 개의 설정 파일을 합쳐서 사용하거나, 변수가 제대로 치환되었는지 등 내가 작성한 `docker-compose.yml`이 실제로 어떻게 해석되었는지 확인하고 싶을 때 사용.

### 3. `w` (Enable Watch)

* **기능:** **파일 변경 감지(Watch Mode)** 기능을 활성화합니다.
* **언제 쓰나요?** `docker-compose.yml`에 `develop: watch` 설정이 되어 있는 경우, 코드를 수정하고 저장할 때마다 컨테이너를 자동으로 업데이트해 줍니다. (현재 설정에는 없으므로 지금은 크게 의미가 없다.)

---

###  `exited with code 0`

로그 마지막의 `python-app-1 exited with code 0`은 에러가 아니라, **"프로그램이 할 일을 다 마치고 정상적으로 종료되었다"**는 뜻입니다. 파이썬 스크립트가 모든 데이터를 출력하고 끝났기 때문에 컨테이너도 함께 종료된 것.



---

### 1. 현재 터미널에서 즉시 빠져나오기 (가장 일반적)

키보드에서 **`Ctrl + C`** 

* **동작:** 실행 중인 모든 컨테이너(db, python-app)에 정지 신호를 보냅니다.
* **결과:** 컨테이너가 정지(Stopped) 상태가 되며 터미널 입력창으로 돌아옵니다.

### 2. 컨테이너를 정지하고 완전히 삭제하기

`Ctrl + C`로 빠져나온 후, 혹은 다른 터미널 창에서 다음 명령어를 입력합니다.

```bash
docker-compose down

```

* **동작:** 생성된 컨테이너와 내부 네트워크를 완전히 제거합니다.
* **특징:** 다음에 다시 `docker-compose up`을 하면 컨테이너를 새롭게 생성합니다. (단, `mysql-data` 폴더와 연결된 DB 데이터는 그대로 유지됩니다.)

### 3. (참고) 실행은 유지하고 터미널만 빠져나오기

만약 앱을 계속 띄워둔 채로 터미널 명령어를 입력하고 싶다면, 실행할 때 **데몬(Background) 모드**를 사용해야 합니다.

* **실행:** `docker-compose up -d`
* **빠져나오기:** 이 모드로 실행하면 실행 즉시 터미널 입력창으로 돌아옵니다.
* **다시 끄고 싶을 때:** `docker-compose stop` 또는 `docker-compose down`을 입력하세요.

```yaml

version: '3.8'  # Docker Compose 파일 버전

services:  # 여러 컨테이너를 정의
  서비스이름:  # 서비스 이름(ex db, database 등 이름 정의)
    image: image_name  # 사용할 이미지
    build: ./path_to_dockerfile  # Dockerfile을 빌드하여 이미지 만들기
    ports:
      - "host_port:container_port"  # 포트 매핑
    environment:  # 환경 변수 설정
      - ENV_VAR_NAME=value
    volumes:  # 데이터 지속성을 위한 볼륨 매핑
      - ./host_path:/container_path
    networks:  # 네트워크 설정
      - network_name
    depends_on:  # 다른 서비스 의존성 정의
      - other_service


```yaml

version: '3.7'
services:
  nginx:
    image: nginx:latest
    container_name: nginx-container
    ports:
      - "8080:80"
    volumes:
      - ./nginx.conf:/etc/nginx/nginx.conf
    depends_on:
      - mysql
      # mysql 서비스가 실행된 후에 nginx 서비스가 시작되도록 지정

  mysql:
    image: mysql:5.7
    container_name: mysql-container
    environment:
      MYSQL_ROOT_PASSWORD: my123 # root 패스워드
      MYSQL_DATABASE: mydb  # database  없으면 생성
      MYSQL_USER: scott  # 계정 생성
      MYSQL_PASSWORD: tiger
    volumes:
      - .\init.sql:/docker-entrypoint-initdb.d/init.sql  # 초기 sql 실행
      - D:\dockerTest\dockercompose_nginx_mysql\mysql_data:/var/lib/mysql  # 컨테이저 재구동 시에도 데이터 유지
    ports:
      - "3306:3306"

volumes:
  mysql_data:
    driver: local  # 호스트 로컬에 저장


```bash

# /etc/nginx/nginx.conf
events {
    worker_connections 1024;
}

http {
    include       /etc/nginx/mime.types;
    default_type  application/octet-stream;
    access_log    /var/log/nginx/access.log;
    error_log     /var/log/nginx/error.log;

    # server 블록은 이 안에 위치해야 합니다.
    server {
        listen       80;
        server_name  localhost;

        location / {
            root   /usr/share/nginx/html;
            index  index.html index.htm;
        }

        # 추가적인 설정들...
    }
}

events: Nginx의 연결 처리 방식을 설정.
http: HTTP 관련 설정을 정의 (MIME 타입, 로그 파일 위치 등).
server: 하나의 Nginx 서버 인스턴스를 정의, 포트 및 도메인 설정, 그리고 요청 처리 방법을 설정 (예: location).


=====================================================================
events 블록: Nginx 서버가 연결을 처리하는 방법을 설정합니다.
worker_connections 1024;: 이 설정은 각 워커 프로세스가 동시에 처리할 수 있는 최대 연결 수를 설정합니다. 즉, 1024는 각 워커가 동시에 1024개의 연결을 처리할 수 있다는 의미입니다. 이 값은 트래픽이 많을 때 성능에 영향을 미칠 수 있습니다.


http {
    include       /etc/nginx/mime.types;
    default_type  application/octet-stream;
    access_log    /var/log/nginx/access.log;
    error_log     /var/log/nginx/error.log;

목적: http 블록은 HTTP 관련 설정을 정의합니다. Nginx는 이 블록 안에서 HTTP 요청을 처리하고, 여러 서버 및 리소스 설정을 포함합니다.
설정 항목:
include /etc/nginx/mime.types;: Nginx는 이 파일을 통해 MIME 타입을 설정합니다. MIME 타입은 서버가 클라이언트에게 어떤 종류의 파일을 보내는지 알려줍니다. 예를 들어, .html 파일은 text/html로, .jpg 파일은 image/jpeg로 전송됩니다.
default_type application/octet-stream;: 파일 유형을 알 수 없을 때 기본적으로 application/octet-stream MIME 타입을 사용합니다. 이는 바이너리 파일로 처리되는 기본 타입입니다.
access_log /var/log/nginx/access.log;: HTTP 요청에 대한 액세스 로그를 저장할 파일 경로입니다.
error_log /var/log/nginx/error.log;: 서버에서 발생한 오류를 기록하는 로그 파일 경로입니다.


server {
    listen       80;
    server_name  localhost;

    location / {
        root   /usr/share/nginx/html;
        index  index.html index.htm;
    }

    # 추가적인 설정들...
}

목적: server 블록은 Nginx 서버가 요청을 어떻게 처리할지 정의합니다. 하나의 Nginx 서버는 여러 server 블록을 가질 수 있으며, 각각 다른 호스트나 포트를 처리할 수 있습니다.

listen 80;: Nginx가 HTTP 요청을 받을 포트를 정의합니다. 여기서는 80 포트, 즉 기본 HTTP 포트에서 요청을 받도록 설정되어 있습니다.
server_name localhost;: 이 서버 블록에 적용되는 도메인 이름을 설정합니다. localhost는 로컬 환경에서만 유효합니다. 실제 서비스에서는 이곳에 실제 도메인 이름을 설정할 수 있습니다.
location / 블록:

location / { ... }: URL 경로에 대한 요청을 처리합니다. 이 블록 내에서 / 경로에 대한 요청을 처리하는 방법을 정의합니다. 여기서는 루트 URL (/)로 들어오는 모든 요청을 처리합니다.
root /usr/share/nginx/html;: Nginx가 요청한 파일을 찾을 디렉터리 경로입니다. 이 경로에서 웹 서버가 클라이언트에게 반환할 파일을 찾습니다.
index index.html index.htm;: 요청된 URL이 디렉터리인 경우, 해당 디렉터리에서 기본적으로 반환할 파일을 설정합니다. 예를 들어, / 경로로 들어오는 요청은 index.html 또는 index.htm 파일을 찾아 반환합니다.



```yaml

version: '3.8'

services:
  mysql:
    image: mysql:8
    environment:
      MYSQL_ROOT_PASSWORD: rootpassword
      MYSQL_DATABASE: product_db
    ports:
      - "3306:3306"
    volumes:
      - mysql-data:/var/lib/mysql
    networks:
      - mysql-network
    restart: always

  app:
    build: ./app
    environment:
      DATABASE_URL: "mysql://root:rootpassword@mysql:3306/product_db"
    ports:
      - "5000:5000"
    depends_on:
      - mysql
    networks:
      - mysql-network
    restart: on-failure

volumes:
  mysql-data:

networks:
  mysql-network:
    driver: bridge


```yaml

version: '3.8': Docker Compose 파일 버전
mysql: MySQL 서비스 설정, 환경 변수로 MYSQL_ROOT_PASSWORD와 MYSQL_DATABASE를 설정
app: Python 애플리케이션 서비스, DATABASE_URL 환경 변수를 통해 MySQL에 연결
depends_on: app 서비스가 mysql 서비스에 의존한다는 설정
volumes: mysql-data라는 Docker 볼륨을 정의하여 데이터 지속성 제공
networks: mysql-network라는 네트워크를 정의하고, 두 서비스가 이를 공유하도록 설정
각 옵션의 선택적인 사용 예시
build: Dockerfile을 기반으로 이미지를 빌드하고 컨테이너를 실행할 때 유용합니다.
ports: 특정 포트를 외부와 연결하여 서비스를 외부에서 접근할 수 있게 만듭니다.
environment: 환경 변수를 설정하여 애플리케이션의 설정을 동적으로 변경할 수 있습니다.
volumes: 데이터 지속성을 보장하기 위해 데이터를 호스트 시스템과 컨테이너 간에 매핑합니다.
depends_on: 서비스가 다른 서비스가 시작된 후에 실행되도록 순서를 지정합니다.
